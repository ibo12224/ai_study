```python
import ccxt
import pandas as pd
import time
from datetime import datetime
```

```python
# 사용할 거래소 인스턴스 생성 (여기서는 업비트를 사용)
symbol = 'BTC/KRW'    # 가져올 코인 쌍 (비트코인/원화)
timeframe = '1d'      # 데이터 간격 (1d: 일봉, 1h: 1시간봉, 1m: 1분봉)
limit = 0       # 가져올 데이터 개수 (최근 30개)
```

```python
def get_coin(coin,time,limit=10,exc=ccxt.upbit()):
    print(f"[{exc.id}] 거래소에서 [{coin}]의 [{time}] 데이터 {limit}개를 가져옵니다.")
    try:
        ohlcv = exc.fetch_ohlcv(coin, timeframe=time,limit=10000)
        
    except Exception as e:
        print(f"데이터를 가져오는 중 오류 발생: {e}")
        ohlcv = []
    if ohlcv:
        # DataFrame 생성 (컬럼명: 타임스탬프, 시가, 고가, 저가, 종가, 거래량)
        df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
        df['date'] = pd.to_datetime(df['timestamp'], unit='ms')
        df = df.drop('timestamp', axis=1)

        df = df[['date', 'open', 'high', 'low', 'close', 'volume']]
        df.set_index('date',inplace=True)
        df.set_flags=coin
        print(df)
        
    else:
        print("\n데이터를 가져오지 못했습니다.")
    return df
```

```python
# --- 2. OHLCV 데이터 가져오기 ---
df1=get_coin(symbol,timeframe,limit)
df2=get_coin('ETH/KRW',timeframe,limit)
# df3=get_coin(symbol,timeframe,limit)
```
