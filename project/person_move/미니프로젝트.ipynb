```python
import FinanceDataReader as fdr
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import matplotlib
matplotlib.rcParams['font.family']='NanumGothic'
plt.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
xrp = fdr.DataReader('XRP/KRW','2018-01-01','2025-12-01').round(0).astype('Int64').drop(columns=['Adj Close'])
btc = fdr.DataReader('BTC/KRW','2018-01-01','2025-12-01').round(0).astype('Int64').drop(columns=['Adj Close'])
eth = fdr.DataReader('ETH/KRW','2018-01-01','2025-12-01').round(0).astype('Int64').drop(columns=['Adj Close'])
inx = fdr.DataReader('IXIC','2018-01-01','2025-12-01').round(0).astype('Int64').drop(columns=['Adj Close'])
# dxy = fdr.DataReader('DX-Y.NYB','2018-01-01','2025-12-01').round(0).astype('Int64').drop(columns=['Adj Close'])
```

```python
plt.plot(btc.index, btc['Close'])
plt.show()
```

```python
plt.plot(xrp.index, xrp['Close'])
plt.show()
```

```python
plt.plot(eth.index, eth['Close'])
plt.show()
```

```python
fig, axes = plt.subplots()
axes.set_xlabel( 'x')
axes.set_ylabel( '',color='b')
plt.plot(btc.index, btc['Close'],color='#ffd700')
axes.plot()

ax2 = axes.twinx()
ax2.set_ylabel( '',color='r')
plt.plot(inx.index, inx['Close'],'r--')
plt.show()
```

```python
btcc = btc[['Close']].rename(columns={'Close':''})
ethc = eth[['Close']].rename(columns={'Close':''})
xrpc = xrp[['Close']].rename(columns={'Close':''})
inxc = inx[['Close']].rename(columns={'Close':''})
result = pd.concat([inxc,btcc,ethc,xrpc], axis=1)
result
```

```python
print(result.corr())
sb.heatmap(result.corr(),vmin=-1,vmax=1,annot=True,cmap='seismic',linewidths=0.2)
plt.show()
```

```python
plt.boxplot([btc['Volume'],inx['Volume'],xrp['Volume'],eth['Volume']],tick_labels=['','','',''])
plt.show()
```
