```python
from bs4 import BeautifulSoup
import requests
import warnings
warnings.simplefilter('ignore')
# 제목 title,
# 예매율 reservation_rate ,
# 누적 관객수 total_audience ,
# 장르 genre ,제한 rating, 
# 감독 director
```

```python
url='http://moviechart.co.kr/rank/realtime/index/image'
response=requests.get(url)
html=BeautifulSoup(response.content,'html.parser')
html
```

```python
href=[]
for a in html.select('.movieBox-item>a'):
    print(a['href'])
    href.append(a['href'])
```

```python
list=[]
for h in href:
    baseurl='https://www.moviechart.co.kr/'
    response=requests.get(baseurl+h)
    html=BeautifulSoup(response.content,'html.parser')
    title=html.select_one('.movieIner__title>div').contents[0].strip()
    reservation_rate=html.select_one('dt').string
    total_audience=html.select_one('dl:nth-of-type(2) > dt')
    if(total_audience==None):
        total_audience='0명'
    else:
        total_audience=total_audience.string
    mv=html.select_one('li:nth-of-type(2) dt').string
    rating=mv.split('/')[-1]
    director=html.select_one('li:nth-of-type(3) dt').string

    reservation_rate=float(reservation_rate.split(' ')[1].replace('%',''))
    print(reservation_rate)
    total_audience = int(total_audience.replace('명','').replace(',',''))
    print(total_audience)
    list.append((title,reservation_rate,total_audience,rating,director))
    print(title)
# 제목 title,
# 예매율 reservation_rate ,
# 누적 관객수 total_audience ,
# 장르 genre ,제한 rating, 
# 감독 director
```
