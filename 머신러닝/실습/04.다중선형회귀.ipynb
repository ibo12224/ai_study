------------------------------------------
1. w(행,열): [컬럼의 개수, 라벨의 갯수]
2. 특성데이터: 행x열(컬럼의갯수)
3. 라벨데이터: 행x열(대부분1열)
4. 예측: 행x열

```python
df = pd.read_csv('data/electric.csv',index_col='Unnamed: 0')
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,[-1]].values
model = LinearRegression()
model.fit( x, y)
```

```python
model.predict([[4],[5]])
```

```python
plt.scatter(x,y)
plt.plot(x,model.coef_*x+model.intercept_)
plt.show()
```
