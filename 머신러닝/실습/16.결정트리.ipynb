#  불순도 측도 : 지니계수(Gini Index)와 엔트로피(Entropy)

---

## 1. 불순도(Impurity)란?
  - **불순도** = 데이터가 섞여 있는 정도`
  - 한 그룹에 **하나의 클래스만 있으면** → 불순도 **0** (완전히 깨끗)`
- 여러 클래스가 **비슷한 비율로 섞이면** → 불순도 **높음** (더럽게 섞임)
- 결정트리는 **불순도를 줄이는 방향**으로 데이터를 나눔

 **비유**:  
주머니 속에 **빨간 공**만 있으면 색깔 맞히기 100% → 불순도 0  
빨간 공과 파란 공이 반반이면 맞히기 어려움 → 불순도 높음

---

## 2. 지니계수 (Gini Index)
**정의**:
$
\text{Gini} = 1 - \sum_{i=1}^{k} p_i^2
$
- $(p_i$): 클래스 $(i$)의 비율
- 의미: **주머니에서 공 2개를 랜덤으로 뽑았을 때, 색이 다를 확률**

**특징**:
- 한 색만 있으면 Gini = 0 (불순도 없음)
- 색이 섞여 있으면 값이 커짐 (최대 0.5, 2클래스일 경우)

**예시**:
- 빨간 100%, 파란 0% → \( Gini = 0 \)
- 빨간 50%, 파란 50% →  
  $
  Gini = 1 - (0.5^2 + 0.5^2) = 0.5
  $

---

## 3. 엔트로피 (Entropy)
**정의**:
$
\text{Entropy} = - \sum_{i=1}^{k} p_i \log_2 p_i
$
- 의미: **정보의 불확실성**을 나타내는 지표
- 값이 클수록 "다음에 나올 클래스 맞히기 어려움"

**특징**:
- 한 색만 있으면 Entropy = 0 (불확실성 없음)
- 모든 색이 비슷한 비율로 있으면 Entropy 최대 (2클래스일 때 최대 1)

**예시**:
- 빨간 100%, 파란 0% → Entropy = 0
- 빨간 50%, 파란 50% →  
  $
  Entropy = -[0.5\log_2 0.5 + 0.5\log_2 0.5] = 1
  $

---

## 4. 지니계수 vs 엔트로피

| 지표       | 최대값 (2클래스) | 해석 방식                | 특징              |
|------------|----------------|--------------------------|-------------------|
| 지니계수   | 0.5            | 잘못 분류될 확률          | 계산이 더 단순    |
| 엔트로피   | 1              | 정보량(불확실성)          | 정보이론 기반     |

---

## 5. 간단한 비유
- **지니계수**: "주머니에서 공 2개 뽑았을 때, 색이 다를 확률"
- **엔트로피**: "주머니 속 공 색깔을 맞히기 어려운 정도"

---

## 6. 한 줄 요약
> 지니계수와 엔트로피는 **데이터가 얼마나 섞여 있는지(불순도)**를 재는 척도이며,  
> 결정트리는 이를 줄이는 방향으로 분할을 진행한다.

#  지니지수(Gini Index) 
---

## 1. 공식
$
\text{Gini} = 1 - \sum_{i=1}^{k} p_i^2
$
- \(k\) : 클래스(종류) 개수  
- \(p_i\) : 클래스 \(i\)의 비율 (0~1 사이 값)

---

## 2. 의미
- **"무작위로 2개를 뽑았을 때, 서로 다른 클래스일 확률"**
- 한 클래스만 있으면 Gini = 0 (**완전히 깨끗**)
- 여러 클래스가 비슷한 비율로 섞이면 Gini 값이 커짐 (**많이 섞임**)

---

## 3. 계산 과정
1. 각 클래스의 비율 \(p_i\)를 **제곱**한다 → 같은 클래스일 확률
2. 그 값을 전부 더한다 → 전체가 같은 클래스일 확률
3. **1에서 빼기** → 다른 클래스일 확률

---

## 4. 예시
주머니 속 공:
- 빨간 50% (0.5)
- 파란 50% (0.5)

계산:
$
\text{Gini} = 1 - (0.5^2 + 0.5^2)  
= 1 - (0.25 + 0.25)  
= 1 - 0.5 = 0.5
$

---

## 5. 직관적으로 보기

| 클래스 비율 | 계산식                               | Gini 값 |
|-------------|-------------------------------------|---------|
| 100% / 0%   | \(1 - (1^2 + 0^2) = 0\)              | 0       |
| 50% / 50%   | \(1 - (0.5^2 + 0.5^2) = 0.5\)        | 0.5     |
| 70% / 30%   | \(1 - (0.7^2 + 0.3^2) = 0.42\)       | 0.42    |

---

## 6. 한 줄 요약
> 지니지수는 **데이터가 얼마나 섞여 있는지**를 수치로 표현한 값이며,  
> 0이면 완전히 한 종류, 값이 커질수록 클래스가 고르게 섞여 있다는 뜻이다.


---

## 3. 엔트로피 (Entropy)
**정의**:
$
\text{Entropy} = - \sum_{i=1}^{k} p_i \log_2 p_i
$
- **의미**: 불확실성(정보량)을 나타내는 지표  
- **값의 특징**:
  - 한 클래스만 있으면 Entropy = 0
  - 모든 클래스 비율이 동일하면 Entropy 최대

**예시**:
- A: 100%, B: 0% → \( Entropy = 0 \)
- A: 50%, B: 50% →  
  $
  Entropy = -[0.5 \log_2 0.5 + 0.5 \log_2 0.5] = 1
  $

---

#  엔트로피(Entropy) 

---

## 1. 공식
$
\text{Entropy} = - \sum_{i=1}^{k} p_i \log_2 p_i
$
- \(k\) : 클래스(종류) 개수  
- \(p_i\) : 클래스 \(i\)의 비율 (0~1 사이 값)

---

## 2. 의미
- **"데이터의 불확실성(혼란도)"를 나타내는 값**
- 한 클래스만 있으면 Entropy = 0 (**확실함, 예측 쉬움**)
- 모든 클래스 비율이 같으면 Entropy 최대 (**예측 어렵고 불확실함**)

---

## 3. 계산 과정
1. 각 클래스 비율 $(p_i$)에 대해 $(p_i \log_2 p_i$) 계산
2. 모두 더한 뒤, **마이너스(-)**를 붙인다  
   → 불확실성이 양수 값으로 표현됨

---

## 4. 예시
주머니 속 공:
- 빨간 50% (0.5)
- 파란 50% (0.5)

계산:
$
\text{Entropy} = - [0.5 \log_2 0.5 + 0.5 \log_2 0.5]  
= - [0.5 \times (-1) + 0.5 \times (-1)]  
= - [-0.5 - 0.5] = 1
$

---

## 5. 직관적으로 보기

| 클래스 비율 | 계산식                                                                | Entropy 값 |
|-------------|-----------------------------------------------------------------------|------------|
| 100% / 0%   | $(-[1 \log_2 1 + 0 \log_2 0] = 0$)                                     | 0          |
| 50% / 50%   | $(-[0.5 \log_2 0.5 + 0.5 \log_2 0.5] = 1$)                             | 1          |
| 70% / 30%   | $(-[0.7 \log_2 0.7 + 0.3 \log_2 0.3] \approx 0.88$)                    | 0.88       |

---

## 6. 지니지수와 비교
| 지표       | 최대값 (2클래스) | 해석 방식              | 특징               |
|------------|----------------|-----------------------|--------------------|
| 지니계수   | 0.5            | 잘못 분류될 확률       | 계산이 단순        |
| 엔트로피   | 1              | 정보량(불확실성)       | 정보이론 기반      |

---

## 7. 한 줄 요약
> 엔트로피는 데이터가 **얼마나 불확실하게 섞여 있는지**를 나타내며,  
> 값이 0이면 완전히 한 종류, 값이 클수록 예측이 어려운 상태이다.



## 4. 비교 표

### Decision Tree Classifier의 파라미터

#### 핵심 파라미터

##### min_samples_split	: 노드를 분할하기 위한 최소한의 샘플 데이터수 
( 과적합을 제어하는데 사용 )
- Default = 2 → 작게 설정할 수록 분할 노드가 많아져 과적합 가능성 증가<br>
##### min_samples_leaf: 리프노드가 되기 위해 필요한 최소한의 샘플 데이터수
- min_samples_split과 함께 과적합 제어 용도<br>
- 불균형 데이터의 경우 특정 클래스의 데이터가 극도로 작을 수 있으므로 작게 설정 필요<br>
##### max_features: 최적의 분할을 위해 고려할 최대 feature 개수<
- Default = None → 데이터 세트의 모든 피처를 사용<br>
- int형으로 지정 →피처 갯수 / float형으로 지정 →비중<br>
- sqrt 또는 auto : 전체 피처 중 √(피처개수) 만큼 선정<br>
- log : 전체 피처 중 log2(전체 피처 개수) 만큼 선정<br>
##### max_depth	: 트리의 최대 깊이<br>
- default = None<br>
→ 완벽하게 클래스 값이 결정될 때 까지 분할<br>
또는 데이터 개수가 min_samples_split보다 작아질 때까지 분할<br>
- 깊이가 깊어지면 과적합될 수 있으므로 적절히 제어 필요<br>
##### max_leaf_nodes:	리프노드의 최대 개수<br>

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
iris = load_iris(as_frame=True)
iris
```

```python
x_data = iris['data']
y_data = iris['target']
```

```python
model = DecisionTreeClassifier()
model.fit( x_data, y_data)
```

```python
model.tree_.impurity #불순도 지표
```

```python
model.predict( [[5.1,3.5,1.4,0.2]] )
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
pred = model.predict( x_data)
pred
```

```python
accuracy_score(y_data, pred )
```

```python
precision_score( y_data, pred, average='macro')
```

```python
recall_score(  y_data, pred, average='macro')
```

```python
f1_score(  y_data, pred, average='macro')
```

```python
confusion_matrix( y_data, pred)
```

##  트리의 복잡도(깊이)를 조절하여 **과적합(Overfitting)을 제어**하는 데 사용.

---

### 1.  분할 조건 제어 파라미터 (가지치기: Pruning)이 파라미터들은 노드가 더 이상 분할되는 것을 막고, 트리의 깊이가 무한정 깊어지는 것을 방지하여 과적합을 제어합니다.

| 파라미터 | 역할 및 설명 | 기본값 | 주요 사용 목적 |
| --- | --- | --- | --- |
| **`min_samples_split`** | **노드를 분할하는 데 필요한 최소 샘플 수** | 2 | 이 값보다 샘플이 적은 노드는 더 이상 분할되지 않습니다. **값이 클수록** 트리가 덜 복잡해져 **과적합을 방지**합니다. |
| **`min_samples_leaf`** | **리프 노드가 되기 위한 최소 샘플 수** | 1 | 분할 후 생성된 자식 노드(리프 노드)가 이 값보다 샘플이 적으면 분할하지 않습니다. `min_samples_split`과 함께 과적합 제어에 매우 중요합니다. |
| **`max_depth`** | **트리의 최대 깊이** | None | 트리가 가질 수 있는 최대 깊이를 제한합니다. **값이 낮을수록** 트리가 얕아져 일반화됩니다. (`None`은 완벽한 분할이 될 때까지 분할한다는 의미입니다.) |
| **`max_leaf_nodes`** | **리프 노드의 최대 개수** | None | 전체 리프 노드의 총개수를 제한하여 트리의 크기를 직접적으로 제어합니다. |

---

### 2.  분할 Feature 선택 제어 파라미터트리가 분할될 때, 모든 Feature(특성)를 고려할지 아니면 일부 Feature만 무작위로 고려할지를 결정하여 계산 효율성을 높이고 과적합을 방지합니다.

| 파라미터 | 역할 및 설명 | 기본값 | 주요 사용 목적 |
| --- | --- | --- | --- |
| **`max_features`** | **최적의 분할을 위해 고려할 최대 Feature 개수** | None | 분할 시 매번 Feature 전체(N)를 사용하는 대신, 일부(K)만 고려하도록 제한합니다. **랜덤 포레스트** 같은 앙상블 기법에서 핵심적으로 사용됩니다. |
| **옵션** | **설명** |  |  |
| `int` 형 | 직접 고려할 **Feature의 개수** 지정 |  |  |
| `float` 형 | 전체 Feature 개수 대비 **비율** 지정 |  |  |
| `sqrt`, `auto` | \sqrt{N} (전체 Feature 개수의 제곱근)만큼 선정 |  |  |
| `log` | \log_2(N) (전체 Feature 개수의 이진 로그)만큼 선정 |  |  |

---

### 파라미터 설정의 일반적인 원칙 (과적합 제어)
| 문제 상황 | 해결 방법 (파라미터 조절) |
| --- | --- |
| **과적합 (Overfitting)** | 트리가 너무 복잡해지고 깊어짐 |
| **과소적합 (Underfitting)** | 트리가 너무 단순하여 학습이 충분하지 않음 |

**핵심은** `min_samples_split`, `min_samples_leaf`, `max_depth` 세 가지를 조정하여 트리의 크기와 깊이를 제어하는 것입니다.
