#  주성분 분석 (Principal Component Analysis, PCA)

주성분 분석(PCA)은 **데이터의 차원을 줄이면서도 정보 손실을 최소화**하는 통계적 기법입니다. 쉽게 말해, 데이터가 가진 가장 중요한 특징(정보)만 뽑아서 **요약**하는 '데이터 다이어트 기술'이라고 할 수 있습니다.

---

### 1. PCA가 필요한 이유: 

'차원의 저주' 우리가 분석하는 데이터에는 보통 여러 개의 **변수(특성, Feature)**가 있습니다. 예를 들어, 꽃의 종류를 구분할 때 '꽃잎 길이', '꽃잎 너비', '꽃받침 길이', '꽃받침 너비' 같은 4가지 변수를 사용할 수 있습니다.

* **문제점:** 변수의 개수(차원)가 너무 많으면 다음과 같은 문제가 생깁니다.
* **분석이 어려워짐:** 시각화가 거의 불가능하고, 해석이 복잡해집니다.
* **모델 성능 저하:** 불필요한/중복된 정보로 인해 분석 모델의 성능이 떨어지거나 과적합(Overfitting) 위험이 커집니다. (이것을 **차원의 저주**라고 부릅니다.)



> **핵심:** PCA는 이 많은 변수들 중 **서로 중복되거나 유사한 정보를 담고 있는 변수**들을 찾아내어, 그 정보를 함축하는 새로운 '축(주성분)'을 만들어서 차원을 줄입니다.

---

### 2. PCA의 작동 원리: 

새로운 축 찾기 PCA는 원래 데이터의 분산(정보량)을 가장 잘 설명하는 새로운 축(주성분)을 찾습니다.

1. **변수들의 상관관계 파악:** 먼저 변수들 간의 관계를 분석합니다.
2. **새로운 축(주성분) 생성:**
* **첫 번째 주성분 (PC1):** 데이터의 분산(정보)을 가장 많이 설명할 수 있는 방향으로 새로운 축을 설정합니다. (데이터를 이 축에 정사영(Projection)했을 때 분산이 가장 크게 되는 축입니다.)
* **두 번째 주성분 (PC2):** PC1과는 직각(수직, 직교)이면서, PC1이 설명하지 못한 나머지 분산을 가장 많이 설명하는 방향으로 축을 설정합니다.
* 이 과정을 반복하여 데이터 차원의 개수만큼 주성분(PC1, PC2, PC3, ...)을 만듭니다.

3. **차원 축소:** 생성된 주성분들 중에서 **분산(정보량)을 많이 설명하는 순서대로** 필요한 개수(예: 2개)만 선택하고 나머지는 버립니다.
* 예를 들어, 10개의 변수(차원)를 가진 데이터를 PCA를 통해 PC1과 PC2, 단 2개의 새로운 축으로 압축하여 시각화나 분석에 사용합니다.


<img src='image/pca.png'>


> **주성분의 특징:** 각 주성분($PC_i$)은 원래 변수들의 선형 결합($PC_i = a_1 X_1 + a_2 X_2 + \dots$)으로 표현되며, **주성분들끼리는 서로 독립적**입니다.

---

### 3. PCA의 주요 활용 분야* 

**데이터 시각화:** 고차원 데이터를 2차원 또는 3차원으로 줄여서 산점도(Scatter Plot) 등으로 표현하여 데이터의 구조를 한눈에 파악할 수 있게 합니다.
* **이미지 압축:** 이미지의 중요한 특징을 유지하면서 크기를 줄이는 데 사용됩니다.
* **노이즈 제거:** 데이터에서 불필요한 노이즈나 중복된 정보를 제거하여 분석의 정확도를 높입니다.

---



## “분산이 가장 큰 방향”

### 1. 한 줄 요약

**데이터가 가장 넓게 퍼져 있는 방향**,  
즉 **정보(차이)가 가장 많이 드러나는 방향**을 말합니다.

---

### 2. 직관적인 비유 (사람 키 예시)

여러 사람의 **키** 데이터를 생각해봅시다.

- 키가 거의 비슷하면 → 데이터가 한 점 근처에 모여 있음  
- 키 차이가 크면 → 데이터가 넓게 퍼져 있음  

이때,

- **키 방향으로 보면** 사람들 차이가 잘 보임  
- **아무 의미 없는 방향(예: 사람 번호)**으로 보면 차이가 거의 안 보임  

 **차이가 가장 잘 보이는 방향 = 분산이 가장 큰 방향**

---

### 3. 2차원 예시로 이해하기 (시험 점수)

데이터가 이렇게 있다고 가정합시다.

- $x$축: 수학 점수  
- $y$축: 영어 점수  

학생들이 대각선 방향으로 길게 늘어서 있다면:

- $x$축만 보면 → 정보 일부만 보임  
- $y$축만 보면 → 정보 일부만 보임  
- **대각선 방향으로 보면 → 점수 차이가 가장 잘 보임**

이 **대각선 방향**이 바로  
 **분산이 가장 큰 방향**

---

### 4. 왜 “정보가 가장 많다”고 하나?

분산의 의미는 다음과 같습니다.

$
\text{분산} = \text{데이터가 평균에서 얼마나 떨어져 있는가}
$

- 분산이 크다 → 값들이 다양하다  
- 분산이 작다 → 값들이 비슷하다  

따라서,

> **분산이 가장 큰 방향 = 데이터의 차이를 가장 잘 설명하는 방향**

---

### 5. PCA와 연결하면

PCA에서:

- **첫 번째 주성분(PC1)**  
  → 분산이 **가장 큰 방향**
- **두 번째 주성분(PC2)**  
  → PC1과 직교하면서 그다음으로 분산이 큰 방향

즉,

> PCA는  
> **“데이터를 가장 잘 설명하는 방향부터 차례대로 찾는 방법”**

---

### 6. 정리

> **분산이 가장 큰 방향이란,  
> 데이터를 바라봤을 때 값들의 차이가 가장 크게 드러나는 방향이다.**



## 주성분 분석 (PCA) 동작 원리 

PCA의 목표는 **데이터의 분산(정보)을 가장 잘 보존**하면서 차원을 줄이는 것입니다. 이를 위해 데이터의 분산이 최대가 되는 새로운 축을 찾습니다.

### 1. 데이터 준비: 

표준화 (Standardization)PCA를 시작하기 전, 모든 변수(특성)의 스케일을 통일합니다.

* **원리:** PCA는 분산이 큰 축에 가중치를 높게 두기 때문에, 단위가 다른 변수가 있으면 분산이 큰 변수가 불공평하게 더 중요하게 취급될 수 있습니다.
* **작동:** 각 변수의 평균을 0, 표준편차를 1로 맞춥니다 (표준화).

### 2. 핵심 단계: 

분산이 최대인 축 찾기PCA의 핵심은 데이터를 **정사영(Projection)**했을 때 **분산이 가장 크게** 나타나는 새로운 축을 찾는 것입니다.

* **정보 = 분산:** 통계학적으로 분산이 크다는 것은 데이터가 넓게 퍼져 있다는 뜻이며, 이는 곧 **정보량(패턴, 특징)**이 많다는 것을 의미합니다.
* **작동:**

1. **첫 번째 주성분 (PC1) 찾기:** 데이터의 분산을 최대로 만들 수 있는 방향으로 첫 번째 새로운 축(PC_1)을 설정합니다. 이 축이 데이터의 정보를 가장 많이 담게 됩니다.

2. **두 번째 주성분 (PC2) 찾기:** PC_1과는 **직교(수직)**하면서, PC_1이 설명하지 못한 **남은 분산**을 최대로 설명할 수 있는 두 번째 축(PC_2)을 설정합니다.

3. 이 과정을 반복하여 데이터의 차원 수만큼 주성분을 생성합니다.



### 3. 수학적 도구: 

공분산 행렬, 고유 벡터/값이 '분산이 최대인 축'을 수학적으로 찾는 도구입니다.

| 도구 | 역할 |
| --- | --- |
| **공분산 행렬** | 변수들 간의 관계(상관성)와 데이터의 전체적인 분산 구조를 파악합니다. |
| **고유 벡터** | 분산이 최대가 되는 **방향**, 즉 우리가 찾는 **주성분 축(PC_i)**이 됩니다. |
| **고유 값** | 고유 벡터(주성분 축) 방향으로 투영했을 때의 **분산의 크기**입니다. 즉, 해당 주성분이 담고 있는 **정보량**입니다. |

### 4. 차원 축소 및 선택주성분을 모두 생성한 후, 실제로 사용할 주성분의 개수를 결정.

* **선택 기준:** 주성분들이 가진 고유 값(분산 설명력)이 큰 순서대로 정렬합니다.
* **작동:**

1. **정보량이 많은** (고유 값이 큰) 상위 k개의 주성분만 선택합니다.
2. 이 k개의 주성분을 사용하여 원본 데이터를 새로운 저차원 공간에 투영(Mapping)합니다.  (예: 4차원 $\rightarrow$ 2차원)

---

### 요약 정리PCA는 표준화된 데이터를 기반으로 **'분산을 최대로 보존하는 축'** (고유 벡터)을 찾고, 그 축이 담고 있는 **정보량** (고유 값)을 기준으로 필요한 개수만 선택하여 차원을 줄이는 기술.


## 회귀선과 비교

---

## 1. 공통점

- 둘 다 **데이터의 전반적인 방향**을 나타냄
- 산점도에서 **대각선 형태의 직선**으로 보이는 경우가 많음
- 데이터가 길게 늘어진 방향을 “대표”하는 선처럼 보임

그래서 시각적으로는 거의 같은 선처럼 느껴질 수 있다.

---

## 2. 핵심 차이점 (아주 중요)

| 구분 | PCA의 주성분(분산 최대 방향) | 회귀선(Regression Line) |
|---|---|---|
| 목적 | 데이터 **구조 요약** | **예측** |
| 기준 | **전체 분산 최대화** | **오차 최소화** |
| 오차 방향 | 선에 **수직 거리** | **y축 방향 거리** |
| 종속변수 | 없음 | 있음 (y) |
| 대칭성 | x, y 대칭 | x → y 방향성 있음 |

---

## 3. 가장 중요한 차이: “오차를 어디서 재느냐”

### (1) 회귀선
- 목표:  
  $
  \min \sum (y_i - \hat{y}_i)^2
  $
- **세로 거리(y축 방향)**만 최소화
- “x로 y를 예측”

 그래서 **x와 y를 바꾸면 회귀선이 달라진다**

---

### (2) PCA 주성분
- 목표:  
  $
  \max \text{Var(투영된 데이터)}
  $
- 선에 대한 **직교 거리(수직 거리)** 기준
- “데이터 전체를 가장 잘 설명하는 방향”

 그래서 **x와 y를 바꿔도 동일한 주성분 방향**

---

## 4. 언제 거의 같아질까?

다음 조건에서는 **회귀선 ≈ PCA 첫 주성분**이 됩니다.

- x, y가 **강한 선형 관계**
- 노이즈가 작음
- x와 y의 **스케일이 비슷함**
- 평균이 제거(중심화)된 경우

이때는 두 선이 거의 겹쳐 보입니다.

---

## 5. 직관적 한 문장 정리

- **회귀선:**  
  > “x를 알면 y를 얼마나 잘 맞출 수 있을까?”
- **PCA 방향:**  
  > “데이터가 전체적으로 어느 방향으로 퍼져 있을까?”

---

## 6. 핵심 문장

> “회귀는 예측의 문제이고, PCA는 표현의 문제다.  
> 선은 비슷해 보여도 목적이 다르다.”



## 결론

**“그 선에 데이터를 투영했을 때의 분산”이 가장 큰 선**

---

## 1. 왜 그냥 ‘선의 분산’이 아니라 “투영된 분산”일까?

임의의 직선 하나를 그었다고 가정합시다.

- 각 데이터 점을 **그 선에 수직으로 내려서 붙입니다**  
- 그러면 선 위에 1차원 데이터가 생깁니다

이때 계산하는 것이:

$
\text{Var}(\text{선 위로 투영된 값})
$

이 값이 **가장 커지도록 선택된 선**이  
**PCA의 첫 번째 주성분(분산이 가장 큰 방향)**입니다.

---

## 2. 중요한 오해 정리

❌ “선 자체에서 점까지의 거리 분산이 가장 큰 선”  
❌ “데이터와 선 사이 거리가 큰 선”

⭕ **“선 위에 놓인 데이터 값들의 퍼짐이 가장 큰 선”**
---

## 3. 직관적으로 다시 말하면

- 선을 회전시켜 가며 본다고 생각하면
- 어떤 방향에서는 점들이 선 위에 **길게 늘어져 보이고**
- 어떤 방향에서는 **뭉쳐 보입니다**

 **가장 길게 늘어져 보이는 방향**  
= **투영했을 때 분산이 최대인 방향**

## 2차원 -> 1차원 차원 축소 과정

### 1. 원본 데이터 (2 Dimensions)이미지의 왼쪽은 원본 데이터의 상태

* **축 (Features):** 데이터가 두 개의 축, 즉 **2개의 변수(차원)**로 표현되어 있습니다.
* `X축`: 집의 **크기 (size)**
* `Y축`: **방의 개수 (number of rooms)**


* **데이터 포인트:** 각 집(데이터 포인트)은 크기와 방의 개수라는 두 가지 정보를 동시에 가지고 있습니다.
* **관찰:** 그림을 보면, 집의 크기가 커질수록 방의 개수도 많아지는 **강한 상관관계**가 있을 것으로 보입니다. 즉, 두 변수가 **겹치는 정보**를 많이 가지고 있습니다.

> **PCA의 목표:** 겹치거나 중복된 정보를 제거하고, 가장 중요한 정보만 남겨서 하나의 새로운 축으로 요약하는 것입니다.

### 2. PCA를 통한 최적의 축 찾기PCA는 2차원 공간에서 데이터의 **분산(공분산정보)**이 가장 크게 퍼져 있는 방향을 찾습니다. 

이 방향이 새로운 축, 즉 **주성분 1 (PC1)**이 됩니다.

* **PC1의 의미:** 이 최적의 축(PC1)은 '크기'와 '방의 개수'라는 두 변수의 정보를 가장 효율적으로 결합한 **새로운 종합적인 특징**을 나타냅니다. 예를 들어, **'집의 전반적인 규모'**와 같은 의미를 가질 수 있습니다.

### 3. 차원 축소 (1 Dimension)데이터를 찾아낸 새로운 축(PC1) 위에 **정사영(Projection)** 시킵니다.

* **정사영:** 2차원 데이터 포인트들을 PC1 축 위로 **수직으로 내리꽂아(그림자처럼)** 1차원의 점으로 만듭니다.
* **결과:** 모든 2차원 집들이 이제 단 하나의 축 위에 있는 점들로 표현됩니다. 이것이 바로 이미지의 오른쪽 그림입니다.

### 4. 축소된 데이터 (1 Dimension)이미지의 오른쪽은 차원 축소의 결과를 보여줍니다.

* **새로운 축 (Feature):** 이제 데이터는 **'size feature'**라는 단 하나의 축(1차원)으로만 표현됩니다. 여기서 'size feature'는 PCA를 통해 만들어진 **주성분(PC1)**을 의미합니다.
* **정보 보존:** 원본 2차원 데이터의 정보(각 집의 상대적인 순서, 즉 '규모'의 순서)는 1차원 축 위에서도 **대부분 그대로 유지**됩니다. 즉, 가장 작은 집은 1차원 축에서도 왼쪽에, 가장 큰 집은 오른쪽에 위치하게 됩니다.

---

### 절차 요약

| 단계 | PCA 작업 내용 | 그림에서의 의미 |
| --- | --- | --- |
| **1. 최적 축 발견** | 데이터의 분산이 최대가 되는 새로운 축(PC1)을 찾습니다. | 2차원 산점도 사이를 대각선으로 가로지르는 최적의 직선을 찾는 것. |
| **2. 데이터 정사영** | 원본 데이터 포인트들을 이 새로운 축(PC1) 위로 투영(Projection)합니다. | 2차원 집들을 직선(PC1) 위로 수직으로 내려 점을 찍는 것. |
| **3. 차원 축소** | PC1 축만 남기고, PC1과 직교하는 나머지 축(정보량이 적은 축)은 제거합니다. | 2차원 평면을 없애고 1차원 직선만 남기는 것. |


````python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sb

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

iris = load_iris(as_frame=True)
X = iris.data.values
y = iris.target.values

iris_df = iris['frame']
scaler = StandardScaler()
x_scale = scaler.fit_transform(X)

# 2개의 주성분:2차원
pca = PCA( n_components=2)
X_pca = pca.fit_transform( x_scale)
pca_df = pd.DataFrame( X_pca, columns=['pc1','pc2'])
pca_df['target'] = y
pca_df
````

````python
sb.scatterplot( pca_df, x='pc1', y='pc2', hue='target')
plt.show()
````

````python
pca.explained_variance_ratio_
````

````python
sum( pca.explained_variance_ratio_ )
````

````python
iris.feature_names
````

````python
pca.components_ #주성분 계수
````

````python
pd.DataFrame( pca.components_, columns=iris.feature_names,
              index=['pc1','pc2'])
````

결과 해석:

PC1을 보면 'petal length (꽃잎 길이)'와 'petal width (꽃잎 너비)'의 계수 절댓값이 매우 큽니다. 이는 PC1이 주로 꽃잎의 크기에 대한 정보를 담고 있다는 의미이며, 꽃잎 크기가 품종을 구별하는 데 가장 중요한 요소였다는 것을 보여줍니다.

PC2는 'sepal width (꽃받침 너비)'에 상대적으로 높은 양의 계수를 가집니다.

- 누적 분산 설명력: 0.96

| **PC2 분산 설명력** | 0.22850762 (약 22.9%) | 두 번째 주성분(PC2)이 나머지 분산을 설명하는 데 기여합니다. |

### 2. 상세 해석
#### A. 개별 주성분의 설명력* **PC1 (Principal Component 1): 73.0%**

* 첫 번째 주성분인 PC1은 **원본 데이터가 가진 전체 정보(분산)의 약 73.0%**를 담고 있습니다.
* 이는 PCA가 찾아낸 새로운 축 중에서 데이터의 특징을 가장 잘 대표하고, 데이터의 구조나 패턴을 파악하는 데 **가장 중요한 역할**을 한다는 의미입니다.


* **PC2 (Principal Component 2): 22.9%**
* 두 번째 주성분인 PC2는 PC1이 설명하지 못한 **나머지 분산 중에서 가장 큰 부분**인 약 22.9%를 설명합니다.
* PC2는 PC1과는 **직교(독립적)**하는 방향으로, PC1이 놓칠 수 있는 **두 번째로 중요한 패턴**에 대한 정보를 제공합니다.



#### B. 누적 분산 설명력 (Accumulated Explained Variance)* **누적 분산 설명력: 96.0%**

* 원래 4차원(4개의 변수)이었던 데이터를 2개의 주성분(PC1, PC2)으로 축소했음에도 불구하고, **원본 데이터가 가진 정보의 96%를 그대로 유지**하고 있다는 뜻입니다.
* **정보 손실은 단지 4%**에 불과하며, 매우 효율적으로 차원 축소가 이루어졌다고 볼 수 있습니다.



### 3. 결론 및 의의

효율적인 차원 축소:

** 4차원 데이터를 2차원으로 축소하면서 96%의 정보를 보존했으므로, 이 2개의 주성분을 사용해 데이터를 분석하거나 시각화해도 **원본 데이터의 특징과 패턴을 거의 완벽하게 파악**할 수 있습니다.
* **시각화의 유효성:** 4차원 데이터를 2차원 평면(PC1과 PC2를 축으로 하는)에 나타내는 산점도(Scatter Plot)는 원본 데이터의 품종 구별이나 군집 형성 여부를 매우 정확하게 보여줄 것입니다.

> 이 결과는 붓꽃(Iris) 데이터처럼 변수 간의 상관관계가 높아 정보의 중복성이 큰 데이터셋에서 PCA가 얼마나 강력하게 작동하는지를 잘 보여줍니다.



이 분석 결과를 바탕으로 **PC1과 PC2를 구성하는 원래 변수들의 기여도(Component Loadings)**를 확인해 보면, 이 96%의 정보가 구체적으로 어떤 변수(예: 꽃잎 길이, 꽃받침 너비)에 의해 결정되었는지 더 깊이 이해할 수 있다


```python
주성분 계수 (Component Loadings):
     sepal length (cm)  sepal width (cm)  petal length (cm)  petal width (cm)
PC1           0.521066         -0.269347           0.580413          0.564857
PC2           0.377418          0.923296           0.024492          0.066942
```

 **주성분 계수(Component Loadings)** 또는 **고유 벡터(Eigenvectors)**이며, 각 주성분이 **원본 변수들로부터 어떤 기여를 받아 구성되었는지**를 나타내는 값입니다.

이 표를 통해 4차원 데이터의 정보(96% 분산)가 2개의 주성분에 어떻게 압축되었는지  해석.

##  주성분 계수 (Component Loadings) 해석
### 1. 표 요약

| 변수 | PC1 계수 | PC2 계수 |
| --- | --- | --- |
| **sepal length (cm)** | 0.521 (중) | 0.377 (중) |
| **sepal width (cm)** | -0.269 (소) | **0.923 (대)** |

> **해석 원칙:**
> * **절댓값의 크기:** 절댓값이 클수록 해당 주성분을 형성하는 데 **중요한 기여**를 했다는 뜻입니다.
> * **부호 (+/-):** 부호는 해당 변수와 주성분 간의 **방향성**을 나타냅니다. 같은 부호는 양의 상관관계, 다른 부호는 음의 상관관계를 의미합니다.

### 2. PC1 (Principal Component 1) 해석

PC1은 전체 분산의 약 73%를 설명하는 **가장 중요한 주성분**입니다.

| 변수 | 계수 | 해석 |
| --- | --- | --- |
| **petal length (cm)** | **0.580** | **가장 큰 양의 기여.** PC1을 정의하는 데 가장 중요한 변수입니다. |
| **petal width (cm)** | **0.565** | 두 번째로 큰 양의 기여. |
| **sepal length (cm)** | 0.521 | 세 번째로 큰 양의 기여. |
| **sepal width (cm)** | -0.269 | **음의 기여**를 하는 유일한 변수입니다. |

* **PC1의 특징:** PC1은 주로 **꽃잎의 크기(petal length, petal width)**와 **꽃받침의 길이(sepal length)**에 의해 정의됩니다. 이 세 변수는 모두 PC1과 양의 상관관계를 가지며, 이 값들이 커지면 PC1의 값도 커지는 경향이 있습니다.
* **PC1의 역할:** 붓꽃 데이터에서 PC1은 **'꽃의 전체적인 크기' 또는 '성숙도'**와 관련된 정보를 포착하고 있으며, 이 정보가 품종을 구분하는 데 73%의 분산 설명력으로 가장 중요했다는 것을 의미합니다.

### 3. PC2 (Principal Component 2) 해석

PC2는 PC1과 독립적이며, 나머지 분산의 약 23%를 설명하는 **두 번째로 중요한 주성분**입니다.

| 변수 | 계수 | 해석 |
| --- | --- | --- |
| **sepal width (cm)** | **0.923** | **압도적으로 가장 큰 양의 기여.** PC2를 정의하는 핵심 변수입니다. |
| **sepal length (cm)** | 0.377 | 양의 기여를 하지만, sepal width보다는 낮습니다. |
| **petal length (cm)** | 0.024 | 거의 기여하지 않습니다. |
| **petal width (cm)** | 0.067 | 거의 기여하지 않습니다. |

* **PC2의 특징:** PC2는 거의 전적으로 **꽃받침 너비(sepal width)**에 의해 정의됩니다. 나머지 꽃잎 관련 변수들의 기여는 매우 미미합니다.
* **PC2의 역할:** PC2는 PC1이 포착하지 못한 **'꽃받침의 너비'**라는 독자적인 정보를 담당합니다. 예를 들어, Setosa 품종은 다른 품종에 비해 꽃받침 너비가 넓은 경향이 있는데, PC2는 이러한 미묘한 차이를 포착하는 데 기여합니다.

### 4. 종합 결론

PCA 결과, 붓꽃 품종을 구별하는 96%의 정보는 다음과 같이 압축되었습니다.

1. **가장 중요한 정보 (PC1):** **꽃잎의 크기**와 꽃받침의 길이에 대한 종합적인 척도입니다. (가장 넓게 품종을 분리하는 주축)
2. **두 번째 중요한 정보 (PC2):** **꽃받침의 너비**에 대한 독자적인 척도입니다. (PC1만으로는 구분이 잘 안되는 미묘한 차이를 보정하는 부축)

이처럼 주성분 계수를 해석하면, 단순한 차원 축소를 넘어 **데이터셋에서 실제로 어떤 물리적인 변수들이 가장 중요한 역할을 했는지**에 대한 통찰을 얻을 수 있다.
