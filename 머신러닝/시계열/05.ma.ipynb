#  AR과 MA 모델 비교 및 MA의 오차 사용 이유

---

##  질문
**Xt가 실제 관측된 데이터인데, MA 모델은 왜 "오차(ε)"로 구성되어 있느냐?**

---

##  AR과 MA 비교

| 모델 | 수식 | 의존 대상 | 직관 |
|------|------|-----------|------|
| AR(1) | $X_t = \phi X_{t-1} + \varepsilon_t$ | 과거 값 $X_{t-1}$ | "자기 자신" 기반 예측 |
| MA(1) | $X_t = \mu + \varepsilon_t + \theta \varepsilon_{t-1}$ | 과거 오차 $\varepsilon_{t-1}$ | "과거 충격" 기반 예측 |

---

##  핵심 개념

### MA 모델은 **예측 불가능한 외부 충격들(오차)**이 현재 값을 만든다는 가정.

즉:

- $X_t$: 우리가 관측한 실제 값
- $\varepsilon_t$: 예측 오차 (백색잡음)
- MA 모델은 $\varepsilon$들의 조합이 $X_t$를 만든다고 가정합니다.

---

##  직관적 비교

### 🔸 AR 모델 예
$$
X_t = 0.8 \cdot X_{t-1} + \varepsilon_t
$$
→ 과거 데이터에 의존해서 예측

---

### 🔸 MA 모델 예
$$
X_t = \mu + \varepsilon_t + \theta \cdot \varepsilon_{t-1}
$$
→ 과거 "예측 오차"에 의존해서 예측

즉, **MA는 $X_t$를 직접 예측하는 구조가 아니라, 오차 간의 상관 구조를 모델링**하는 것

---

##  오차는 어떻게 알 수 있나?

> 오차는 **모델이 맞다고 가정하고 역산**하는 방식으로 추정합니다.

예:

1. $\varepsilon_0$을 0으로 가정
2. $X_1 = \mu + \varepsilon_1 + \theta \cdot \varepsilon_0$
3. $\Rightarrow \varepsilon_1 = X_1 - \mu - \theta \cdot \varepsilon_0$
4. 이 과정을 반복하여 $\varepsilon_t$를 계산

---

##  결론 요약

| 질문 | 답변 요약 |
|------|-----------|
| 관측값 $X_t$가 있는데 왜 MA는 오차 $\varepsilon$로 구성되나? | MA 모델은 데이터가 오차들의 조합으로 생성된다고 가정함 |
| 오차는 어떻게 구하나? | 모델 식을 알고 있다면 역산하여 추정 가능 |
| AR은 데이터 기반, MA는 왜 오차 기반인가? | AR은 자기상관 기반, MA는 오차 간 상관 기반 모델이기 때문 |

---





MA 모델의 핵심은 **"과거의 실수가 현재에 영향을 준다"**는 것입니다. 수식만 보면 복잡해 보이지만, 우리가 일상에서 **'어제 틀린 만큼 오늘 보정하는 과정'**으로 생각하면 이해하기 쉽다.

가장 기본이 되는 **MA(1) 모델** (어제의 오차만 반영하는 모델) 기준으로

---

1. MA(1) 모델 수식 뜯어보기
$$Y_t = \mu + \epsilon_t + \theta \epsilon_{t-1}$$

이 수식의 구성 요소를 하나씩 의미로 바꿔보면 다음과 같습니다.
- $Y_t$ (현재 값): 우리가 예측하거나 관찰하려는 오늘의 값입니다.
- $\mu$ (평균): 이 데이터가 보통 유지하는 기본적인 평균값입니다.
- $\epsilon_t$ (오늘의 충격): 오늘 새롭게 발생한 예측 불가능한 사건(백색 잡음)입니다.
- $\theta$ (가중치): 과거의 오차가 오늘에 얼마나 영향을 줄지 결정하는 비율입니다.
- $\epsilon_{t-1}$ (어제의 오차): 어제 발생했던 '예상 밖의 충격'입니다.
---

## 2.  예시 :(붕어빵 가게 매출)

어느 붕어빵 가게의 평소 하루 평균 매출이 **10만원($\mu=10$)**이고, 어제의 실수나 충격을 오늘 매출에 **50%($\theta=0.5$)**만큼 반영한다고 가정해 봅시다.

### 상황 설정:

1. **어제 ($t-1$): 갑작스러운 폭설(충격)로 손님이 안 와서 평소보다 -4만원의 오차가 발생했습니다. ($\epsilon_{t-1} = -4$)
2. 오늘 ($t$): 오늘은 특별한 일이 없어 새로운 충격이 **0**입니다.  ($\epsilon_t = 0$)

### 계산 (오늘의 매출 Y_t):

> **해석:** 평균은 10만원이지만, 어제 발생한 **'-4만원'이라는 충격의 여파**가 오늘까지 남아 매출이 **8만원**이 된 것입니다.

---

## 3. MA 모델의 핵심 특징

1. 기억력이 짧다: MA(1) 모델은 딱 어제의 충격($\epsilon_{t-1}$)까지만 기억합니다. 그저께 발생한 충격($\epsilon_{t-2}$)은 오늘 수식에 포함되지 않으므로 오늘 값에 아무런 영향을 주지 못합니다.

2. 보정의 의미: 예보보다 실제 기온이 낮았다면(오차), 그 오차만큼 다음 예보를 수정하는 것과 비슷합니다. 그래서 **"이동평균(Moving Average)"**이라는 이름이 붙었습니다. 오차가 발생할 때마다 그 오차를 반영하며 평균값이 이동하기 때문이죠.

---

###  정리하자면

* **AR 모델**은 "어제 매출이 높았으니 오늘도 높겠지"라고 **값 자체**를 따라가는 것이고,
* **MA 모델**은 "어제 예상치 못한 사건(오차)이 있었으니, 그 여파를 오늘 반영하자"라고 **충격의 잔상**을 계산하는 것.


```python
# 설정값
mu = 10                # 평균 매출
theta = 0.5            # 어제 오차 반영 비율 (50%)
errors = [2, -4, 3]    # 3일 동안 발생한 예상치 못한 충격들 (랜덤 오차)

# 첫날(t=1)은 어제의 오차가 없다고 가정 (0)
# 둘째날(t=2) 매출 계산
# 매출 = 평균 + 오늘의 충격 + (어제의 충격 * 반영비율)
day2_sales = mu + errors[1] + (errors[0] * theta)

print(f"2일차 매출: {day2_sales}") 
# 결과: 10 + (-4) + (2 * 0.5) = 7.0
```


패턴의 차이:  

$X_t = 0.8 \times X_{t-1}$ 이라는 AR(1) 규칙을 완벽하게 따르고 있습니다.

MA의 특징: 반면 MA 모델은 "과거의 충격이 곧 사라질 것"이라고 가정합니다. MA(1) 모델은 딱 한 시점 전의 오차만 보기 때문에, 데이터가 계속해서 일정하게 감소하는 '추세'를 따라가는 데는 적합하지 않다.

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# 1. 데이터 준비( AR용 데이터임)
X = [100, 80, 64, 51.2, 40.96, 32.768]

# 2. MA(1) 모델 학습 (Order: AR=0, I=0, MA=1)
# 관측치가 적어 경고가 발생할 수 있으나 계산은 가능합니다.
model = ARIMA(X, order=(1, 0, 0))
model_fit = model.fit()

# 3. 다음 한 스텝 예측
forecast = model_fit.forecast(steps=1)

print(f"다음 예상 값: {forecast[0]:.4f}")
```

MA(이동평균) 모델에 가장 적합한 데이터:

 "특정한 방향으로 계속 가는 성질(추세)은 없지만, 한 번 발생한 충격이 다음 시점까지 잔상처럼 남아있는 데이터".
 
---

## 1. MA 모델에 적합한 데이터 생성 원리

MA(1) 모델 수식: $Y_t = 0 + \epsilon_t + 0.8 \epsilon_{t-1}$ (평균은 0, 어제의 충격을 80% 반영)

| 시점(t) | 랜덤 충격($\epsilon_t$) | 계산 식 ($Y_t$) | 최종 데이터 값 |
| --- | --- | --- | --- |
| 1 | **+10** (대박 뉴스) | 10 + (0.8 $\times$ 0) | **10** |
| 2 | **0** (보통) | 0 + (0.8 $\times$ 10) | **8** |
| 3 | **-5** (악재) | -5 + (0.8 $\times$ 0) | **-5** |
| 4 | **0** (보통) | 0 + (0.8 $\times$ -5) | **-4** |

위 표를 보면, 1시점에 터진 충격(+10)이 2시점까지 영향을 주고, 3시점의 충격(-5)이 4시점까지 영향을 줍니다. 이렇게 **충격이 한 박자 늦게 반영되는 데이터**가 MA에 딱 맞는 데이터.


```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# 1. MA 적합 데이터 생성 (평균 20, Theta 0.8)
np.random.seed(42)
errors = np.random.normal(0, 2, 10) # 랜덤한 충격들
X = []
mu = 20
theta = 0.8

for i in range(1, len(errors)):
    val = mu + errors[i] + theta * errors[i-1]
    X.append(val)

print(f"생성된 MA 적합 데이터: {[round( int(x), 2) for x in X]}")

# 2. MA(1) 모델로 학습 및 예측
model = ARIMA(X, order=(0, 0, 1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=1)

print(f"다음 시점 예측값: {forecast[0]:.2f}")
```

## 3. 이 데이터가 MA에 적합한 이유 (AR과 비교)

1. **평균 회귀:** 값이 위아래로 출렁이지만 결국 평균(20) 근처로 돌아오려 합니다. 
(전형적인 MA의 특징)

2. **단기 기억:** 100에서 80, 64로 계속 떨어지던 이전 데이터(AR 성향)와 달리, 이 데이터는 **바로 직전의 오차**만 중요하게 여깁니다.

3. **예측의 정확도:** 이런 데이터에서 MA 모델을 쓰면 "어제 예상보다 높았으니, 오늘도 그 잔상이 남아 평균보다 조금 높겠군"이라는 논리가 완벽하게 들어맞습니다.

### 정리하자면

* **이전 데이터(100 $\rightarrow$ 80 $\dots$):** "어제 값이 오늘의 기준이 되는" **AR 모델**용 데이터.

* **이번 데이터(출렁이는 데이터):** "어제의 충격이 오늘 살짝 남아있는" **MA 모델**용 데이터.

- 아니면 **평균 주변에서 출렁이는 형태**인가?
