````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
#     
np.random.seed(42)
x = np.arange(1, 21) + np.random.normal(0, 1, 20)
print('x=', x)
#  1  (Differencing)
diff_x = np.diff(x)
print('diff_x=', diff_x)

#    vs   
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, marker='o')
plt.title("   (Non-Stationary)")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(diff_x, marker='o', color='orange')
plt.title(" 1   (Stationary?)")
plt.grid(True)

plt.tight_layout()
plt.show()
````
