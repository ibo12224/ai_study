#   군집분석 핵심 개념 정리

## 1. 군집분석(Clustering) 이란?
- **정의**: 비슷한 특성을 가진 데이터들을 묶어서 **그룹(Cluster)**으로 만드는 분석 방법
- **목적**: 같은 그룹 안의 데이터는 서로 **유사**하고, 다른 그룹 간에는 **차이**가 크게 하도록 함
- **활용 예시**
  - 고객 세분화 (VIP, 일반, 잠재고객)
  - 이미지 분류
  - 문서 토픽 분류

---

## 2. 군집분석의 주요 개념

| 개념 | 설명 | 비유 |
|------|------|------|
| **군집(Cluster)** | 서로 비슷한 데이터들의 모임 | 같은 취향의 사람끼리 만든 동아리 |
| **유사도(Similarity)** | 데이터 간의 '가깝다'의 정도 | 집과 집 사이 거리 |
| **거리(Distance)** | 유사도의 반대 개념, 가까울수록 값이 작음 | 친구 집까지 걸리는 시간 |
| **내부 응집도(Within-cluster similarity)** | 같은 군집 내의 데이터가 얼마나 비슷한지 | 같은 동아리 사람끼리 얼마나 친한지 |
| **군집 간 분리도(Between-cluster separation)** | 다른 군집끼리 얼마나 떨어져 있는지 | 동아리끼리 얼마나 다른 취미를 가지는지 |

---

## 3. 거리 계산 방법
| 거리 종류 | 계산 방식 | 특징 |
|-----------|-----------|------|
| **유클리드 거리(Euclidean)** | 직선 거리 계산 | 가장 많이 사용 |
| **맨해튼 거리(Manhattan)** | 축을 따라 이동한 거리 합 | 바둑판 길 이동 |
| **코사인 유사도(Cosine Similarity)** | 두 벡터 방향 비교 | 텍스트 분석에서 자주 사용 |

---

## 4. 군집화 방법
| 방법 | 특징 | 장점 | 단점 |
|------|------|------|------|
| **계층적 군집(Hierarchical)** | 트리 구조(덴드로그램)로 표현 | 군집 수를 미리 몰라도 됨 | 데이터 크면 느림 |
| **비계층적 군집(대표: K-means)** | 군집 수 K를 먼저 지정 | 계산 빠름, 대용량 가능 | K를 사전에 알아야 함 |

---

## 5. K-means 알고리즘 절차
1. **K** (군집 개수) 결정
2. 임의의 **K개 중심점(Centroid)** 설정
3. 각 데이터 → 가장 가까운 중심점에 할당
4. 각 군집의 중심점 재계산
5. 중심점 변화가 없을 때까지 3~4 반복

---

## 6. 군집 수 결정 방법
- **엘보우(Elbow) 방법**: 군집 수에 따른 SSE(오차 제곱합) 감소 그래프에서 꺾이는 지점 선택
- **실루엣(Silhouette) 계수**: 군집 내 응집도와 군집 간 분리도를 종합적으로 평가 (값이 클수록 좋음)

---

## 7. 해석 시 유의점
- 군집 분석은 **지도학습이 아님** → 정답이 없는 분석
- 변수의 **스케일 차이**에 따라 거리 계산 결과가 달라짐 → 표준화 필요
- 군집 개수는 **분석 목적과 해석 가능성** 고려

---

✅ **핵심 요약**
- 군집분석 = 데이터끼리 '비슷한 것끼리 묶기'
- 유사도는 가깝고, 거리 값은 작을수록 유사
- K-means vs 계층적 군집 → 속도 vs 시각화
- 표준화 필수, 군집 수 결정은 엘보우/실루엣 활용
