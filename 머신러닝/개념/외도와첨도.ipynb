회귀 분석 결과를 해석하다 보면 데이터가 정규 분포를 따르는지 확인해야 할 때가  사용하는 핵심 지표가 바로 **왜도(Skewness)**와 **첨도(Kurtosis)**

**왜도는 '좌우로 치우친 정도'**, 
**첨도는 '위로 뾰족한 정도'**

---

### 1. 왜도 (Skewness): 좌우 비대칭성

데이터의 분포가 한쪽으로 얼마나 쏠려 있는지를 나타냅니다.

* **왜도 = 0:** 좌우가 대칭인 종 모양(정규 분포)입니다.
* **왜도 > 0 (Positive Skew):** 꼬리가 **오른쪽**으로 길게 늘어진 형태입니다. 데이터가 왼쪽(낮은 값)에 몰려 있습니다. (예: 가구당 소득 분포)
* **왜도 < 0 (Negative Skew):** 꼬리가 **왼쪽**으로 길게 늘어진 형태입니다. 데이터가 오른쪽(높은 값)에 몰려 있습니다. (예: 난이도가 매우 쉬운 시험 점수)

---

### 2. 첨도 (Kurtosis): 뾰족함과 꼬리의 두께

분포가 중심부에서 얼마나 뾰족하게 솟아 있는지, 그리고 양 끝(꼬리)에 데이터가 얼마나 몰려 있는지를 나타냅니다.

* **첨도 = 3 (또는 보정 후 0):** 정규 분포의 전형적인 모습입니다. (이를 **중첨**, Mesokurtic이라 합니다.)
* **첨도 > 3 (Positive Kurtosis):** 정규 분포보다 **중앙이 더 뾰족**하고 꼬리가 두껍습니다. 극단적인 값(Outliers)이 많이 존재할 가능성이 높습니다. (이를 **고첨**, Leptokurtic이라 합니다.)
* **첨도 < 3 (Negative Kurtosis):** 정규 분포보다 **중앙이 완만**하고 꼬리가 얇습니다. 데이터가 전체적으로 넓게 퍼져 있는 형태입니다. (이를 **저첨**, Platykurtic이라 합니다.)

> **참고:** 많은 통계 프로그램(Python 포함)에서는 정규 분포의 첨도를 0으로 맞추기 위해 실제 계산값에서 3을 뺍니다. 이를 **초과 첨도(Excess Kurtosis)**라고 부르기도 합니다.

### 4. 왜 이 수치들이 중요한가?

회귀 분석의 전제 조건 중 하나는 **"오차항이 정규 분포를 따라야 한다"**는 것입니다.

* 왜도나 첨도가 너무 크면(절대값 기준 왜도 > 2, 첨도 > 7 등 기준은 학술마다 다름), 데이터가 정규성을 잃었다고 판단.
* 이 경우 로그 변환(Log Transform) 등을 통해 데이터를 정규 분포에 가깝게 만든 후 분석해야 모델의 신뢰도가 높아진다.


### 외도(Skewness)와 첨도(Kurtosis)의 계산 공식

기본적으로 데이터가 평균으로부터 떨어진 정도인 **'적률(Moment)'**을 활용.


---

### 1. 기본 개념: n차 적률

공식에 공통적으로 들어가는 핵심 요소는 **표준 점수(Z-score)의 n승**을 평균낸 값입니다.

* 평균 ($\mu$): 데이터의 중심
* 표준편차 ($\sigma$): 데이터가 퍼진 정도
* 편차 ($\frac{X - \mu}{\sigma}$): 각 데이터가 평균에서 몇 표준편차만큼 떨어져 있는지 나타내는 수치

---

### 2. 왜도 (Skewness) 공식

왜도는 **3차 적률**을 활용합니다. 편차를 세 제곱()하면 양수와 음수의 부호가 살아있기 때문에, 어느 방향으로 꼬리가 긴지 알 수 있습니다.

$$Skewness = E\left[\left(\frac{X-\mu}{\sigma}\right)^3\right]$$

**[풀어쓴 공식]**


$$G_1 = \frac{\frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^3}{\left(\frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^2\right)^{3/2}}$$

* **분자:** 편차의 세제곱 합 (비대칭성 측정)
* **분모:** 표준편차의 세제곱 (단위를 맞추기 위한 표준화)

---

### 3. 첨도 (Kurtosis) 공식

첨도는 **4차 적률**을 활용합니다. 편차를 네 제곱()하면 모든 값이 양수가 되며, 평균에서 아주 멀리 떨어진 값(이상치)의 영향력이 극대화되어 그래프의 '뾰족함'이나 '꼬리의 두께'를 나타냅니다.

$$Kurtosis = E\left[\left(\frac{X-\mu}{\sigma}\right)^4\right]$$

**[풀어쓴 공식]**

$$G_2 = \frac{\frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^4}{\left(\frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^2\right)^2}$$

#### 💡 초과 첨도 (Excess Kurtosis)

통계 소프트웨어(Python, R 등)에서는 보통 위 공식에서 **3을 뺀 값**을 사용합니다. 정규 분포의 첨도가 3이기 때문에, 이를 0으로 맞춰서 보기 편하게 만들기 위함입니다.


---

### 4. 한눈에 보는 요약표

| 구분 | 차수 | 공식의 핵심 | 의미 |
| --- | --- | --- | --- |
| **평균** | 1차 적률 | $E[X]$ | 데이터의 중심 위치 |
| **분산** | 2차 적률 | $E[(X-\mu)^2]$ | 데이터가 퍼진 정도 |
| **왜도** | 3차 적률 | $E[(X-\mu)^3]$ | 좌우 비대칭 (0이면 대칭) |
| **첨도** | 4차 적률 | $E[(X-\mu)^4]$ | 뾰족함/이상치 비중 (3이면 정규분포) |

---





### 5. 파이썬으로 직접 계산해보기

````python
from scipy.stats import skew, kurtosis
import numpy as np

data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

print(f"왜도: {skew(data)}") 
print(f"초과 첨도: {kurtosis(data)}") # scipy는 기본적으로 3을 뺀 값을 반환합니다.
````
