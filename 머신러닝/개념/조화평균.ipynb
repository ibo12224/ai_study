##  조화평균 (Harmonic Mean) 

조화평균은 **속도**나 **비율**과 같이 역수의 산술평균을 구하는 것이 자연스러운 상황에서 사용되는 특별한 종류의 평균입니다.

### 1. 조화평균이 필요한 이유

가장 흔한 예시는 **평균 속도**를 계산할 때입니다.

####  예시 상황: 왕복 이동

* **갈 때:** 시속 **10 km**
* **올 때:** 시속 **30 km**
* **거리:** 갈 때와 올 때의 거리는 **동일**하다고 가정합니다.

이때, 당신의 **평균 속도**는 얼마일까요?

####  잘못된 계산 (산술평균)

단순히 속도 두 개를 더해서 2로 나누면:
$$
\text{산술평균} = \frac{10 + 30}{2} = 20 \text{ km/h}
$$
이것은 **틀린** 답입니다. 왜냐하면 느린 속도($10\text{ km/h}$)로 움직이는 시간이 빠른 속도($30\text{ km/h}$)로 움직이는 시간보다 **훨씬 길기** 때문에, 평균 속도는 $20\text{ km/h}$보다 **느려야** 합니다.

####  올바른 계산 (조화평균)

평균 속도는 **전체 이동 거리**를 **전체 소요 시간**으로 나눈 값입니다.

1.  **전체 이동 거리:** 거리를 $D$라고 하면, 왕복 거리는 $2D$입니다.
2.  **전체 소요 시간:**
    * 갈 때 시간: $T_{\text{갈}}$
    * 올 때 시간: $T_{\text{올}}$
    $$
    T_{\text{갈}} = \frac{D}{10}, \quad T_{\text{올}} = \frac{D}{30}
    $$
    총 시간:
    $$
    T_{\text{총}} = \frac{D}{10} + \frac{D}{30} = D \left( \frac{1}{10} + \frac{1}{30} \right) = D \left( \frac{3+1}{30} \right) = D \left( \frac{4}{30} \right) = \frac{2D}{15}
    $$
3.  **평균 속도:**
    $$
    \text{평균 속도} = \frac{\text{전체 거리}}{\text{총 시간}} = \frac{2D}{\frac{2D}{15}} = 2D \times \frac{15}{2D} = 15 \text{ km/h}
    $$
    따라서 평균 속도는 **$15\text{ km/h}$** 입니다.

### 2. 조화평균 공식

이러한 **평균 속도**를 구하는 과정을 일반화한 것이 조화평균 공식입니다. $n$개의 데이터($x_1, x_2, \dots, x_n$)에 대한 조화평균($H$)은 다음과 같습니다.

$$
H = \frac{n}{\frac{1}{x_1} + \frac{1}{x_2} + \dots + \frac{1}{x_n}} = \frac{n}{\sum_{i=1}^{n} \frac{1}{x_i}}
$$

쉽게 말해, **'데이터의 개수($n$)'** 를 **'각 데이터의 역수들의 합'** 으로 나눈 값입니다.

####  예시 적용

앞선 예시($10\text{ km/h}$와 $30\text{ km/h}$)에 공식을 적용하면:

$$
H = \frac{2}{\frac{1}{10} + \frac{1}{30}} = \frac{2}{\frac{3}{30} + \frac{1}{30}} = \frac{2}{\frac{4}{30}} = 2 \times \frac{30}{4} = \frac{60}{4} = 15 \text{ km/h}
$$

공식으로 계산해도 동일한 **$15\text{ km/h}$**가 나옵니다.

### 3. 조화평균의 특징

* **역수의 산술평균:** 조화평균을 구하는 것은, **역수들의 산술평균**을 구한 다음 **다시 역수**를 취하는 것과 같습니다.
    $$
    \text{조화평균} = \frac{1}{\text{산술평균 of 역수들}}
    $$
* **가장 작은 값에 민감:** 조화평균은 데이터 값 중 **가장 작은 값**에 가장 크게 영향을 받습니다. 이 때문에 느린 속도나 낮은 생산성과 같이 **병목 현상**이 발생하는 상황을 더 잘 반영합니다.
* **크기 관계:** 일반적으로 조화평균($H$), 기하평균($G$), 산술평균($A$) 사이에는 다음과 같은 관계가 성립합니다.
    $$
    H \le G \le A
    $$

##  F1 스코어와 조화평균

F1 스코어는 모델의 성능을 하나의 숫자로 나타내는 점수입니다. 이 점수를 계산할 때 **산술평균** 대신 **조화평균**을 사용하는 이유.

### 비유: 자동차의 두 바퀴

모델의 성능을 자동차라고 생각해 보세요. 이 자동차를 움직이는 데는 **정밀도 (Precision)**와 **재현율 (Recall)**이라는 두 개의 중요한 바퀴가 필요합니다.

* **정밀도 (Precision) 바퀴:** '모델이 맞다고 한 것'의 정확성을 담당합니다.
* **재현율 (Recall) 바퀴:** '실제 정답'을 얼마나 놓치지 않고 잘 찾아냈는지 담당합니다.

#### 1. 산술평균으로 계산한다면? (위험한 평균)

만약 우리가 두 바퀴의 **산술평균**으로 자동차 성능을 평가한다면, 한쪽 바퀴가 찌그러져 있어도(예: Recall이 0점) 다른 쪽 바퀴가 매우 크면(예: Precision이 100점) 평균 점수는 50점이 됩니다.

$$
\text{산술평균} = \frac{100 (\text{P}) + 0 (\text{R})}{2} = 50 \text{점}
$$

**문제점:** 바퀴가 하나 망가진 자동차는 달릴 수 없습니다! 점수는 50점으로 **'그럭저럭'**처럼 보이지만, 실제 모델은 **전혀 쓸모가 없습니다.**

#### 2. 조화평균으로 계산한다면? (엄격한 균형 감시자)

**조화평균**은 이 문제를 해결합니다. 조화평균은 마치 두 바퀴 중 **가장 작은 바퀴의 크기**를 중심으로 평균을 계산하는 것과 비슷합니다.

* **조화평균의 특징:** 두 값 중 **어느 하나라도 0에 가까우면, 전체 평균 점수도 0에 가깝게 떨어집니다.**

위의 예시를 조화평균(F1 스코어)으로 계산하면:

$$
F1 \approx \text{매우 낮은 점수} \text{ (거의 0)}
$$

* **결과:** 조화평균은 한쪽 바퀴(지표)가 망가지면, 전체 성능 점수도 **처참하게** 낮추어 "이 자동차(모델)는 균형이 망가져서 사용할 수 없다"고 정확하게 경고합니다.

###  요약

F1 스코어에서 조화평균을 쓰는 이유는, **Precision과 Recall이 반드시 함께 높아야** 좋은 모델이라고 인정하기 위해서입니다.

**"둘 중 하나라도 낮으면, 전체 점수는 낮아진다!"**
