# 📐 유클리드 거리 = 피타고라스 정리

## 1. 피타고라스 정리

- 직각삼각형에서 빗변의 길이 $c$는 다음과 같이 계산됩니다:

$$
c = \sqrt{a^2 + b^2}
$$

- 여기서 $a$, $b$는 직각을 이루는 두 변입니다.

---

## 2. 유클리드 거리와의 관계 (2차원)

- 두 점 A($x_1$, $y_1$), B($x_2$, $y_2$) 사이의 거리:

$$
\text{거리} = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

- 이것은 $a = x_2 - x_1$, $b = y_2 - y_1$인 피타고라스 정리와 동일합니다.

---

## 3. 고차원(3차원 이상) 확장

- 3차원에서는 $z$ 좌표도 포함되어 다음과 같이 확장됩니다:

$$
\sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2}
$$

- 즉, 피타고라스 정리를 **n차원 벡터 공간**에 일반화한 것이 유클리드 거리입니다.

---

✅ **결론**:  
> 유클리드 거리는 피타고라스 정리를 바탕으로 정의된 거리입니다.


#  거리 개념 쉽게 이해하기

## 1. 유클리드 거리 (Euclidean Distance)

- **정의**: 두 점 사이의 직선 거리. 우리가 흔히 말하는 "직선 거리"입니다.
- **공식**:
  $$
  d(\mathbf{x}, \mathbf{y}) = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2 + \cdots + (x_n - y_n)^2}
  $$

- **예시 (2차원)**:
  - 점 A = (3, 4), 점 B = (0, 0)
  - 거리: $\sqrt{(3-0)^2 + (4-0)^2} = \sqrt{9 + 16} = \sqrt{25} = 5$

---

## 2. 맨해튼 거리 (Manhattan Distance)

- **정의**: 격자(바둑판) 위에서 가로와 세로만 이동할 때의 거리.
- **공식**:
  $$
  d(\mathbf{x}, \mathbf{y}) = |x_1 - y_1| + |x_2 - y_2| + \cdots + |x_n - y_n|
  $$

- **예시 (2차원)**:
  - 점 A = (3, 4), 점 B = (0, 0)
  - 거리: $|3-0| + |4-0| = 3 + 4 = 7$

---

## 3. 민코프스키 거리 (Minkowski Distance)

- **정의**: 일반화된 거리 공식. $p$ 값에 따라 유클리드 또는 맨해튼 거리로 바뀜.
- **공식**:
  $$
  d(\mathbf{x}, \mathbf{y}) = \left( \sum_{i=1}^{n} |x_i - y_i|^p \right)^{\frac{1}{p}}
  $$

- **특별한 경우**:
  - $p = 1$ → 맨해튼 거리
  - $p = 2$ → 유클리드 거리

---

## 4. 코사인 유사도 (Cosine Similarity)

- **정의**: 두 벡터의 방향이 얼마나 유사한지 측정. 거리라기보단 **유사도**입니다.
- **공식**:
  $$
  \cos(\theta) = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{x}\| \cdot \|\mathbf{y}\|}
  $$

- **유사도가 1에 가까울수록 같은 방향**, -1에 가까울수록 반대 방향

---

## 실생활 비유

| 거리 종류     | 비유                                               |
|---------------|----------------------------------------------------|
| 유클리드 거리 | 두 지점 사이를 직선으로 이동한 거리                |
| 맨해튼 거리   | 바둑판 위에서 오직 가로/세로로만 걷는 거리          |
| 코사인 유사도 | 두 사람이 같은 주제를 얼마나 비슷하게 말하는지    |

---

##  정리

- **유클리드 거리**: 실제 물리적 거리 측정
- **맨해튼 거리**: 도시 블록 구조처럼 경로가 제한적일 때
- **코사인 유사도**: 문서, 텍스트, 벡터 간 유사성 측정
