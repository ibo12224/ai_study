## 데이터 마이닝

데이터 마이닝은 방대한 데이터를 분석하여 유의미한 패턴과 규칙을 발견하고, 이를 예측 및 의사결정에 활용하는 것을 목표로 합니다. 다음은 데이터 마이닝의 핵심 개념과 상세 프로세스, 그리고 주요 기법에 대한  정리입니다.

---

### 1. 데이터 마이닝의 기본 이해

#### 데이터 마이닝의 목적
데이터 마이닝은 데이터 속에 숨겨진 **규칙**이나 **패턴**을 찾아내어 **예측**이나 **설명**을 가능하게 하는 것이 중요합니다.
* **예측**: 발견된 패턴을 통해 일종의 **함수나 모형**을 만들어 미래를 예측함으로써, 기업의 **사업 성공 확률을 높이는** 데 기여합니다.
* **설명**: 발견된 규칙을 통해 현상을 설명할 수 있게 되어, 기업의 **의사결정에 큰 도움**을 주고 신제품 개발 등 **신사업 성공**에 활용될 수 있습니다.

#### 통계분석과의 비교
| 특징 | 데이터 마이닝 | 통계분석 |
| :--- | :--- | :--- |
| **목표** | 데이터 속 숨겨진 **규칙, 패턴**을 찾아 **예측/설명** | **표본**을 통해 모집단의 특징을 **추정/검정** |
| **접근** | **가설과 검정**을 굳이 하지 않음 | **가설과 검정**을 포함하며, 표본이 필수 |
| **과정** | 인사이트를 얻어 활용하는 과정 | 수집 $\rightarrow$ 정제 $\rightarrow$ 추정 $\rightarrow$ 검정의 과정 |

### 2. 데이터 마이닝의 5단계 프로세스

데이터 마이닝은 분석의 목적을 정의하고 최종적으로 정보를 검증하는 5단계의 순서로 진행됩니다.

| 단계 | 설명 | 주요 활동 |
| :---: | :--- | :--- |
| **① 목적 정의** | 데이터 마이닝의 결과로 무엇을 알고자 하는지 **분명한 목적을 설정**합니다. | 이해관계자 동의 하에 데이터 마이닝에 필요한 **기반 및 필요 데이터를 정의**합니다. |
| **② 데이터 준비** | 필요한 데이터를 수집하고 데이터의 품질을 보장합니다. | 고객정보, 고객거래정보 등 데이터를 수집하고, **데이터 정제** 및 **보강 작업**을 거쳐 데이터 부족이 없도록 합니다. |
| **③ 데이터 가공** | 분석 기법 적용이 가능한 형태로 데이터를 준비합니다. | 데이터 마이닝 목적에 따른 **목적 변수**를 정의하고 데이터를 변환합니다. |
| **④ 데이터 마이닝 기법 적용** | 준비된 데이터를 활용하여 분석 기법을 적용해 목적하는 정보를 추출합니다. | 소프트웨어와 라이브러리를 사용하며, 알고리즘 적용에 따른 결과의 차이를 분석할 전문가가 필요합니다. |
| **⑤ 검증** | 추출한 정보를 검증합니다. | 얻은 모델의 **성능을 검증**하고, 평가가 우수하면 IT 부서와 협의하여 업무에 적용하며 **모델의 기대효과를 알립니다**. |

### 3. 데이터 마이닝의 종류 및 기법

데이터 마이닝 기법은 학습 방식과 분석 목적에 따라 다양하게 분류됩니다.

#### (1) 학습 방법에 따른 분류

| 구분 | 지도학습 (Supervised Learning) | 비지도학습 (Unsupervised Learning) |
| :---: | :--- | :--- |
| **개념** | **정답(출력 데이터)**이 있는 데이터를 활용해 분석 모델을 학습시킵니다. | **정답을 알려주지 않고** 입력 데이터만으로 숨겨진 **패턴**을 찾아 학습합니다. |
| **예시** | 고양이/강아지 사진 입력 시, 각 사진에 대한 **정답지** 정보가 필요. | 정답 레이블 없는 사진 100장을 입력 시, 컴퓨터는 특징이 비슷한 동물끼리 **군집화**를 실시. |
| **변수** | **독립 변수**에 따른 **종속 변수**가 존재합니다. | **종속 변수**가 없습니다. |
| **주요 기법** | **회귀** (선형회귀, 릿지, 라쏘), **분류** (로지스틱 회귀, 의사결정나무, SVM) | **군집** (K-means, DBSCAN), **연관** (Apriori), **차원 축소** (PCA, SVD) |

---

#### (2) 분석 목적에 따른 분류

| 분석 유형 | 목적 | 특징 및 주요 기법 |
| :---: | :--- | :--- |
| **분류 (Classification) 분석** | 데이터가 어느 그룹에 속하는지 **판별**하여 예측합니다. | 지도학습의 일종. 동/식물 종 분류, 신용 등급 평가 등에 활용. **로지스틱 회귀분석, 의사결정나무, 앙상블 분석, k-NN** 등 |
| **군집 (Clustering) 분석** | 데이터들 사이의 **유사성을 측정**하여 유사성이 높은 객체끼리 **하나의 그룹으로 묶습니다**. | 비지도학습의 일종. **병합적 방법, 분할적 방법, K-평균 군집(K-means)** 등 |
| **연관 (Association) 분석** | 데이터(상품)의 **연관성을 파악**합니다 (장바구니 분석). | 비지도학습의 일종. 맥주와 기저귀의 연관성 파악 등. **연관 규칙**으로 결과가 나타나며, 교차판매 등에 활용될 수 있습니다. |
