# 로그 함수(logarithmic function)


로그 함수는 **지수 함수(exponential function)의 역함수**

---

## 로그는 "어떤 수 b를 얻기 위해, 밑(base) a를 **몇 번** 거듭제곱해야 하는가?"라는 질문에 대한 답.

수식으로는 다음과 같이 표현합니다:

$$\log_{a} b = x \quad \iff \quad a^x = b$$

* **a (밑, base):** 거듭제곱을 하는 기준이 되는 수입니다. (a > 0, a \neq 1 조건을 만족해야 합니다.)
* **b (진수, argument):** 로그를 취하는 대상이 되는 수입니다. (b > 0 조건을 만족해야 합니다.)
* **x (로그 값):** 밑 a를 x번 제곱하면 b가 된다는 뜻으로, 로그 함수의 **결과**.

예를 들어, $log_{2} 8$은 "2를 몇 번 제곱해야 8이 될까?"라는 질문이고, 

2^3 = 8이므로 답은 **3**.


---

## 특별한 밑을 가지는 로그 (2, 10, e)로그 함수에서 특히 자주 사용되는 세 가지 밑.

### 1. 밑이 2인 로그: 이진 로그 (Binary Logarithm, $log_{2})$* **표기:** $log_{2} x$
* **사용처:** 주로 **컴퓨터 과학**이나 **정보 이론**에서 많이 사용됩니다. 

컴퓨터는 0과 1의 이진 시스템으로 작동하기 때문에, 데이터의 양(비트, bit)이나 알고리즘의 복잡도를 나타낼 때 $\log_{2}$가 핵심적으로 사용됩니다.

* **예시:** 2^4 = 16 이므로, $log_{2} 16 = 4$ 입니다. (2를 4번 제곱하면 16이 됩니다.)

### 2. 밑이 10인 로그: 상용 로그 (Common Logarithm, $log_{10}$)* **표기:** $log_{10} x$ 이지만, 밑 10은 **자주 생략**하고 그냥 $log{x}$ 라고 씁니다.


* **사용처:** 우리가 일상적으로 사용하는 수가 10진법이기 때문에, 큰 수의 **자릿수**나 **크기 비교**를 쉽게 할 때 사용됩니다.


* **지진의 리히터 규모:** 에너지가 10배 증가할 때마다 규모가 1씩 증가합니다.
* **소리의 데시벨(dB):** 소리의 세기를 상용 로그로 나타냅니다.
* **용액의 산성도(pH):** 수소 이온 농도의 역수에 상용 로그를 취한 값입니다.


* **예시:** $log 100$은 "10을 몇 번 제곱해야 100이 될까?"라는 질문이고, 10^2 = 100이므로 답은 **2**입니다.

### 3. 밑이 e인 로그: 자연 로그 (Natural Logarithm, $log_{e}$)* **표기:** $log_{e} x$ 이지만, 특별히 **$ln x$** 라고 씁니다.
* **e (자연 상수):** e는 약 $2.71828...$인 무리수이며, 복리 계산이나 미적분학에서 '자연적으로' 발생하는 매우 중요한 상수입니다.
* **사용처:** 주로 **미적분학**, **물리학**, **금융** (연속 복리), **자연 현상** (인구 증가, 방사성 물질 붕괴) 등에서 변화율을 다룰 때 가장 '계산이 편한' 밑이 됩니다.

미분($\frac{d}{dx} \ln x = \frac{1}{x}$)이 가장 간단한 형태를 가지기 때문

* **예시:** \ln e는 "e를 몇 번 제곱해야 e가 될까?"라는 질문이고, e^1 = e이므로 답은 **1**입니다.

## 요약 및 비교
| 종류 | 밑 (Base, a) | 표기법 | 의미하는 바 | 주로 사용되는 곳 |
| --- | --- | --- | --- | --- |
| **이진 로그** | 2 | $log_{2} x$ | $2^{\text{몇}} = x$ | 컴퓨터 과학, 정보 이론 |
| **상용 로그** | 10 |$\log x$ ($\log_{10} x$) | $10^{\text{몇}} = x$ | 일상적인 측정 (pH, 데시벨, 규모) |
| **자연 로그** | $e \approx 2.718$ | $\ln x$ ($\log_{e} x$) | $e^{\text{몇}} = x$ | 미적분학, 자연 현상 모델링 |
