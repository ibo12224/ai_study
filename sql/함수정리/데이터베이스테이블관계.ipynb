# 🔗 데이터베이스의 Relationship (관계) 종류와 예시

데이터베이스에서는 테이블 간의 관계를 정의할 수 있습니다.  
주로 3가지 관계 유형이 있으며, 실제 예시를 통해 쉽게 이해해 봅시다.

---

## 1️⃣ 1:1 관계 (One-to-One)

### 💡 개념  
한 테이블의 한 레코드가 다른 테이블의 한 레코드와만 연결되는 관계

### 📌 예시  
**사원(Employee)** - **사원증(EmployeeCard)** 관계  
- 한 사원은 하나의 사원증만 가짐  
- 하나의 사원증은 오직 한 사원에게만 발급됨

#### 테이블 구성 예
| Employee |           | EmployeeCard |
|----------|-----------|--------------|
| emp_id   | 🔗 1:1     | card_id      |
| name     |           | emp_id (FK)  |
| dept     |           | issue_date   |

---

## 2️⃣ N:1 관계 (Many-to-One)

### 💡 개념  
여러 레코드가 하나의 레코드와 연결되는 관계 (1:N 관계의 반대 방향)

### 📌 예시  
**학생(Student)** - **학과(Department)** 관계  
- 여러 학생이 같은 학과에 소속됨  
- 한 학과에는 여러 학생이 존재할 수 있음

#### 테이블 구성 예
| Student          |           | Department       |
|------------------|-----------|------------------|
| student_id       | 🔗 N:1     | dept_id (PK)     |
| name             |           | dept_name        |
| **dept_id (FK)** |           |                  |

---

## 3️⃣ M:N 관계 (Many-to-Many)

### 💡 개념  
한 테이블의 여러 레코드가 다른 테이블의 여러 레코드와 연결될 수 있는 관계  
(중간 테이블로 해결)

### 📌 예시  
**학생(Student)** - **강의(Course)** 관계  
- 한 학생이 여러 강의를 수강할 수 있음  
- 한 강의도 여러 학생이 수강할 수 있음

➡ 중간 테이블 **수강신청(Enrollment)** 을 사용하여 연결

#### 테이블 구성 예

| Student          |           | Enrollment               |           | Course          |
|------------------|-----------|---------------------------|-----------|------------------|
| student_id (PK)  |           | student_id (FK)           |           | course_id (PK)  |
| name             | 🔗 M:N     | course_id (FK)            | 🔗 M:N     | title           |
| major            |           | grade                     |           | credit          |

---

## ✅ 요약 정리

| 관계 유형 | 설명 | 예시 | 중간 테이블 필요 여부 |
|-----------|------|------|------------------------|
| 1:1       | 하나 ↔ 하나 | 사원 - 사원증 | ❌ |
| N:1       | 여러 ↔ 하나 | 학생 - 학과 | ❌ |
| M:N       | 여러 ↔ 여러 | 학생 - 강의 | ✅ (예: 수강신청) |
