```python
# import mysql.connector as MY
import pymysql

try:
    db = pymysql.connect(host="127.0.0.1", user='root' ,
                     passwd='1234',database='test')
    cur = db.cursor()
    sql = "insert into student values('이황',40,'1989-07-07')"
    cur.execute( sql )
    db.commit()
    db.close()
    print('추가성공')
except Exception as err:
    print('에러', err)

```
