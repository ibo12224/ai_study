```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib

matplotlib.rcParams['font.family']='NanumGothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df=pd.read_csv("../data/cars.csv")
df.set_index('Unnamed: 0',inplace=True)
```

```python
#X=speed
#Y=dist
y=df.iloc[:,1]
x=df.iloc[:,0]
```

1. 회귀 계수값을 구하시요( w, b)

```python
w=np.sum((x-x.mean())*(y-y.mean()))/np.sum((x-x.mean())**2)
#y=wx+b
#b=y-wx
np.pl
b=y.mean()-w*x.mean()
print(w,b)
```

2. 자동차속도가 10인경우 예측값을 구하시요

```python
print(w*10+b)
```

3. 자동차속도X축, 제동거리 y축 scatter를 그리시요

```python
fig, ax = plt.subplots(figsize=(14, 6))
ax.scatter(x,y)
```

4. 자동차속도에 따른 예측선을 scatter위에 그리시요.

```python
fig, ax = plt.subplots(figsize=(14, 6))
ax.scatter(x,y)
ax.plot(x,w*x+b,linestyle='-')
plt.show()
```
