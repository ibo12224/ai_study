#  통계학이란?

통계학(Statistics)은 **데이터를 수집, 정리, 분석, 해석**하여  
의사결정이나 예측에 활용하는 학문이다.  
즉, 불확실한 현상 속에서 **의미 있는 정보를 도출**하기 위한 체계적인 방법론이다.

---

##  통계학의 두 가지 주요 분야

| 구분 | 기술통계 (Descriptive Statistics) | 추론통계 (Inferential Statistics) |
|:----:|:---------------------------------:|:----------------------------------:|
| **개념** | 수집된 데이터를 요약·정리하여 한눈에 파악할 수 있게 하는 통계 | 표본 데이터를 이용해 모집단의 특성을 추정하거나 가설을 검정하는 통계 |
| **목적** | 데이터의 **현황 파악 및 요약** | 모집단에 대한 **추론 및 예측** |
| **주요 방법** | 평균, 중앙값, 분산, 표준편차, 그래프(히스토그램, 상자그림 등) | 추정(점추정, 구간추정), 가설검정, 회귀분석, 분산분석 등 |
| **데이터 활용 범위** | 주어진 **표본 또는 전체 데이터 자체** | 표본을 통해 **모집단 전체에 대한 결론**을 도출 |
| **예시** | 반 학생 30명의 수학 점수의 평균 계산 | 30명 점수를 바탕으로 전체 학년의 평균 점수를 추정 |
| **결과 형태** | 요약된 수치나 시각화 결과 | 확률적 결론 (예: 신뢰구간, p-value 등) |

---


- **기술통계**는 데이터를 요약하고 보여주는 단계  
- **추론통계**는 데이터를 바탕으로 보이지 않는 <span style='color:red;font-size:15pt'>모집단</span>을 예측하는 단계
