*F-검정(F-test)* 은  **"두 집단이 얼마나 다른지, 혹은 분산(데이터가 퍼진 정도)이 차이가 나는지"** 를 확인하는 통계 방법.


'평균'을 비교하는 T-검정과 달리, F-검정은 **'분산(Variance)'** 을 비교하는 것이 핵심.

---

## 1. F-검정의 핵심 개념

데이터가 얼마나 넓게 퍼져 있는지를 '분산', F-검정은 **두 개의 분산을 서로 나누어서(비율을 구해서)** 그 값이 1에 가까운지, 아니면 아주 큰지를 봅니다.

* **비율이 1에 가깝다면?** → "두 집단의 퍼진 정도가 비슷하네!"
* **비율이 1보다 훨씬 크다면?** → "두 집단의 퍼진 정도가 확실히 다르네!"

---

## 2. F-검정의 공식

공식은 생각보다 아주 간단. 두 집단의 분산을 서로 나누기만 하면 됨.

* : 첫 번째 집단의 분산
* : 두 번째 집단의 분산
* (보통 계산하기 편하게 분산이 더 큰 값을 분자에 둡니다.)

$$F = \frac{s_1^2}{s_2^2}$$

이 결과로 나온 **F-값**을 통계표와 비교해서, 우리가 정한 기준(유의수준)보다 큰지 확인하여 "차이가 있다/없다"를 결론 .

---

## 3. 언제 사용하나? 

### ① 두 학교 학생들의 성적 편차 비교

* **상황:** A학교와 B학교의 수학 점수 평균은 비슷합니다. 하지만 A학교는 학생들이 골고루 잘하고, B학교는 잘하는 학생과 못하는 학생의 격차가 매우 큽니다.
* **F-검정 활용:** "두 학교의 성적 **분산(격차)**이 통계적으로 진짜 차이가 나는가?"를 확인할 때 씁니다.

### ② ANOVA (분산분석) - 3개 이상 집단 비교

* **상황:** 빨강, 파랑, 노랑 세 가지 약의 효과를 비교하고 싶습니다.
* **F-검정 활용:** "약의 종류에 따라 효과가 **유의미하게** 다른가?"를 측정할 때 전체적인 차이를 한 번에 계산하기 위해 F-검정을 사용합니다.

---

## 4. 결과 해석하는 법 (쉬운 요약)

1. **F-값 확인:** 공식으로 구한 값입니다.
2. **p-value(유의확률) 확인:** 보통 **0.05**를 기준으로 삼습니다.
* **p-value < 0.05:** "두 집단은 유의미하게 차이가 난다!" (우연이 아니다)
* **p-value > 0.05:** "두 집단은 별 차이가 없다!" (우연일 가능성이 높다)



---

###  요약하자면!

F-검정은 **"두 집단의 흩어진 정도(분산)를 비교해서, 그 차이가 진짜인지 우연인지 판별하는 도구"**라고 이해하시면 가장 정확합니다.

**F-분포(F-distribution)**는 통계학에서 두 집단의 '흩어진 정도(분산)'를 비교할 때 사용하는 아주 중요한 도구입니다. 이 복잡해 보이는 곡선이 어떻게 태어났는지, 그 도출 과정.

---

## 1. F-분포의 탄생: "두 집단의 비율"

F-분포는 근본적으로 두 개의 독립적인 카이제곱($\chi^2$) 분포의 비율로 정의됩니다.

- 준비물: 서로 독립인 두 확률변수 $U$와 $V$가 있다고 가정합니다.
- 조건: $U$는 자유도가 $d_1$인 카이제곱 분포를 따르고, $V$는 자유도가 $d_2$인 카이제곱 분포를 따릅니다.
- 합체: 각 변수를 자신의 자유도로 나눈 뒤, 그 비율을 구하면 그것이 바로 F-분포가 됩니다.


$$F = \frac{U/d_1}{V/d_2}$$
여기서 $d_1$은 분자의 자유도, $d_2$는 분모의 자유도라고 부릅니다.

---

## 2. 왜 이런 계산을 하나요? (현실적인 배경)

우리가 실제로 궁금한 것은 "A 집단의 분산($\sigma_1^2$)과 B 집단의 분산($\sigma_2^2$)이 같은가?"임.

1. **정규분포에서 시작:** 두 집단이 정규분포를 따른다고 가정하고 표본을 뽑습니다.
2. **카이제곱과의 연결:** 표본분산()은 카이제곱 분포와 밀접한 관련이 있습니다. 구체적으로 $\frac{(n-1)S^2}{\sigma^2}$는 자유도가 인 카이제곱 분포를 따릅니다.
3. **비율 계산:** 만약 두 집단의 실제 분산()이 같다면, 두 표본분산의 비율을 구했을 때 그 값은 **F-분포**를 따르게 됩니다.

즉, **"두 표본분산의 비율이 F-분포의 예상 범위 안에 있는가?"**를 확인하여 두 집단이 통계적으로 차이가 나는지 결정하는 것입니다.

---

## 3. F-분포의 특징

그래프를 보면 몇 가지 특징을 알 수 있습니다.


* **0보다 큽니다:** 분산의 비율이므로 음수가 나올 수 없습니다. 그래서 항상 0에서 시작합니다.

* **오른쪽으로 긴 꼬리:** 대부분의 분산 비율은 1 근처에 몰려 있지만(두 분산이 비슷할 확률이 높으므로), 한쪽이 유독 클 경우 값이 매우 커질 수 있어 오른쪽으로 꼬리가 길게 늘어집니다.

* 면적($\alpha$): 그림에서 색칠된 부분($\alpha$)은 '우연히 이 정도로 큰 F-값이 나올 확률'을 의미합니다. 이 면적이 너무 작으면(보통 0.05 미만) "이건 우연이 아니라 진짜 두 집단이 다르구나!"라고 결론을 내립니다.

---

###  요약

F-분포는 **"두 개의 독립적인 카이제곱 변수를 각각의 자유도로 나누어 비율을 만든 것"**에서 도출되었다. 이 원리를 이용해 우리는 서로 다른 여러 집단의 분산을 비교하고, 실험 결과가 유의미한지 판단할 수 있다.

**F-분포표(F-distribution table)**는 통계 검정을 할 때 내가 구한 F-값이 얼마나 드문 일인지(유의미한지) 판단하기 위한 '기준점'을 찾는 도구.

표를 보는 방법은 크게 **3가지 정보**만 확인하면 됩니다.

---

### 1. 표를 보기 위해 필요한 3가지 값

표에서 값을 찾기 전에 먼저 다음 정보들을 알고 있어야 합니다.

1. 분자자유도 ($\nu_1$): 표의 **가로축(맨 윗줄)**에 있는 숫자입니다. 보통 비교하려는 집단의 수에서 1을 뺀 값($k-1$)입니다.
2. 분모자유도 ($\nu_2$): 표의 **세로축(맨 왼쪽)**에 있는 숫자입니다. 보통 전체 데이터 개수에서 집단 수를 뺀 값($N-k$)입니다.
3. 유의수준 ($q$ 또는 $\alpha$): 그림에서 색칠된 영역의 넓이입니다. 보통 **0.05(5%)**나 **0.01(1%)**을 가장 많이 사용합니다.

---

### 2. 실제 표 읽는 순서 (예시)

만약 조건이 분자자유도($\nu_1$) = 3, 분모자유도($\nu_2$) = 5, 유의수준($q$) = 0.05라면 어떻게 찾을까요?


1. 가로줄에서 찾기: 맨 윗줄(분자자유도 $\nu_1$)에서 **'3'**을 찾습니다.
2. 세로줄에서 찾기: 맨 왼쪽 열(분모자유도 $\nu_2$)에서 **'5'**를 찾습니다.
3. 유의수준 고르기: $\nu_2$가 5인 칸 옆을 보면 $q$값이 0.10, 0.05, 0.025, 0.01 네 줄이 있습니다. 그중 '0.05' 행을 선택합니다.
4. **교차점 읽기:** 가로 '3'과 세로 '5'의 '0.05' 행이 만나는 지점의 숫자를 읽습니다.
* 표를 확인해 보면 값은 **5.41**입니다.



> **해석:** 이 경우, 내가 계산한 F-값이 **5.41보다 크다면** "이 결과는 우연이 아닐 확률(유의미할 확률)이 높다"고 결론 내립니다.

---

### 3. 표 상단의 공식 이해하기

이미지 상단에 적힌 공식들은 표의 성질을 설명합니다.

* $P\{F \geq F_{(q; \nu_1, \nu_2)}\} = q$: F-값이 표에서 찾은 기준값보다 클 확률이 바로 색칠된 넓이 $q$라는 뜻입니다.

* 역수 공식 ($F_{1-q} = 1/F_q$): 만약 분자/분모의 자유도가 바뀐 값을 찾고 싶을 때 사용하는 수학적 규칙입니다. (예: $\nu_1=3, \nu_2=5$의 값은 $\nu_1=5, \nu_2=3$ 값의 역수와 관련이 있습니다.)

---

### 

* 표를 볼 때 가장 많이 실수하는 것이 분자자유도와 분모자유도를 반대로 보는 것입니다. 가로가 분자($\nu_1$), 세로가 분모($\nu_2$)라는 점만 기억

* 보통 논문이나 리포트에서는 유의수준 **0.05**에 해당하는 값을 기준으로 많이 삼는다.
