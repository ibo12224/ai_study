##  확률 질량 함수 (PMF)와 확률 밀도 함수 (PDF)의 차이점

확률 질량 함수 (PMF)와 확률 밀도 함수 (PDF)는 모두 확률 분포를 설명하는 데 사용되지만, 각각 **이산 확률 변수**와 **연속 확률 변수**에 적용된다는 근본적인 차이점이 있습니다.

---

### 1. 확률 질량 함수 (Probability Mass Function, PMF)

| 특징 | 설명 |
| :--- | :--- |
| **적용 대상** | **이산 확률 변수** (Discrete Random Variable) |
| **변수 예시** | 주사위 눈금, 동전 던지기 횟수, 불량품 개수 등 **셀 수 있는 값** |
| **함수의 의미** | 특정 값 $x$가 나올 **정확한 확률**을 나타냅니다. |
| **표현 (수학적)** | $P(X = x) = f(x)$ |
| **값의 범위** | $0 \le f(x) \le 1$ |
| **전체 확률** | 모든 가능한 값에 대한 함수 값의 **합**은 1입니다. $$\sum_{x} f(x) = 1$$ |

> **예시:** 주사위를 던졌을 때, $P(X=4)$는 정확히 4가 나올 확률 (1/6)입니다.

---

### 2. 확률 밀도 함수 (Probability Density Function, PDF)

| 특징 | 설명 |
| :--- | :--- |
| **적용 대상** | **연속 확률 변수** (Continuous Random Variable) |
| **변수 예시** | 키, 몸무게, 온도, 시간 등 **특정 범위 내의 모든 값**을 가질 수 있는 값 |
| **함수의 의미** | 특정 값 $x$ 자체의 확률이 아닌, **$x$ 주변의 확률 밀도**를 나타냅니다. |
| **표현 (수학적)** | $f(x)$ |
| **값의 범위** | $f(x) \ge 0$. **1을 초과할 수 있습니다.** |
| **전체 확률** | 모든 가능한 값에 대한 함수 값의 **적분**은 1입니다. $$\int_{-\infty}^{\infty} f(x) dx = 1$$ |

> **예시:** 어떤 사람의 키가 정확히 175.0000...cm 일 확률은 0입니다. 대신, 키가 **170cm 초과 180cm 이하일 확률**처럼 **특정 구간**의 확률을 구해야 합니다. 이는 해당 구간에 대해 PDF를 **적분**하여 계산합니다.



---

### 3. 주요 요약 비교

| 구분 | 확률 질량 함수 (PMF) | 확률 밀도 함수 (PDF) |
| :--- | :--- | :--- |
| **변수 종류** | **이산** (Discrete) | **연속** (Continuous) |
| **$P(X=x)$ 의 의미** | **특정 값의 확률** | **항상 0** (확률 밀도) |
| **확률 계산** | **합산** (Summation) | **적분** (Integration) |
