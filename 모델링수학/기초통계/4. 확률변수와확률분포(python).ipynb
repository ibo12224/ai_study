## 2. Ïó∞ÏÜçÌôïÎ•†Î∂ÑÌè¨ (Continuous Probability Distribution) 

**Ïó∞ÏÜçÌôïÎ•†Î∂ÑÌè¨**Îäî Ïó∞ÏÜç ÌôïÎ•†Î≥ÄÏàòÏùò ÌôïÎ•†Î∂ÑÌè¨Î•º **ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò(Probability Density Function, PDF)**Î°ú ÌëúÌòÑÌï©ÎãàÎã§. ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÎäî ÌäπÏ†ï Íµ¨Í∞ÑÏóêÏÑúÏùò ÌôïÎ•†ÏùÑ Íµ¨ÌïòÎäî Îç∞ ÏÇ¨Ïö©ÎêòÎ©∞, ÌôïÎ•†ÏùÄ Ìï®ÏàòÏùò Í∑∏ÎûòÌîÑ ÏïÑÎûò Î©¥Ï†ÅÏúºÎ°ú Í≥ÑÏÇ∞Îê©ÎãàÎã§. Ïù¥ Î©¥Ï†ÅÏùÄ Ï†ÑÏ≤¥ 1Ïù¥ ÎêòÏñ¥Ïïº Ìï©ÎãàÎã§.

---

### ÏòàÏãú: ÏÇ¨ÎûåÏùò ÌÇ§ ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò

ÏÇ¨ÎûåÏùò ÌÇ§Í∞Ä 150cmÏóêÏÑú 200cm ÏÇ¨Ïù¥Ïóê ÏûàÎã§Í≥† Í∞ÄÏ†ïÌïòÍ≥†, Ïù¥ Íµ¨Í∞ÑÏóêÏÑúÏùò ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÎ•º Í∑∏Î†§Î≥¥Î©¥, ÏïÑÎûòÏôÄ Í∞ôÏùÄ Í≥°ÏÑ†Ïù¥ ÎÇòÏò¨ Ïàò ÏûàÏäµÎãàÎã§. Ïù¥ Í≥°ÏÑ† ÏïÑÎûò Î©¥Ï†ÅÏùò Ìï©ÏùÄ 1ÏûÖÎãàÎã§. ÌäπÏ†ï Íµ¨Í∞Ñ, ÏòàÎ•º Îì§Ïñ¥ 160cmÏóêÏÑú 170cm ÏÇ¨Ïù¥Ïùò ÌôïÎ•†ÏùÄ Í≥°ÏÑ† ÏïÑÎûò Í∑∏ Íµ¨Í∞ÑÏóê Ìï¥ÎãπÌïòÎäî Î©¥Ï†ÅÏúºÎ°ú Í≥ÑÏÇ∞Îê©ÎãàÎã§.

* $P(160 \le X \le 170)$Îäî 160cmÏóêÏÑú 170cm ÏÇ¨Ïù¥Ïùò ÌÇ§Í∞Ä ÎÇòÏò¨ ÌôïÎ•†Î°ú, Ïù¥ Íµ¨Í∞ÑÏóêÏÑúÏùò Í≥°ÏÑ† ÏïÑÎûò Î©¥Ï†ÅÏùÑ Íµ¨Ìï©ÎãàÎã§.
* ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò ÏïÑÎûò Î©¥Ï†ÅÏùò Ìï©ÏùÄ Ìï≠ÏÉÅ 1ÏûÖÎãàÎã§.

## Ïù¥ÏÇ∞ÌôïÎ•†

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Ï£ºÏÇ¨ÏúÑ Í∞í(ÌôïÎ•†Î≥ÄÏàò)
dice_values = np.arange(1, 7)

# Ï£ºÏÇ¨ÏúÑ Í∞Å Î©¥Ïù¥ ÎÇòÏò¨ ÌôïÎ•† (1/6): ÌôïÎ•†Î∂ÑÌè¨
probabilities = np.repeat(1/6, 6)
print(probabilities)

# ÎßâÎåÄÍ∑∏ÎûòÌîÑ ÏãúÍ∞ÅÌôî(Ïù¥ÏÇ∞ÌôïÎ•†Î∂ÑÌè¨Ïùò ÏãúÍ∞ÅÌôî)
plt.bar(dice_values, probabilities)
plt.xlabel("Dice Value")
plt.ylabel("Probability")
plt.title("Discrete Probability Distribution of a Fair Dice")
plt.show()

# ÌôïÎ•†Î∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞ Ï∂úÎ†•
df = pd.DataFrame({
    "Dice_Value": dice_values,
    "Probability": probabilities
})
df

# ÌôïÎ•†Î≥ÄÏàò: dice_values (Ï£ºÏÇ¨ÏúÑ Îàà 1 ~ 6)
# ÌôïÎ•†Î∂ÑÌè¨: df Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ï†ÑÏ≤¥ ÎòêÎäî dice_valuesÏôÄ probabilities Î∞∞Ïó¥Ïùò ÎåÄÏùë Í¥ÄÍ≥Ñ.
```

### Ïó∞ÏÜçÌôïÎ•†

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞® ÏÑ§Ï†ï
mu = 170   # ÌèâÍ∑†
sigma = 10  # ÌëúÏ§ÄÌé∏Ï∞®

# 100Î™ÖÏùò ÏÇ¨Îûå ÌÇ§Î•º Ï†ïÍ∑úÎ∂ÑÌè¨ÏóêÏÑú ÏÉòÌîåÎßÅ
np.random.seed(42)  # Ïû¨ÌòÑ Í∞ÄÎä•Ìïú Í≤∞Í≥ºÎ•º ÏúÑÌï¥ ÏãúÎìú ÏÑ§Ï†ï

## ÌôïÎ•†Î≥ÄÏàò:sample_heights
#ÌèâÍ∑†Ïù¥ 170Ïù¥Í≥† ÌëúÏ§ÄÌé∏Ï∞®Í∞Ä 10Ïù∏ Î™®ÏßëÎã®ÏóêÏÑú ÏÉòÌîåÎßÅÎêú 100Î™ÖÏùò ÌÇ§ Í∞í. 
#ÌÇ§Îäî Ïó∞ÏÜçÏ†ÅÏù∏ Í∞í(Ïòà: 170cmÎ∂ÄÌÑ∞ 175cm ÏÇ¨Ïù¥Ïùò Î™®Îì† Í∞í)ÏùÑ Í∞ÄÏßà Ïàò ÏûàÎäî Ïó∞ÏÜç ÌôïÎ•†Î≥ÄÏàò.
sample_heights = np.random.normal(loc=mu, scale=sigma, size=100)

#ÌôïÎ•†Î∂ÑÌè¨ (Probability Distribution)
#Ï†ïÏùò: Ïó∞ÏÜç ÌôïÎ•†Î≥ÄÏàòÏùò ÌôïÎ•†Î∂ÑÌè¨Îäî **ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò(PDF)**Î°ú ÌëúÌòÑÎêòÎ©∞, ÌôïÎ•†ÏùÄ Ìï®ÏàòÏùò Í∑∏ÎûòÌîÑ ÏïÑÎûò Î©¥Ï†ÅÏúºÎ°ú Í≥ÑÏÇ∞Îê©ÎãàÎã§.
#ÏΩîÎìú ÏÉÅÏùò ÏúÑÏπò: ÏΩîÎìúÏóêÏÑú ÏÉùÏÑ±ÌïòÍ≥† ÏãúÍ∞ÅÌôîÌïòÎäî **Ï†ïÍ∑úÎ∂ÑÌè¨(Normal Distribution)**Í∞Ä ÌôïÎ•†Î∂ÑÌè¨Ïóê Ìï¥Îãπ

# ÌûàÏä§ÌÜ†Í∑∏Îû® (Î∞ÄÎèÑÌòï)
plt.hist(sample_heights, bins=10, density=True, color='green')
plt.xlabel("Height (cm)")
plt.ylabel("Probability Density")
plt.title("Histogram of 100 Sample Heights with Normal Distribution")

# Ï†ïÍ∑úÎ∂ÑÌè¨ Í≥°ÏÑ† Ï∂îÍ∞Ä
x = np.linspace(min(sample_heights), max(sample_heights), 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x, y)

plt.show()
```

#  ÌôïÎ•† Î∞ÄÎèÑ(Probability Density) Í≥ÑÏÇ∞ Í≥ºÏ†ï

ÌûàÏä§ÌÜ†Í∑∏Îû®(histogram)ÏùÑ Í∑∏Î¶¥ Îïå `density=True` ÏòµÏÖòÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥  
ÎßâÎåÄ ÎÜíÏù¥Í∞Ä ‚ÄúÎπàÎèÑÏàò‚ÄùÍ∞Ä ÏïÑÎãàÎùº ‚ÄúÌôïÎ•† Î∞ÄÎèÑ(Probability Density)‚ÄùÍ∞Ä ÎêúÎã§.

ÌôïÎ•† Î∞ÄÎèÑ Í≥ÑÏÇ∞Í≥ºÏ†ï.

---

## 1. Íµ¨Í∞Ñ(bin)Ïùò ÎÑàÎπÑ(Í≥ÑÍ∏â) Í≥ÑÏÇ∞

ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÄ Îç∞Ïù¥ÌÑ∞Î•º Ïó¨Îü¨ Í∞úÏùò **Íµ¨Í∞Ñ(bin)** ÏúºÎ°ú ÎÇòÎàÑÏñ¥ ÌëúÌòÑ.  
Íµ¨Í∞ÑÏùò Í∞úÏàòÎäî `bins` ÏòµÏÖòÏúºÎ°ú Í≤∞Ï†ï.

Ïòà)
- Îç∞Ïù¥ÌÑ∞ Î≤îÏúÑ: 0 ~ 100  
- `bins=10`  
‚Üí Îç∞Ïù¥ÌÑ∞Î•º 10Í∞ú Íµ¨Í∞ÑÏúºÎ°ú ÎÇòÎàÑÍ≥†,  
‚Üí Í∞Å Íµ¨Í∞ÑÏùò ÎÑàÎπÑ(bin width:Í≥ÑÍ∏â)Îäî:  
$$
\frac{100 - 0}{10} = 10
$$

- Îç∞Ïù¥ÌÑ∞ Î≤îÏúÑ: 150 ~ 190  
- `bins=10`  
‚Üí Îç∞Ïù¥ÌÑ∞Î•º 10Í∞ú Íµ¨Í∞ÑÏúºÎ°ú ÎÇòÎàÑÍ≥†,  
‚Üí Í∞Å Íµ¨Í∞ÑÏùò ÎÑàÎπÑ(bin width)Îäî:  
$$
\frac{190 - 150}{10} = 4
$$

---

## 2. Í∞Å Íµ¨Í∞ÑÏùò ÎπàÎèÑÏàò Í≥ÑÏÇ∞

Í∞Å Íµ¨Í∞ÑÏóê Î™á Í∞úÏùò Îç∞Ïù¥ÌÑ∞Í∞Ä Îì§Ïñ¥Í∞ÄÎäîÏßÄ ÏÑ∏Îäî Í≥ºÏ†ïÏûÖÎãàÎã§.

Ïòà)  
- Ï≤´ Î≤àÏß∏ Íµ¨Í∞ÑÏóê 8Í∞ú Îç∞Ïù¥ÌÑ∞Í∞Ä Îì§Ïñ¥Í∞îÎã§Î©¥ ‚Üí ÎπàÎèÑÏàò 8

---

## 3. ÌôïÎ•† Î∞ÄÎèÑ Í≥ÑÏÇ∞

ÌôïÎ•† Î∞ÄÎèÑÎäî Îã§Ïùå Í≥µÏãùÏúºÎ°ú Í≥ÑÏÇ∞Îê©ÎãàÎã§:

$$
\text{ÌôïÎ•† Î∞ÄÎèÑ} =
\frac{\text{Íµ¨Í∞ÑÏùò ÎπàÎèÑÏàò}}{\text{Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ Ïàò} \times \text{Íµ¨Í∞Ñ ÎÑàÎπÑ}}
$$

Ï¶â,  
- ÎπàÎèÑÏàòÎßåÏúºÎ°úÎäî ÌôïÎ•†Ïùò ÏùòÎØ∏Í∞Ä Î∂ÄÏ°±ÌïòÎØÄÎ°ú  
- Íµ¨Í∞Ñ ÎÑàÎπÑÎ•º Í≥†Î†§ÌïòÏó¨ **Î∞ÄÎèÑ(density)** Î°ú Î∞îÍæº Í∞íÏûÖÎãàÎã§.

---

## 4. Ï†ÑÏ≤¥ Î©¥Ï†ÅÏù¥ 1Ïù¥ ÎêòÎèÑÎ°ù Ï°∞Ï†ï

ÌôïÎ•† Î∞ÄÎèÑÎ•º ÏÇ¨Ïö©Ìïú ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÄ  
Í∞Å ÎßâÎåÄÏùò Î©¥Ï†Å(ÎÜíÏù¥ √ó Íµ¨Í∞Ñ ÎÑàÎπÑ)ÏùÑ Î™®Îëê ÎçîÌïòÎ©¥ **Ìï≠ÏÉÅ 1**Ïù¥ Îê©ÎãàÎã§.  
‚Üí Ï¶â, Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Ï∞®ÏßÄÌïòÎäî ÎπÑÏú® 100%Î•º ÏùòÎØ∏Ìï©ÎãàÎã§.

---

## ‚úÖ ÏöîÏïΩ

- yÏ∂ï Í∞í = ÌôïÎ•† Î∞ÄÎèÑ  
$$
\text{ÌôïÎ•† Î∞ÄÎèÑ} =
\frac{\text{Íµ¨Í∞ÑÏùò ÎπàÎèÑÏàò}}{\text{Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ Ïàò} \times \text{Íµ¨Í∞Ñ ÎÑàÎπÑ}}
$$

- ÌôïÎ•† Î∞ÄÎèÑÎ•º ÏÇ¨Ïö©ÌïòÎ©¥ ÌûàÏä§ÌÜ†Í∑∏Îû®Ïùò Ï†ÑÏ≤¥ Î©¥Ï†ÅÏù¥ 1Ïù¥ ÎêòÎèÑÎ°ù Ïä§ÏºÄÏùºÎßÅ Îê©ÎãàÎã§.

---



```python
#  ÌååÏù¥Ïç¨ ÏòàÏãú ÏΩîÎìú (ÌûàÏä§ÌÜ†Í∑∏Îû® + ÌôïÎ•† Î∞ÄÎèÑ)

import numpy as np
import matplotlib.pyplot as plt

# 1. ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (Ï†ïÍ∑úÎ∂ÑÌè¨)
np.random.seed(42)
data = np.random.normal(loc=170, scale=10, size=100)

# 2. ÌûàÏä§ÌÜ†Í∑∏Îû® Í≥ÑÏÇ∞ (density=False ‚Üí ÎπàÎèÑÏàò)
counts, bin_edges = np.histogram(data, bins=10, density=False)
print( 'counts:',counts)
print( 'bin_edges:',bin_edges)
print('data Í∞ØÏàò:',len(data))
print('-------------'*2)
# 3. Íµ¨Í∞Ñ ÎÑàÎπÑ Í≥ÑÏÇ∞
bin_width = bin_edges[1] - bin_edges[0]

# 4. ÌôïÎ•† Î∞ÄÎèÑ Í≥ÑÏÇ∞
density = counts / (len(data) * bin_width)
print( 'ÎèÑÏàò 1Ïù∏Í≤ΩÏö∞:', 1/(100*4.4720232885986775))
print("ÎπàÎèÑÏàò:", counts)
print("Íµ¨Í∞Ñ ÎÑàÎπÑ:", bin_width)
print("ÌôïÎ•† Î∞ÄÎèÑ:", density )
print('ÎπàÎèÑÏàòÎ°ú Î≥ÄÌôò', ( density*bin_width)*len(data) )
print('Íµ¨Í∞ÑÎÑàÎπÑXÎ∞ÄÎèÑ', (density*bin_width).sum())
print('-------------'*2)

# 5. ÌûàÏä§ÌÜ†Í∑∏Îû® Í∑∏Î¶¨Í∏∞ (ÌôïÎ•† Î∞ÄÎèÑ Ï†ÅÏö©)
plt.hist(data, bins=10, density=True, color='green',
          alpha=0.6, label='Histogram (Density)')
plt.xlabel("Height (cm)")
plt.ylabel("Probability Density")
plt.title("Histogram with Probability Density")

# 6. Ï†ïÍ∑úÎ∂ÑÌè¨ Í≥°ÏÑ† Ï∂îÍ∞Ä
x = np.linspace(min(data), max(data), 100)
from scipy.stats import norm
y = norm.pdf(x, loc=170, scale=10)
plt.plot(x, y, label='Normal PDF')

plt.legend()
plt.show()
```
