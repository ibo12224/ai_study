

###  **정규분포를 표준 정규분포로 변환하는 공식 (Z-변환)**

<img src='https://postfiles.pstatic.net/20160613_20/padosori60_1465811092248us3gf_PNG/1-1.PNG?type=w773'>

정규분포를 표준 정규분포로 변환하는 공식은 **Z-변환**을 사용합니다. 표준 정규분포는 평균이 $0$, 표준편차가 $1$인 정규분포입니다.

---

### **Z-변환 공식:**

정규분포 $X$가 평균 $\mu$와 표준편차 $\sigma$를 가진다면, 이를 표준 정규분포로 변환하는 공식은 다음과 같습니다:

$$Z = \frac{X - \mu}{\sigma}$$

---

### **여기서:**

* **$Z$**는 표준 정규분포 상의 값 (**표준화된 값**),
* **$X$**는 변환 전의 값 (**원래의 정규분포 값**),
* **$\mu$**는 정규분포의 평균,
* **$\sigma$**는 정규분포의 표준편차.

## 정규분포 특징 

평균이 다르고 표준편차도 다르지만,

* $N(50, 5^2)$ 에서 $50 \sim 55$까지의 넓이(확률)나
* $N(30, 2^2)$ 에서 $30 \sim 32$까지의 넓이(확률) 같아요!

즉, **정규분포**에서 $m$이 몇이든 $\sigma$가 몇이든!

$$P(m \le X \le m+k\sigma)$$

은 언제나 **동일합니다**!! ($k$는 정수)

<img src='image/norm.png'>

---



###  **정규분포의 성질**

즉, 정규분포에서 $m$이 몇이든 $\sigma$가 몇이든!

$$P(m \le X \le m+k\sigma)$$

은 **언제나 동일합니다!!** ( $k$는 정수 )

---

이 내용은 **정규분포의 표준화**를 통해 $m$과 $\sigma$에 관계없이 구간 $[m, m+k\sigma]$의 확률이 $Z$-변환을 거치면 항상 $[0, k]$로 표준화되어 동일한 값을 갖게 된다



###  Z-변환 예시 및 의미

평균이 $\mu = 100$이고, 표준편차가 $\sigma = 15$인 정규분포에서 $70$점과 $100$점을 표준 정규분포로 변환하면 다음과 같이 계산합니다.

#### **1. 70점을 표준화 (Z-값으로 변환):**

$$Z = \frac{70 - 100}{15} = \frac{-30}{15} = -2$$

즉, **70점**은 표준 정규분포에서 **$Z = -2$**에 해당합니다.

#### **2. 100점을 표준화:**

$$Z = \frac{100 - 100}{15} = \frac{0}{15} = 0$$

**100점**은 표준 정규분포에서 **$Z = 0$**에 해당합니다.

---

### **Z-변환의 의미:**

* **Z-값**은 **평균으로부터 몇 표준편차만큼 떨어져 있는지**를 나타냅니다. 예를 들어, $Z = -2$는 원래 값이 평균보다 **2 표준편차 아래에 있음**을 의미합니다.
* 이 변환을 통해 모든 정규분포를 **평균 $0$, 표준편차 $1$인 표준 정규분포**로 변환할 수 있습니다.

$$P (0 \le Z \le 2) = 0.4772$$

### 4. **$Z$값 $1.96$의 의미 (신뢰구간)**

표준 정규분포에서 $z$값 $1.96$은 평균으로부터 **약 $1.96$ 표준편차 떨어진 값**을 의미하며, 통계적으로 매우 중요한 지표 중 하나입니다.

#### **4.1. 누적 확률**
* 표준 정규분포에서 $z$값이 $1.96$일 때, **누적 확률**은 약 $0.975$입니다.
* 이는 $z$값이 $1.96$ 이하인 값이 전체 분포에서 약 $97.5\%$를 차지한다는 의미입니다.

#### **4.2. 95% 신뢰구간**
* $z$값 $1.96$은 **$95\%$ 신뢰구간**에서 중요한 값으로 사용됩니다.
* 통계적으로, 정규분포에서 **평균을 중심으로 $\pm 1.96$ 표준편차 범위** 안에 데이터의 약 $95\%$가 포함됩니다.
* 따라서, $95\%$ 신뢰수준으로 어떤 값의 구간을 구할 때, 평균을 기준으로 $\pm 1.96$ 표준편차 범위가 해당 데이터의 신뢰구간이 됩니다.

이 값은 통계적 **가설 검정**이나 **신뢰구간 계산**에서 자주 활용됩니다.

---


<img src='https://upload.wikimedia.org/wikipedia/commons/thumb/8/8c/Standard_deviation_diagram.svg/800px-Standard_deviation_diagram.svg.png'>

평균 100, 표준편차 15인 분포에서 몇 가지 점수(70, 80, 90, 100, 110)를 Z-점수로 변환합니다.

```python
import pandas as pd
import numpy as np

# 평균과 표준편차 설정
mean_value = 100  # 평균
sd_value = 15     # 표준편차

# 변환할 점수들
scores = np.array([70, 80, 90, 100, 110])

# Z-변환 공식 적용 (정규분포 -> 표준 정규분포 변환)
# Z = (X - μ) / σ
z_scores = (scores - mean_value) / sd_value

# 결과 출력
result_df = pd.DataFrame({
    'Original_Scores': scores,
    'Z_Scores': z_scores
})

print("--- 1. Z-변환 결과 데이터프레임 ---")
print(result_df)
```

Z-변환된 값들의 확률 밀도 함수를 계산하고, 표준 정규분포 곡선을 그립니다.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm # 표준 정규분포의 PDF 계산을 위해 scipy.stats 사용
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 평균과 표준편차 설정
mean_value = 100  # 평균
sd_value = 15     # 표준편차

# 원래 점수들을 설정 (여기서는 50 ~ 150 사이의 점수 범위 사용)
# 평균 ±4 표준편차 범위: 100 ± 4*15 = 40 ~ 160
scores = np.linspace(mean_value - 4 * sd_value, mean_value + 4 * sd_value, 1000)

# Z-변환 공식 적용
z_scores = (scores - mean_value) / sd_value

# Z-변환된 값에 대한 확률 밀도 함수 (표준 정규분포: mean=0, sd=1)
z_pdf_values = norm.pdf(z_scores, loc=0, scale=1) # loc: 평균, scale: 표준편차

# Z-변환된 표준 정규분포 곡선 그리기
plt.figure(figsize=(8, 5))
plt.plot(z_scores, z_pdf_values, color='blue', linewidth=2)
plt.title("표준 정규분포 (Z-변환 후)")
plt.xlabel("Z-값")
plt.ylabel("확률 밀도")
plt.axvline(x=0, color='red', linestyle='--', linewidth=2, label='평균 (Z=0)') # 평균 위치에 수직선 추가
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

원래 정규분포와 Z-변환된 표준 정규분포를 위아래 2개의 서브플롯으로 비교

```python
# R 코드 3: 원래 분포와 Z-변환된 분포를 함께 그리기

# 평균과 표준편차 설정
mean_value = 100  # 원래 분포의 평균
sd_value = 15     # 원래 분포의 표준편차

# x 값 범위 설정 (원래 분포 범위: 평균에서 ±4 표준편차 범위)
x_values = np.linspace(mean_value - 4 * sd_value, mean_value + 4 * sd_value, 1000)

# 원래 분포의 확률 밀도 함수 (PDF)
original_pdf_values = norm.pdf(x_values, loc=mean_value, scale=sd_value)

# Z-변환 공식 적용
z_values = (x_values - mean_value) / sd_value

# Z-변환된 값에 대한 확률 밀도 함수 (표준 정규분포)
z_pdf_values = norm.pdf(z_values, loc=0, scale=1)

# 원래 분포와 Z-변환된 분포를 함께 그리기
fig, axes = plt.subplots(2, 1, figsize=(10, 8)) # 그래프 2개를 위아래로 배치 (2행 1열)
plt.subplots_adjust(hspace=0.4) # 서브플롯 간격 조정

# 1. 원래 분포 그리기
axes[0].plot(x_values, original_pdf_values, color='blue', linewidth=2)
axes[0].set_title("1. 원래 정규분포 (평균 100, 표준편차 15)")
axes[0].set_xlabel("점수")
axes[0].set_ylabel("확률 밀도")
axes[0].axvline(x=mean_value, color='red', linestyle='--', linewidth=2, label='평균 (100)') # 평균 위치 표시
axes[0].legend()
axes[0].grid(True, linestyle=':', alpha=0.6)

# 2. Z-변환된 분포 그리기
axes[1].plot(z_values, z_pdf_values, color='green', linewidth=2)
axes[1].set_title("2. Z-변환된 표준 정규분포")
axes[1].set_xlabel("Z-값")
axes[1].set_ylabel("확률 밀도")
axes[1].axvline(x=0, color='red', linestyle='--', linewidth=2, label='평균 (0)') # 평균 위치 표시 (Z-변환된 분포의 평균은 0)
axes[1].legend()
axes[1].grid(True, linestyle=':', alpha=0.6)

plt.show()
```
