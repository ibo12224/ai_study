# 데이터의 종류(Type)

1. 질적 척도 (Qualitative Scale)

1) 정의: 질적 척도는 데이터를 범주나 특성으로 분류하는 척도로, 수치적인 의미보다는 속성이나 특징을 나타내는 데이터에 사용. 일반적으로 측정 단위 없이 명목적(이름 또는 분류)인 값.

2) 특징:
- 수치가 아닌 범주로 데이터를 표현.
- 순위나 크기를 나타내지 않으며, 단순히 구분하는 용도로 사용.
- 계산을 할 수 없으며, 수의 의미보다는 구분의 의미가 크다.
```text
예시:
성별 (남성, 여성)
혈액형 (A형, B형, O형, AB형)
국적 (한국, 미국, 일본)
머리 색깔 (검은색, 갈색, 금발)
```

2. 양적 척도 (Quantitative Scale)

1)정의: 양적 척도는 데이터를 수치화하여 측정할 수 있는 척도로, 숫자로 표현되고 크기나 양을 나타낸다. 양적 데이터는 계산이 가능하며, 다양한 수학적 분석에 사용될 수 있다.

2)특징:
- 숫자로 데이터를 표현하며, 크기나 양을 나타낸다.
- 데이터를 더하고, 빼고, 곱하고, 나누는 수학적 연산이 가능.
- 연속형과 이산형이 있다:

2-1) 연속형(Continuous): 값이 연속적으로 존재할 수 있으며, 소수점까지 표현 가능 
(예: 키, 무게).

2-2) 이산형(Discrete): 값이 분리된 정수 형태로 존재합니다 (예: 사람 수, 자동차 대수).
```text
예시
------------------------
나이 (예: 25세):이산
키 (예: 175.2cm):연속
무게 (예: 70.2kg):연속
시험 점수 (예: 85점):이산
소득 (예: 5,000달러):이산
```

차이점 요약:
---
1. 질적 척도 (Qualitative Scale):
- 범주나 속성을 표현.
- 수치가 아닌 단순한 분류.
- 계산 불가.
- 예시: 성별, 혈액형, 국적.
2. 양적 척도 (Quantitative Scale):
- 수치로 표현되고 크기나 양을 나타냄.
- 수학적 계산 가능.
- 연속형과 이산형으로 나뉨.
- 예시: 나이, 키, 무게, 소득.

# 데이터 측정 수준(Level of Measurement):
### 측정 데이터를 어떤 방식으로 해석할 수 있는가

1. 구간척도 (Interval Scale)

1) 정의: 구간척도는 **수치 간의 간격(차이)**이 일정하고 의미가 있는 척도. 절대적인 0점이 없다는 것이 특징.

2) 특징:
- **차이(간격)**가 일정: 두 값 사이의 차이는 일정하며, 그 차이가 의미를 가진다.
- 절대 영점이 없음: 0이라는 값이 데이터에 포함될 수 있지만, 그 0은 "없음"을 의미하지는 않는다. 즉, 절대적인 기준으로서의 0이 존재하지 않기 때문에 "0"이라고 해도 그 값이 무의미하다는 뜻은 아니다.

- 덧셈과 뺄셈은 가능하지만, 곱셈과 나눗셈으로 비율을 계산하는 것은 불가능합니다. 예를 들어, 섭씨 온도에서 20도와 40도의 차이는 20도지만, 40도가 20도의 두 배로 더 덥다고 할 수는 없다.

3) 예시:
- 섭씨 온도: 0도가 온도가 없는 것이 아니라 특정한 기준일 뿐이다.
- 연도: 예를 들어, 2000년과 2020년은 20년 차이가 있지만, 0년이 무언가가 없는 상태를 나타내지는 않는다.
- IQ 점수: 두 사람의 IQ 차이가 일정하더라도, 절대적인 0이라는 기준이 없다.


2. 비율척도 (Ratio Scale)

1) 정의: 비율척도는 구간척도와 유사하지만, 절대적인 영점이 존재하며, 비율 계산이 가능한 척도. 즉, 0은 값이 없음을 의미하며, 곱셈과 나눗셈을 통해 두 값 간의 비율을 계산할 수 있다.

2) 특징:
- 절대 영점이 존재: 0은 값이 없음을 의미하고, 절대적인 기준으로 사용됨. 예를 들어, 0kg은 무게가 없음을 나타낸다.
- 비율 계산이 가능: 두 값 사이의 비율을 계산할 수 있습니다. 예를 들어, 4kg은 2kg의 두 배이며, 4는 2의 두 배로 의미를 가진다.
- 덧셈, 뺄셈뿐만 아니라 곱셈과 나눗셈이 가능하여, 비율을 통해 데이터를 해석할 수 있다.

3) 예시:
- 무게: 0kg은 무게가 없음을 의미하며, 4kg은 2kg의 두 배이다.
- 키: 0cm는 높이가 없음을 의미하며, 180cm는 90cm의 두 배이다.
- 시간: 0초는 시간이 존재하지 않는 상태를 나타내며, 10초는 5초의 두 배이다.
- 소득: 0달러는 소득이 없음을 나타내며, 1000달러는 500달러의 두 배이다.

```text
차이점 요약:
절대 영점:
구간척도에는 절대적인 0이 없고, 0이라는 값이 단순한 기준입니다.
비율척도에는 절대적인 0이 있으며, 값이 없음을 의미합니다.
연산 가능성:
구간척도에서는 덧셈과 뺄셈만 가능하며, 곱셈이나 비율 계산은 의미가 없습니다.
비율척도에서는 덧셈, 뺄셈, 곱셈, 나눗셈이 모두 가능하고, 비율을 계산할 수 있습니다.
적용 예시:
구간척도: 온도(섭씨), 연도, IQ 점수.
비율척도: 무게, 키, 시간, 소득.
