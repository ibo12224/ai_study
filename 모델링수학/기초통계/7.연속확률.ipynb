## 1. ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò (PDF)

* **Ï†ïÏùò**: ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎäî **Ïó∞ÏÜç ÌôïÎ•† Î∂ÑÌè¨**ÏóêÏÑú ÌäπÏ†ï Í∞í Ï£ºÏúÑÏùò Íµ¨Í∞ÑÏóê Í∞íÏù¥ ÎÇòÏò¨ Í∞ÄÎä•ÏÑ±(**ÌôïÎ•†Ïùò Î∞ÄÎèÑ**)ÏùÑ ÎÇòÌÉÄÎÉÖÎãàÎã§.

* **Î™©Ï†Å**: ÌäπÏ†ï Íµ¨Í∞ÑÏóêÏÑú Í∞íÏù¥ ÎÇòÏò¨ ÌôïÎ•†ÏùÑ Í≥ÑÏÇ∞ÌïòÎäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§.

* **ÌäπÏßï**: Ïó∞ÏÜçÌòï ÌôïÎ•† Î≥ÄÏàòÏùò Í≤ΩÏö∞, **ÌäπÏ†ï Í∞í ÏûêÏ≤¥Ïóê ÎåÄÌïú ÌôïÎ•†ÏùÄ 0**ÏûÖÎãàÎã§. ÎåÄÏã†, ÌäπÏ†ï Î≤îÏúÑÏóê Í±∏Ïπú ÌôïÎ•†ÏùÑ Íµ¨Ìï† Îïå **Ï†ÅÎ∂Ñ**ÏùÑ ÌÜµÌï¥ Í≥ÑÏÇ∞Ìï©ÎãàÎã§.

> ÏòàÎ•º Îì§Ïñ¥, $f(x)$Í∞Ä ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎùºÎ©¥:
$$
P(a \le X \le b) = \int_{a}^{b} f(x)dx
$$
Ïù¥ ÏãùÏùÄ Íµ¨Í∞Ñ $[a, b]$ ÏÇ¨Ïù¥ÏóêÏÑú Í∞íÏù¥ ÎÇòÏò¨ ÌôïÎ•†ÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§. 
**ÏòàÏãú:**

Ïó∞ÏÜçÌòï Í∑†ÏùºÎ∂ÑÌè¨ÏóêÏÑú $0$Í≥º $10$ ÏÇ¨Ïù¥Ïùò ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎäî ÏùºÏ†ïÌïú Í∞í(ÌèâÌèâÌïú ÏßÅÏÑ†)ÏûÖÎãàÎã§. Ïù¥Îïå ÌäπÏ†ï Í∞í $5$ÏóêÏÑúÏùò ÌôïÎ•† Î∞ÄÎèÑÎäî ÏùºÏ†ïÌïòÏßÄÎßå, $P(X = 5) = 0$Ïù¥Í≥†, ÌäπÏ†ï Íµ¨Í∞Ñ(Ïòà: $4$ÏóêÏÑú $6$ ÏÇ¨Ïù¥)Ïùò ÌôïÎ•†ÏùÑ Í≥ÑÏÇ∞Ìï¥Ïïº ÏùòÎØ∏Í∞Ä ÏûàÏäµÎãàÎã§.

## 2. Ï†ÅÎ∂ÑÏùÑ Ïù¥Ïö©Ìïú Í±∞Î¶¨ Í≥ÑÏÇ∞ ÏòàÏ†ú

* Ïñ¥Îñ§ Ï∞®Í∞Ä ÏãúÏÜç $60\text{ km}$Î°ú $2$ÏãúÍ∞Ñ ÎèôÏïà Îã¨Î†∏ÏùÑ Îïå,
$$
\text{Í±∞Î¶¨} = \int_{0}^{2} 60 \,dt = 60 \times (2 - 0) = 120\text{ km}
$$

* Ïó¨Í∏∞ÏÑú **$60$ÏùÄ ÏÜçÎèÑ**, **$dt$Îäî ÏãúÍ∞ÑÏùò ÏïÑÏ£º ÏûëÏùÄ Îã®ÏúÑ**Î•º ÏùòÎØ∏Ìï©ÎãàÎã§.
* **Í≤∞Í≥º**Îäî **ÏÜçÎèÑ $\times$ ÏãúÍ∞Ñ = Í±∞Î¶¨**Ïùò ÏõêÎ¶¨Í∞Ä Ï†ÅÎ∂ÑÏùÑ ÌÜµÌï¥ ÌôïÏù∏ÎêòÎäî Í≤ÉÏûÖÎãàÎã§.

## 2. ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò (CDF)

* **Ï†ïÏùò**: ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò $F(x)$Îäî ÌôïÎ•† Î≥ÄÏàòÍ∞Ä ÌäπÏ†ï Í∞í $x$ **Ïù¥Ìïò**Ïùò Í∞íÏùÑ Í∞ÄÏßà ÌôïÎ•†ÏùÑ ÎÇòÌÉÄÎÉÖÎãàÎã§.
* **Î™©Ï†Å**: $x$ Ïù¥ÌïòÏùò ÌôïÎ•†ÏùÑ ÎàÑÏ†ÅÏ†ÅÏúºÎ°ú Í≥ÑÏÇ∞ÌïòÎäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§. Ï¶â, $X$Í∞Ä Ï£ºÏñ¥ÏßÑ Í∞íÎ≥¥Îã§ ÏûëÍ±∞ÎÇò Í∞ôÏùÄ ÌôïÎ•†ÏùÑ ÎÇòÌÉÄÎÉÖÎãàÎã§.

ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò $F(x)$Îäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï†ïÏùòÎê©ÎãàÎã§:
$$
F(x) = P(X \le x)
$$
Ïù¥Îäî ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò $f(t)$Î•º Ï†ÅÎ∂ÑÌïòÏó¨ ÏñªÏùÑ Ïàò ÏûàÏäµÎãàÎã§:
$$
F(x) = \int_{-\infty}^{x} f(t)dt
$$

* **ÏòàÏãú**: Ïó∞ÏÜçÌòï Í∑†ÏùºÎ∂ÑÌè¨ÏóêÏÑú $0$Í≥º $10$ ÏÇ¨Ïù¥Ïùò ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®ÏàòÎäî $0$ÏóêÏÑú ÏãúÏûëÌïòÏó¨, Í∞íÏù¥ Ï¶ùÍ∞ÄÌï†ÏàòÎ°ù ÎàÑÏ†Å ÌôïÎ•†Ïù¥ $1$Ïóê Í∞ÄÍπåÏõåÏßëÎãàÎã§. ÏòàÎ•º Îì§Ïñ¥, $F(5)$Îäî $0$Í≥º $5$ ÏÇ¨Ïù¥Ïùò Í∞íÏù¥ ÎÇòÏò¨ ÌôïÎ•†ÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.

## 1. Í∑†ÏùºÎ∂ÑÌè¨ (Continuous Uniform Distribution)

Ïó∞ÏÜçÌòï Í∑†ÏùºÎ∂ÑÌè¨Îäî ÌäπÏ†ï Íµ¨Í∞Ñ ÎÇ¥Ïùò **Î™®Îì† Í∞íÏù¥ ÎèôÏùºÌïú ÌôïÎ•† Î∞ÄÎèÑ**Î•º Í∞ÄÏßÄÎäî Î∂ÑÌè¨ÏûÖÎãàÎã§.

### 1. Í∞úÎÖê Î∞è ÌäπÏßï

* **Ï†ïÏùò**: Ï£ºÏñ¥ÏßÑ Íµ¨Í∞Ñ $[a, b]$ ÎÇ¥Ïùò **Î™®Îì† Í∞íÏóêÏÑú ÌôïÎ•† Î∞ÄÎèÑ(Probability Density)Í∞Ä ÏùºÏ†ïÌïú** Î∂ÑÌè¨ÏûÖÎãàÎã§.
* **ÌäπÏßï**:
    * Íµ¨Í∞Ñ $[a, b]$ Î∞ñÏùò ÌôïÎ•† Î∞ÄÎèÑÎäî **0**ÏûÖÎãàÎã§.
    * Íµ¨Í∞Ñ $[a, b]$ ÏïàÏóêÏÑúÎäî ÌôïÎ•† Î∞ÄÎèÑÍ∞Ä $\frac{1}{b-a}$Î°ú **ÏùºÏ†ï**Ìï©ÎãàÎã§.
* **ÌëúÌòÑ**: ÏßÅÏÇ¨Í∞ÅÌòï Î™®ÏñëÏùò Í∑∏ÎûòÌîÑÎ°ú ÎÇòÌÉÄÎÇòÎ©∞, ÌùîÌûà **ÏßÅÏÇ¨Í∞ÅÌòï Î∂ÑÌè¨(Rectangular Distribution)**ÎùºÍ≥†ÎèÑ Î∂àÎ¶ΩÎãàÎã§.



### 2. ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò (PDF)

Ïó∞ÏÜçÌòï Í∑†ÏùºÎ∂ÑÌè¨Ïùò ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò $f(x)$Îäî Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§:

$$
f(x) = \begin{cases} \frac{1}{b-a} & \text{for } a \le x \le b \\ 0 & \text{otherwise} \end{cases}
$$

* **ÌôïÎ•† Í≥ÑÏÇ∞**: Ïó∞ÏÜç Î∂ÑÌè¨Ïù¥ÎØÄÎ°ú, ÌäπÏ†ï Í∞í $X=x$Ïùò ÌôïÎ•†ÏùÄ $0$ÏûÖÎãàÎã§. ÌôïÎ•†ÏùÄ **ÌäπÏ†ï Íµ¨Í∞Ñ**ÏùÑ **Ï†ÅÎ∂Ñ**ÌïòÏó¨ Íµ¨Ìï©ÎãàÎã§.
    $$P(c \le X \le d) = \int_{c}^{d} f(x) dx$$

### 3. ÏòàÏãú



* **Î≤ÑÏä§ ÎèÑÏ∞© ÏãúÍ∞Ñ**: Î≤ÑÏä§Í∞Ä $15$Î∂Ñ Í∞ÑÍ≤©ÏúºÎ°ú Ïö¥ÌñâÎê† Îïå, Ï†ïÎ•òÏû•Ïóê ÎèÑÏ∞©Ìïú ÏäπÍ∞ùÏù¥ Î≤ÑÏä§Î•º Í∏∞Îã§Î¶¨Îäî ÏãúÍ∞Ñ $X$Îäî Íµ¨Í∞Ñ $[0, 15]$ ÏÇ¨Ïù¥Ïùò Î™®Îì† Í∞íÏùÑ Í∞ÄÏßà Ïàò ÏûàÏúºÎ©∞, Ïù¥ Íµ¨Í∞Ñ ÎÇ¥ÏóêÏÑú Í∏∞Îã§Î¶¥ ÌôïÎ•† Î∞ÄÎèÑÎäî Î™®Îëê ÎèôÏùºÌï©ÎãàÎã§.
    * Ïù¥ Í≤ΩÏö∞ $a=0$, $b=15$ Ïù¥ÎØÄÎ°ú, ÌôïÎ•† Î∞ÄÎèÑÎäî $\frac{1}{15-0} = \frac{1}{15}$ÏûÖÎãàÎã§.
* **ÎûúÎç§ Ïà´Ïûê ÏÉùÏÑ±**: Ïª¥Ìì®ÌÑ∞ ÌîÑÎ°úÍ∑∏Îû®Ïù¥ $0$Í≥º $1$ ÏÇ¨Ïù¥Ïùò Ïã§ÏàòÎ•º Î¨¥ÏûëÏúÑÎ°ú ÏÉùÏÑ±Ìï† Îïå ($a=0, b=1$), Ïù¥ Î≤îÏúÑ ÎÇ¥Ïùò Ïñ¥Îñ§ ÌäπÏ†ï Í∞í Ï£ºÏúÑÏùò ÌôïÎ•† Î∞ÄÎèÑÎäî $\frac{1}{1-0}=1$Î°ú Î™®Îëê ÎèôÏùºÌï©ÎãàÎã§.

ÎßåÏùº ($a=0, b=10$)ÏóêÏÑúÎäî ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÍ∞Ä Îã§ÏùåÍ≥º Í∞ôÏù¥ $\frac{1}{10}$Î°ú ÏùºÏ†ïÌï©ÎãàÎã§:
$$
f(x) = \frac{1}{10-0} = \frac{1}{10} \quad \text{for } x \in [0, 10]
$$
Ï¶â, $0$ÏóêÏÑú $10$ ÏÇ¨Ïù¥Ïùò Ïñ¥Îñ§ Í∞íÏù¥ ÏÑ†ÌÉùÎê† ÌôïÎ•† Î∞ÄÎèÑÎäî $\frac{1}{10}$Î°ú ÏùºÏ†ïÌï©ÎãàÎã§.

---


### ÌôïÎ•† Í≥ÑÏÇ∞ Í≥µÏãù (Í∑†ÏùºÎ∂ÑÌè¨)

Ïó∞ÏÜçÌòï Í∑†ÏùºÎ∂ÑÌè¨ÏóêÏÑú ÌäπÏ†ï Íµ¨Í∞Ñ $[a, b]$Ïóê Í∞íÏù¥ ÎÇòÏò¨ ÌôïÎ•†ÏùÄ Îã§ÏùåÍ≥º Í∞ôÏù¥ Í≥ÑÏÇ∞Îê©ÎãàÎã§:

$$
P(a \le X \le b) = \frac{b-a}{\text{max} - \text{min}}
$$

**ÏòàÏãú**: Ïö∞Ïú†Ïùò ÏñëÏù¥ $0\text{ml}$ÏóêÏÑú $10\text{ml}$ ÏÇ¨Ïù¥ÏóêÏÑú Í∑†ÏùºÌïòÍ≤å Î∂ÑÌè¨Ìï† Îïå, $5\text{ml}$ÏóêÏÑú $6\text{ml}$ ÏÇ¨Ïù¥Ïùº ÌôïÎ•† Í≥ÑÏÇ∞ÏùÄ Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§.
* $\text{max} = 10$ (ÏµúÎåìÍ∞í)
* $\text{min} = 0$ (ÏµúÏÜüÍ∞í)
* $a = 5, b = 6$ (Ïö∞Î¶¨Í∞Ä Íµ¨ÌïòÎ†§Îäî Íµ¨Í∞Ñ)

$$
P(5 \le X \le 6) = \frac{6-5}{10-0} = \frac{1}{10} = 0.1
$$
Îî∞ÎùºÏÑú, Ïö∞Ïú†Ïùò ÏñëÏù¥ $5\text{ml}$ÏóêÏÑú $6\text{ml}$ ÏÇ¨Ïù¥Ïóê ÏûàÏùÑ ÌôïÎ•†ÏùÄ **$0.1$ ($10\%$)**ÏûÖÎãàÎã§.

ÎÑ§, Ïù¥ÎØ∏ÏßÄÏùò ÌÖçÏä§Ìä∏Î•º Ï£ºÌîºÌÑ∞ ÎÖ∏Ìä∏Î∂Å ÎßàÌÅ¨Îã§Ïö¥ ÌòïÏãùÏúºÎ°ú Î≥ÄÌôòÌï¥ ÎìúÎ¶ΩÎãàÎã§.

---

###  ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò(CDF) Í≥µÏãù:

Í∑†ÏùºÎ∂ÑÌè¨ÏóêÏÑú Íµ¨Í∞Ñ $[\text{min}, \text{max}]$ ÏÇ¨Ïù¥Ïùò ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®ÏàòÎäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï†ïÏùòÎê©ÎãàÎã§:

$$
F(x) = \frac{x - \text{min}}{\text{max} - \text{min}}
$$

Ïó¨Í∏∞ÏÑú:

* $\text{min} = 0$ (ÏµúÏÜüÍ∞í)
* $\text{max} = 10$ (ÏµúÎåìÍ∞í)
* $x$Îäî Ïö∞Î¶¨Í∞Ä Íµ¨ÌïòÎ†§Îäî ÌäπÏ†ï Í∞íÏûÖÎãàÎã§.

---


---

### ü•õ $5\text{ml}$ÏóêÏÑú $6\text{ml}$ ÏÇ¨Ïù¥Ïùò ÌôïÎ•†ÏùÑ ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®ÏàòÎ°ú Í≥ÑÏÇ∞:

#### 1. $6\text{ml}$ Ïù¥ÌïòÏùº ÌôïÎ•† $F(6)$:
Í∑†ÏùºÎ∂ÑÌè¨Ïùò ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò(CDF) Í≥µÏãù $F(x) = \frac{x - \text{min}}{\text{max} - \text{min}}$ÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§. ($\text{min}=0$, $\text{max}=10$)

$$
F(6) = \frac{6 - 0}{10 - 0} = \frac{6}{10} = 0.6
$$

Ï¶â, Ïö∞Ïú†Ïùò ÏñëÏù¥ $6\text{ml}$ Ïù¥ÌïòÏùº ÌôïÎ•†ÏùÄ **$60\%$**ÏûÖÎãàÎã§.

---

#### 2. $5\text{ml}$ Ïù¥ÌïòÏùº ÌôïÎ•† $F(5)$:

$$
F(5) = \frac{5 - 0}{10 - 0} = \frac{5}{10} = 0.5
$$

Ï¶â, Ïö∞Ïú†Ïùò ÏñëÏù¥ $5\text{ml}$ Ïù¥ÌïòÏùº ÌôïÎ•†ÏùÄ **$50\%$**ÏûÖÎãàÎã§.

---

### Îëê Í∞íÏùò Ï∞®Î°ú ÌôïÎ•†ÏùÑ Í≥ÑÏÇ∞:

$$
P(5 \le X \le 6) = F(6) - F(5) = 0.6 - 0.5 = 0.1
$$

Îî∞ÎùºÏÑú, ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥ÎèÑ Ïö∞Ïú†Ïùò ÏñëÏù¥ $5\text{ml}$ÏóêÏÑú $6\text{ml}$ ÏÇ¨Ïù¥Ïùº ÌôïÎ•†ÏùÄ **$0.1 (10\%)$**Î°ú Í≥ÑÏÇ∞Îê©ÎãàÎã§.

---


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
plt.rcParams['font.family'] ='Malgun Gothic'

# Ïö∞Ïú†Ïùò Ïñë Î≤îÏúÑ ÏÑ§Ï†ï
min_value = 0
max_value = 10
x_values = np.linspace(min_value, max_value, 1000)

# SciPyÏùò uniform Í∞ùÏ≤¥ ÏÉùÏÑ± (Í∑†ÏùºÎ∂ÑÌè¨)
# loc = min_value, scale = max_value - min_value
uniform_dist = uniform(loc=min_value, scale=max_value - min_value)

# ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò (PDF)ÏôÄ ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò (CDF) Í∞í Í≥ÑÏÇ∞
# uniform.pdf(x, loc=a, scale=b-a) -> RÏùò dunif(x, min=a, max=b)ÏôÄ ÎèôÏùº
pdf_values = uniform_dist.pdf(x_values) # PDF

# uniform.cdf(x, loc=a, scale=b-a) -> RÏùò punif(x, min=a, max=b)ÏôÄ ÎèôÏùº
cdf_values = uniform_dist.cdf(x_values) # CDF

# 1. ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò (PDF) Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1) # 1Ìñâ 2Ïó¥ Ï§ë Ï≤´ Î≤àÏß∏ ÏòÅÏó≠
plt.plot(x_values, pdf_values, color="blue", linewidth=2)
plt.title("Í∑†ÏùºÎ∂ÑÌè¨ PDF (0, 10)")
plt.xlabel("x")
plt.ylabel("Î∞ÄÎèÑ")
plt.ylim(0, 0.2)
plt.grid(True, linestyle='--', alpha=0.6)

# 2. ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò (CDF) Í∑∏Î¶¨Í∏∞
plt.subplot(1, 2, 2) # 1Ìñâ 2Ïó¥ Ï§ë Îëê Î≤àÏß∏ ÏòÅÏó≠
plt.plot(x_values, cdf_values, color="red", linewidth=2)
plt.title("Í∑†ÏùºÎ∂ÑÌè¨ CDF (0, 10)")
plt.xlabel("x")
plt.ylabel("ÎàÑÏ†Å ÌôïÎ•†")
plt.ylim(0, 1)
plt.grid(True, linestyle='--', alpha=0.6)

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.tight_layout() # ÏÑúÎ∏åÌîåÎ°Ø Í∞ÑÏùò Í∞ÑÍ≤© ÏûêÎèô Ï°∞Ï†ï
plt.show()
```

```python
from scipy.stats import uniform

# Ïö∞Ïú†Ïùò Ïñë Î≤îÏúÑ ÏÑ§Ï†ï
min_value = 1
max_value = 10

# ÌäπÏ†ï Íµ¨Í∞Ñ ÏÑ§Ï†ï (Ïòà: 3mlÏóêÏÑú 7ml ÏÇ¨Ïù¥Ïùò ÌôïÎ•†)
lower_bound = 3
upper_bound = 7

# SciPyÏùò uniform Í∞ùÏ≤¥ ÏÉùÏÑ±
# loc = min_value, scale = max_value - min_value
uniform_dist = uniform(loc=min_value, scale=max_value - min_value)

# ÌäπÏ†ï Íµ¨Í∞ÑÏóê ÎåÄÌïú ÌôïÎ•† Í≥ÑÏÇ∞
# P(lower_bound <= X <= upper_bound) = CDF(upper_bound) - CDF(lower_bound)
probability = uniform_dist.cdf(upper_bound) - uniform_dist.cdf(lower_bound)

# Í≤∞Í≥º Ï∂úÎ†•
print(f"3mlÏóêÏÑú 7ml ÏÇ¨Ïù¥Ïùò ÌôïÎ•†: {probability}")
```

##  Ï†ïÍ∑úÎ∂ÑÌè¨(Normal Distribution) ÏÑ§Î™Ö

Ï†ïÍ∑úÎ∂ÑÌè¨(Normal Distribution)Îäî **Ïó∞ÏÜç ÌôïÎ•† Î∂ÑÌè¨ Ï§ë Í∞ÄÏû• ÎÑêÎ¶¨ ÏÇ¨Ïö©ÎêòÎäî Î∂ÑÌè¨** Ï§ë ÌïòÎÇòÏûÖÎãàÎã§. Ï†ïÍ∑úÎ∂ÑÌè¨Îäî **ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞®**Î•º Ïù¥Ïö©Ìï¥ ÌäπÏ†ï Îç∞Ïù¥ÌÑ∞Í∞Ä Ïñ¥ÎñªÍ≤å Î∂ÑÌè¨ÎêòÎäîÏßÄÎ•º ÎÇòÌÉÄÎÉÖÎãàÎã§.

---

### Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò

Ï†ïÍ∑úÎ∂ÑÌè¨Îäî Îã§ÏùåÍ≥º Í∞ôÏùÄ ÏãùÏúºÎ°ú ÌëúÌòÑÎê©ÎãàÎã§:

$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

Ïó¨Í∏∞ÏÑú:

* $\mu$Îäî **ÌèâÍ∑†(Mean)**
* $\sigma$Îäî **ÌëúÏ§ÄÌé∏Ï∞®(Standard Deviation)**

ÏûÖÎãàÎã§.

---

## Ï†ïÍ∑úÎ∂ÑÌè¨ ÌäπÏßï 

ÌèâÍ∑†Ïù¥ Îã§Î•¥Í≥† ÌëúÏ§ÄÌé∏Ï∞®ÎèÑ Îã§Î•¥ÏßÄÎßå,

* $N(50, 5^2)$ ÏóêÏÑú $50 \sim 55$ÍπåÏßÄÏùò ÎÑìÏù¥(ÌôïÎ•†)ÎÇò
* $N(30, 2^2)$ ÏóêÏÑú $30 \sim 32$ÍπåÏßÄÏùò ÎÑìÏù¥(ÌôïÎ•†) Í∞ôÏïÑÏöî!

Ï¶â, **Ï†ïÍ∑úÎ∂ÑÌè¨**ÏóêÏÑú $m$Ïù¥ Î™áÏù¥Îì† $\sigma$Í∞Ä Î™áÏù¥Îì†!

$$P(m \le X \le m+k\sigma)$$

ÏùÄ Ïñ∏Ï†úÎÇò **ÎèôÏùºÌï©ÎãàÎã§**!! ($k$Îäî Ï†ïÏàò)

Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò **ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò(PDF: Probability Density Function)**Îäî ÌäπÏ†ï Í∞í Ï£ºÏúÑÏóêÏÑú ÌôïÎ•†Ïù¥ ÏñºÎßàÎÇò ÏßëÏ§ëÎêòÏñ¥ ÏûàÎäîÏßÄÎ•º ÎÇòÌÉÄÎÉÖÎãàÎã§. Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò PDFÎäî ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞®Î•º Í∏∞Ï§ÄÏúºÎ°ú Ï¢Ö Î™®ÏñëÏùò Í≥°ÏÑ†ÏùÑ Í∞ÄÏßÄÎ©∞, ÎåÄÎ∂ÄÎ∂ÑÏùò Í∞íÏù¥ ÌèâÍ∑† Í∑ºÏ≤òÏóê Î∂ÑÌè¨Ìï©ÎãàÎã§.

### Ï†ïÍ∑úÎ∂ÑÌè¨ PDF ÏàòÏãù

Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï†ïÏùòÎê©ÎãàÎã§.

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

Ïó¨Í∏∞ÏÑú:

* $\mu$Îäî ÌèâÍ∑† (mean),
* $\sigma$Îäî ÌëúÏ§ÄÌé∏Ï∞® (standard deviation),
* $f(x)$Îäî ÌäπÏ†ï Í∞í $x$ÏóêÏÑúÏùò **ÌôïÎ•† Î∞ÄÎèÑ**Î•º ÎÇòÌÉÄÎÉÖÎãàÎã§.

PDFÎäî ÌäπÏ†ï Í∞í ÏûêÏ≤¥Ïùò ÌôïÎ•†ÏùÑ ÏùòÎØ∏ÌïòÎäî Í≤ÉÏùÄ ÏïÑÎãàÍ≥†, **ÌäπÏ†ï Íµ¨Í∞ÑÏóêÏÑú Í∞íÏù¥ ÎÇòÏò¨ Í∞ÄÎä•ÏÑ±**ÏùÑ ÎÇòÌÉÄÎÉÖÎãàÎã§. Ïù¥ Í∞íÏùÄ ÌäπÏ†ï Íµ¨Í∞ÑÏóê ÎåÄÌïú **Ï†ÅÎ∂Ñ**ÏùÑ ÌÜµÌï¥ Íµ¨Í∞Ñ ÌôïÎ•†Î°ú Î≥ÄÌôòÎê©ÎãàÎã§.

Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò **ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò(CDF: Cumulative Distribution Function)**Îäî ÌäπÏ†ï Í∞í Ïù¥ÌïòÏùò ÌôïÎ•†ÏùÑ ÎàÑÏ†ÅÏúºÎ°ú Í≥ÑÏÇ∞ÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§. CDFÎäî ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò(PDF)Î•º **Ï†ÅÎ∂Ñ**Ìïú Í≤∞Í≥ºÎ°ú, **ÌäπÏ†ï Í∞í $x$ Ïù¥ÌïòÏùò ÌôïÎ•†**ÏùÑ ÎÇòÌÉÄÎÉÖÎãàÎã§.

### Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò CDF ÏàòÏãù

Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®ÏàòÎäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï†ïÏùòÎê©ÎãàÎã§.

$$
F(x) = P(X \le x) = \int_{-\infty}^{x} f(t)dt
$$

Ïó¨Í∏∞ÏÑú $f(t)$Îäî Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò(PDF)ÏûÖÎãàÎã§. Ïù¥ Ï†ÅÎ∂ÑÏùÑ ÌíÄÎ©¥, ÌëúÏ§Ä Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò CDFÎäî Îã§ÏùåÍ≥º Í∞ôÏù¥ ÌëúÌòÑÎê©ÎãàÎã§:

$$
F(x) = \frac{1}{2} \left[1 + \text{erf} \left(\frac{x - \mu}{\sigma\sqrt{2}}\right)\right]
$$

Ïó¨Í∏∞ÏÑú:

* $\mu$Îäî ÌèâÍ∑† (mean),
* $\sigma$Îäî ÌëúÏ§ÄÌé∏Ï∞® (standard deviation),
* $\text{erf}(z)$Îäî **Ïò§Ï∞® Ìï®Ïàò**Î°ú, ÌäπÎ≥ÑÌïú Ìï®ÏàòÎ°ú Ï†ïÏùòÎêú Í∞íÏûÖÎãàÎã§.

Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò CDFÎäî ÏßÅÏ†ëÏ†ÅÏù∏ Ìï¥ÏÑùÏù¥ Ïñ¥Î†§Ïõå, Î≥¥ÌÜµ **Ï†ïÍ∑úÎ∂ÑÌè¨ ÌÖåÏù¥Î∏î**Ïù¥ÎÇò **Í≥ÑÏÇ∞ ÎèÑÍµ¨**Î•º ÏÇ¨Ïö©ÌïòÏó¨ Í∞íÏùÑ Í≥ÑÏÇ∞Ìï©ÎãàÎã§.



### 2. Íµ¨Í∞Ñ ÌôïÎ•† Í≥ÑÏÇ∞: PDF vs. CDF

Ï†ïÍ∑úÎ∂ÑÌè¨ÏóêÏÑú **PDF(ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò)**Î•º Ïù¥Ïö©ÌïòÏó¨ 70Ï†êÏóêÏÑú 100Ï†ê ÏÇ¨Ïù¥Ïùò ÌôïÎ•†ÏùÑ Í≥ÑÏÇ∞ÌïòÎäî Î∞©Î≤ïÏùÑ ÏÑ§Î™ÖÌïòÍ≤†ÏäµÎãàÎã§. PDFÎ°ú ÌäπÏ†ï Íµ¨Í∞ÑÏùò ÌôïÎ•†ÏùÑ Í≥ÑÏÇ∞ÌïòÎ†§Î©¥ Ìï¥Îãπ Íµ¨Í∞ÑÏóêÏÑúÏùò ÌôïÎ•† Î∞ÄÎèÑÎ•º **Ï†ÅÎ∂Ñ**Ìï¥Ïïº Ìï©ÎãàÎã§. ÌïòÏßÄÎßå ÏùºÎ∞òÏ†ÅÏúºÎ°ú Ïö∞Î¶¨Îäî **Ï†ÅÎ∂Ñ ÎåÄÏã† ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò(CDF)**Î•º ÏÇ¨Ïö©ÌïòÏó¨ Íµ¨Í∞ÑÏùò ÌôïÎ•†ÏùÑ Íµ¨Ìï©ÎãàÎã§. PDFÎäî ÌôïÎ•† Î∞ÄÎèÑÎ•º ÎÇòÌÉÄÎÇ¥Îäî Ìï®ÏàòÏù¥Í≥†, CDFÎäî ÎàÑÏ†Å ÌôïÎ•†ÏùÑ Í≥ÑÏÇ∞ÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§.

#### 1. PDFÎ•º ÏÇ¨Ïö©Ìïú ÏàòÏãù Ï†ÑÍ∞ú

Ï†ïÍ∑úÎ∂ÑÌè¨ÏóêÏÑú ÌäπÏ†ï Íµ¨Í∞Ñ $70 \le X \le 100$Ïùò ÌôïÎ•†ÏùÑ Í≥ÑÏÇ∞ÌïòÍ∏∞ ÏúÑÌïú Î∞©Î≤ïÏùÄ ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò(PDF)Î•º Ï†ÅÎ∂ÑÌïòÎäî Í≤ÉÏûÖÎãàÎã§:

$$
P(70 \le X \le 100) = \int_{70}^{100} f(x)dx
$$

Ïó¨Í∏∞ÏÑú, Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò $f(x)$Îäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï†ïÏùòÎê©ÎãàÎã§:

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$



#### 3. ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò(CDF)Î•º Ïù¥Ïö©Ìïú Í≥ÑÏÇ∞

ÌïòÏßÄÎßå ÏùºÎ∞òÏ†ÅÏúºÎ°ú Ïö∞Î¶¨Îäî Ï†ÅÎ∂Ñ ÎåÄÏã† **ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò(CDF)**Î•º ÏÇ¨Ïö©ÌïòÏó¨ Íµ¨Í∞ÑÏùò ÌôïÎ•†ÏùÑ Íµ¨Ìï©ÎãàÎã§. CDFÎ•º ÏÇ¨Ïö©ÌïòÎ©¥, ÌäπÏ†ï Íµ¨Í∞ÑÏùò ÌôïÎ•†ÏùÑ Îã§ÏùåÍ≥º Í∞ôÏù¥ Í∞ÑÎã®ÌïòÍ≤å Í≥ÑÏÇ∞Ìï† Ïàò ÏûàÏäµÎãàÎã§:

$$
P(70 \le X \le 100) = F(100) - F(70)
$$

Ïó¨Í∏∞ÏÑú $F(x)$Îäî Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®ÏàòÎ°ú, $x$ Í∞í Ïù¥ÌïòÏùò ÎàÑÏ†Å ÌôïÎ•†ÏùÑ ÎÇòÌÉÄÎÉÖÎãàÎã§. Ïù¥Î•º ÌÜµÌï¥ Ï†ÅÎ∂Ñ Í≥ÑÏÇ∞ÏùÑ Í∞ÑÎã®ÌïòÍ≤å Ìï† Ïàò ÏûàÏäµÎãàÎã§.

---


```python
import numpy as np

# ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞® ÏÑ§Ï†ï
mean_value = 100  # ÌèâÍ∑†
sd_value = 15     # ÌëúÏ§ÄÌé∏Ï∞®

# Í≤∞Í≥º Ïû¨ÌòÑÏÑ±ÏùÑ ÏúÑÌïú ÏãúÎìú ÏÑ§Ï†ï
np.random.seed(123) 

# Ï†ïÍ∑úÎ∂ÑÌè¨ÏóêÏÑú 10Í∞úÏùò ÏÉòÌîå ÏÉùÏÑ±
# locÏùÄ ÌèâÍ∑†(mean), scaleÏùÄ ÌëúÏ§ÄÌé∏Ï∞®(sd)Î•º ÏùòÎØ∏
samples_10 = np.random.normal(loc=mean_value, scale=sd_value, size=10)

# Í≤∞Í≥º Ï∂úÎ†•
print("--- 10Í∞ú ÏÉòÌîå Í≤∞Í≥º ---")
print(samples_10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'

# ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞® ÏÑ§Ï†ï
mean_value = 100  # ÌèâÍ∑†
sd_value = 15     # ÌëúÏ§ÄÌé∏Ï∞®

# Í≤∞Í≥º Ïû¨ÌòÑÏÑ±ÏùÑ ÏúÑÌïú ÏãúÎìú ÏÑ§Ï†ï
np.random.seed(123) 

# Ï†ïÍ∑úÎ∂ÑÌè¨ÏóêÏÑú 1000Í∞úÏùò ÏÉòÌîå ÏÉùÏÑ±
samples_1000 = np.random.normal(loc=mean_value, scale=sd_value, size=1000)

# ÌûàÏä§ÌÜ†Í∑∏Îû® Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(8, 5))
plt.hist(samples_1000, 
         bins=30,  # ÌûàÏä§ÌÜ†Í∑∏Îû®Ïùò Íµ¨Í∞Ñ Ïàò (breaks)
         color="lightblue",  # ÎßâÎåÄ ÏÉâÍπî (col)
         edgecolor="black")  # ÎßâÎåÄ ÌÖåÎëêÎ¶¨ ÏÉâÍπî (border)

plt.title("ÌèâÍ∑† 100, ÌëúÏ§ÄÌé∏Ï∞® 15Ïù∏ Ï†ïÍ∑úÎ∂ÑÌè¨ ÏÉòÌîå", fontsize=15) # Ï†úÎ™© (main)
plt.xlabel("Í∞í", fontsize=12) # XÏ∂ï ÎùºÎ≤® (xlab)
plt.ylabel("ÎπàÎèÑ", fontsize=12) # YÏ∂ï ÎùºÎ≤® (ylab)
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
from scipy.stats import norm

# ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞® ÏÑ§Ï†ï
mean_value = 100  # ÌèâÍ∑†
sd_value = 15     # ÌëúÏ§ÄÌé∏Ï∞®

# Íµ¨Í∞Ñ ÏÑ§Ï†ï
lower_bound = 70  # ÌïòÌïú
upper_bound = 100 # ÏÉÅÌïú

# 70Ï†êÏóêÏÑú 100Ï†ê ÏÇ¨Ïù¥Ïùò ÌôïÎ•† Í≥ÑÏÇ∞: P(70 < X < 100) = CDF(100) - CDF(70)
# pnorm(upper_bound) - pnorm(lower_bound)
probability = norm.cdf(upper_bound, loc=mean_value, scale=sd_value) - \
              norm.cdf(lower_bound, loc=mean_value, scale=sd_value)

# Í≤∞Í≥º Ï∂úÎ†•
print("--- Íµ¨Í∞Ñ ÌôïÎ•† Í≥ÑÏÇ∞ Í≤∞Í≥º ---")
print(f"70Ï†êÏóêÏÑú 100Ï†ê ÏÇ¨Ïù¥Ïùò ÌôïÎ•†: {probability:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞® ÏÑ§Ï†ï
mean_value = 100  # ÌèâÍ∑†
sd_value = 15     # ÌëúÏ§ÄÌé∏Ï∞®

# x Í∞í Î≤îÏúÑ ÏÑ§Ï†ï (ÌèâÍ∑†ÏóêÏÑú ¬±4 ÌëúÏ§ÄÌé∏Ï∞® Î≤îÏúÑÎ°ú ÏÑ§Ï†ï)
x_values = np.linspace(mean_value - 4 * sd_value, mean_value + 4 * sd_value, 1000)

# ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò Í∞í Í≥ÑÏÇ∞ (PDF)
# dnorm(x_values)
pdf_values = norm.pdf(x_values, loc=mean_value, scale=sd_value)

# Ï†ïÍ∑úÎ∂ÑÌè¨ Í≥°ÏÑ† Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(8, 5))
plt.plot(x_values, pdf_values, color="blue", linewidth=2, label="PDF Curve") # type = "l" (ÏÑ†)
         
plt.title("Ï†ïÍ∑úÎ∂ÑÌè¨ Í≥°ÏÑ† (ÌèâÍ∑† 100, ÌëúÏ§ÄÌé∏Ï∞® 15)", fontsize=15) # Ï†úÎ™© (main)
plt.xlabel("Í∞í", fontsize=12) # XÏ∂ï ÎùºÎ≤® (xlab)
plt.ylabel("ÌôïÎ•† Î∞ÄÎèÑ", fontsize=12) # YÏ∂ï ÎùºÎ≤® (ylab)

# ÌèâÍ∑† ÏúÑÏπòÏóê ÏàòÏßÅÏÑ† Ï∂îÍ∞Ä (abline(v = mean_value))
plt.axvline(x=mean_value, color="red", linestyle="--", linewidth=2, label="ÌèâÍ∑†") 

plt.legend()
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞® ÏÑ§Ï†ï
mean_value = 100  # ÌèâÍ∑†
sd_value = 15     # ÌëúÏ§ÄÌé∏Ï∞®

# x Í∞í Î≤îÏúÑ ÏÑ§Ï†ï (ÌèâÍ∑†ÏóêÏÑú ¬±4 ÌëúÏ§ÄÌé∏Ï∞® Î≤îÏúÑÎ°ú ÏÑ§Ï†ï)
# RÏùò seq()ÏôÄ Ïú†ÏÇ¨ÌïòÍ≤å ÏùºÏ†ïÌïú Í∞ÑÍ≤©Ïùò Î∞∞Ïó¥ ÏÉùÏÑ±
x_values = np.linspace(mean_value - 4 * sd_value, mean_value + 4 * sd_value, 1000)

# ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò Í∞í Í≥ÑÏÇ∞ (PDF)
# RÏùò dnorm() Ìï®ÏàòÎäî PythonÏùò norm.pdf() Ìï®ÏàòÎ°ú ÎåÄÏ≤¥
pdf_values = norm.pdf(x_values, loc=mean_value, scale=sd_value)

# Ï†ïÍ∑úÎ∂ÑÌè¨ Í≥°ÏÑ† Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(10, 6))

# 1. Ï†ïÍ∑úÎ∂ÑÌè¨ Í≥°ÏÑ† ÌîåÎ°Ø
plt.plot(x_values, pdf_values, color="blue", linewidth=2, label="PDF Í≥°ÏÑ†") 

# 2. 70Ï†êÏóêÏÑú 100Ï†ê ÏÇ¨Ïù¥Ïùò Íµ¨Í∞Ñ Ï±ÑÏö∞Í∏∞
lower_bound = 70
upper_bound = 100

# Ï±ÑÏö∞Í∏∞Ïóê ÏÇ¨Ïö©Ìï† x Í∞í Íµ¨Í∞ÑÏùÑ 70ÏóêÏÑú 100 ÏÇ¨Ïù¥Î°ú ÏÑ§Ï†ï
x_fill = np.linspace(lower_bound, upper_bound, 1000)

# Ìï¥Îãπ x Í∞íÏóê ÎåÄÌïú PDF Í∞í Í≥ÑÏÇ∞
y_fill = norm.pdf(x_fill, loc=mean_value, scale=sd_value)

# plt.fill_between() Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÏòÅÏó≠ Ï±ÑÏö∞Í∏∞
# where: xÍ∞Ä 70Í≥º 100 ÏÇ¨Ïù¥Ïóê ÏûàÏùÑ ÎïåÎßå Ï±ÑÏö∞Í∏∞
# alpha: Ìà¨Î™ÖÎèÑ ÏÑ§Ï†ï (RÏùò rgb(1, 0, 0, 0.4)ÏôÄ Ïú†ÏÇ¨)
plt.fill_between(x_fill, y_fill, 
                 color='red', alpha=0.4, 
                 label=f'{lower_bound} ~ {upper_bound} ÌôïÎ•†')


# 3. ÌèâÍ∑† ÏúÑÏπòÏóê ÏàòÏßÅÏÑ† Ï∂îÍ∞Ä
# RÏùò abline(v=mean_value)
plt.axvline(x=mean_value, color="red", linestyle="--", linewidth=2, label=f"ÌèâÍ∑† $\mu$={mean_value}") 

# 4. Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï (Ï†úÎ™©, ÎùºÎ≤® Îì±)
plt.title(f"Ï†ïÍ∑úÎ∂ÑÌè¨ Í≥°ÏÑ† (ÌèâÍ∑† {mean_value}, ÌëúÏ§ÄÌé∏Ï∞® {sd_value})\n{lower_bound} ~ {upper_bound}Ï†ê ÏÇ¨Ïù¥Ïùò ÌôïÎ•†", fontsize=15)
plt.xlabel("Ï†êÏàò", fontsize=12)
plt.ylabel("ÌôïÎ•† Î∞ÄÎèÑ", fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.5)
plt.show()
```
