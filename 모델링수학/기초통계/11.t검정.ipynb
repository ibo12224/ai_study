#  t분포란 무엇인가요?

t분포는 통계학에서 표본의 크기가 **작고** 모분산(모집단 전체의 분산)을 **모를 때** 주로 사용되는 확률분포입니다.

##  배경: 왜 t분포가 필요했을까요?

통계학의 근간에는 **정규분포(Normal Distribution)**가 있습니다. 표본의 크기가 충분히 크거나(보통 $n \ge 30$) 모분산을 알고 있다면, 표본 평균의 분포는 **중심극한 정리(Central Limit Theorem)**에 의해 정규분포를 따릅니다. 이때는 **Z-분포(표준정규분포)**를 사용하여 가설 검정이나 신뢰 구간 추정을 할 수 있습니다.

**하지만, 다음과 같은 현실적인 문제가 발생했습니다:**

1.  **표본의 크기가 작을 때:** 실험이나 조사가 어려워 표본을 많이 얻지 못할 때가 많습니다. ($n < 30$)
2.  **모분산을 모를 때:** 모집단 전체의 분산을 아는 경우는 거의 없습니다. 우리는 표본 분산을 대신 사용해야 합니다.

표본의 크기가 작고 표본 분산을 사용하게 되면, 기존의 Z-분포를 사용했을 때보다 **변동성**이 커지고 정확도가 떨어집니다. 이 문제를 해결하기 위해 **윌리엄 실리 고셋(William Sealy Gosset)**이라는 통계학자(기네스 맥주 공장 직원!)가 1908년에 'Student'라는 필명으로 t분포를 개발했습니다.

# 💡 t분포의 원리 및 특징

t분포는 Z-분포와 매우 유사하지만, 다음과 같은 중요한 특징들을 가집니다.

## 1. Z-분포 vs. t-분포

| 특징 | Z-분포 (표준정규분포) | t-분포 |
| :--- | :--- | :--- |
| **모분산 사용 여부** | 모분산($\sigma^2$)을 **알고 있을 때** | 모분산($\sigma^2$)을 **모를 때** |
| **분포 형태** | 항상 동일한 형태 | **자유도($\nu$)**에 따라 형태가 변함 |
| **꼬리** | 상대적으로 얇은 꼬리 | **Z-분포보다 두꺼운 꼬리** (Fat Tail) |

## 2. 두꺼운 꼬리 (Fat Tail)의 의미

<img src='https://encrypted-tbn2.gstatic.com/licensed-image?q=tbn:ANd9GcTuX3b4Ntj5c2OJe4pOhxfltz1ym0K9SVE9QTaun2Bm_HKQXDJR1nuiN30XsH3Fn0FfIFW0iS3bfQMWu8SAvMPAH23YDv9lciK5fl6WrBnZVBFAYYE'>


t분포는 Z-분포보다 중앙이 **낮고** 꼬리 부분이 **더 두껍습니다**.

* **원리:** 우리가 모분산 대신 표본 분산을 사용하기 때문에, 표본 분산이 모분산을 정확히 반영하지 못할 위험이 있습니다. 이 **불확실성** 때문에, t분포는 Z-분포보다 **극단적인 값(꼬리 부분)**이 나올 확률을 더 높게 책정합니다.
* **결과:** t분포를 사용하면 Z-분포를 사용할 때보다 더 **보수적인** 결론(예: 신뢰 구간이 더 넓어짐)을 내리게 됩니다.

## 3. 자유도 ($\nu$)의 역할

t분포의 가장 큰 특징은 **자유도($\nu$, degrees of freedom, df)**라는 매개변수에 의해 그 형태가 결정된다는 것입니다.

* **자유도의 정의:** 보통 표본의 크기에서 1을 뺀 값입니다.
    $$\nu = n - 1$$
    (여기서 $n$은 표본의 크기입니다.)
* **자유도의 증가에 따른 변화:**
    * **자유도가 작을수록:** 표본의 크기가 작다는 뜻이며, 불확실성이 크므로 꼬리가 **더 두꺼워집니다.**
    * **자유도가 커질수록:** 표본의 크기가 커진다는 뜻이며, 표본 분산이 모분산에 가까워지므로 t분포는 점점 **Z-분포(표준정규분포)의 형태**로 수렴합니다.

## 4. t-통계량의 계산

가설 검정 등에서 사용되는 **t-통계량($t$)**은 다음과 같은 공식으로 계산됩니다.

$$
t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
$$

* $\bar{X}$: 표본 평균
* $\mu$: 모집단의 평균 (가설에서 주장하는 값)
* $s$: **표본의 표준편차** (모집단의 표준편차($\sigma$) 대신 사용)
* $n$: 표본의 크기

이 공식에서 모표준편차($\sigma$) 대신 표본 표준편차($s$)를 사용하기 때문에, 이 통계량이 t분포를 따르게 됩니다.

#  요약 및 활용

t분포는 **작은 표본**과 **모분산을 모르는 상황**에서 불확실성을 고려하여 통계적 추론을 가능하게 하는 중요한 도구입니다.

* **언제 사용하나요?**
    * **t-검정(t-test):** 표본 평균이 특정 값과 같은지 다른지 검정할 때.
    * **신뢰 구간 추정:** 모평균이 포함될 구간을 추정할 때.

t분포 덕분에 우리는 제한된 데이터만으로도 합리적이고 신뢰할 수 있는 통계적 결론을 내릴 수 있습니다!

##  t분포 사용의 기준

t분포를 사용해야 하는 표본 개수의 **정확한 기준은 없습니다.** 하지만 통계학에서 일반적으로 사용되는 **경험적인 기준(rule of thumb)**이 있으며, 표본의 개수가 **30개 미만**일 때 t분포를 사용하는 것이 일반적입니다.


### 1. 경험적 기준: $n < 30$

* **표본의 크기가 작을 때 ($n < 30$):**
    * **모분산($\sigma^2$)을 모를 경우,** 표본 분산($s^2$)을 모분산 대신 사용하게 됩니다.
    * 표본의 크기가 작으면, 표본 분산이 모분산과 차이가 클 가능성이 높아지며, 이로 인해 발생하는 불확실성을 보정하기 위해 **t분포**를 사용합니다.

* **표본의 크기가 클 때 ($n \ge 30$):**
    * 표본의 크기가 충분히 크면 **중심극한 정리(Central Limit Theorem)**에 의해 표본 평균의 분포가 정규분포에 매우 가깝게 됩니다.
    * 또한, 표본 분산($s^2$)이 모분산($\sigma^2$)을 상당히 잘 추정하므로, 이때는 **Z-분포(표준정규분포)**를 사용해도 큰 문제가 없습니다. 실제로 자유도($\nu = n-1$)가 30 이상인 t분포는 Z-분포와 거의 구분할 수 없을 정도로 유사해집니다. 

### 2. 더 중요한 기준: 모분산의 유무와 분포 가정

t분포를 사용하는지 여부를 결정하는 더 근본적인 기준은 표본의 크기보다 **모분산의 유무**입니다.

| 상황 | 모분산 ($\sigma^2$) | 표본 분산 ($s^2$) | 사용하는 분포 |
| :--- | :--- | :--- | :--- |
| **이상적인 경우** | **앎** ($\sigma^2$ 사용) | - | **Z-분포** (표본 크기 무관) |
| **현실적인 경우** | **모름** ($\sigma^2$ 대신 $s^2$ 사용) | **사용** ($s^2$ 사용) | **t분포** (일반적으로 $n < 30$일 때) |

**요약하자면,** 표본의 크기가 작고 (대개 **30개 미만**이며) 모분산을 모를 때, 불확실성을 보정하기 위해 t분포를 사용합니다. 표본의 크기가 커질수록 (30개 이상) t분포는 Z-분포에 수렴하므로, 이 경우에는 Z-분포를 사용해도 무방합니다.



#  t분포의 확률 밀도 함수 (PDF) 공식

자유도(degrees of freedom)가 $\nu$인 **t분포**의 확률 변수 $T$에 대한 확률 밀도 함수 $f(t)$는 다음과 같습니다.

$$
f(t) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu\pi} \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu + 1}{2}}
$$

여기서 각 기호는 다음과 같은 의미를 가집니다.

* $f(t)$: t-통계량 $t$에서의 **확률 밀도 값**입니다.
* $t$: 확률 변수 (t-통계량의 값)로, 범위는 $-\infty < t < \infty$ 입니다.
* $\nu$ (뉴, nu): **자유도(degrees of freedom)**를 나타내며, 보통 표본 크기 $n$에 대해 $\nu = n-1$ 입니다. $\nu$는 양의 정수($\nu > 0$)입니다.
* $\pi$: 원주율 **파이**($\approx 3.14159$)입니다.
* $\Gamma$: **감마 함수(Gamma Function)**를 나타냅니다.

##  감마 함수 ($\Gamma$)란?

공식에 사용된 $\Gamma(z)$는 팩토리얼($n!$)의 개념을 실수와 복소수로 확장한 함수입니다. 통계학의 다양한 분포(정규분포, 카이제곱분포, F분포 등)의 공식에 사용되어 분포가 확률의 기본 규칙(총 면적 = 1)을 만족하도록 조정하는 역할을 합니다.

* 양의 정수 $n$에 대해서는 $\Gamma(n) = (n-1)!$ 의 관계가 있습니다.
* $\Gamma(1/2) = \sqrt{\pi}$ 입니다.

##  공식의 의미

이 복잡한 공식 전체를 외울 필요는 없지만, 핵심적인 구성 요소들을 통해 t분포의 특징을 알 수 있습니다.

1.  **앞쪽 항 (상수 부분):**
    $$
    \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu\pi} \Gamma\left(\frac{\nu}{2}\right)}
    $$
    이 부분은 $t$와 관련 없는 **상수**로, 전체 면적이 **1**이 되도록 만들어주는 **정규화 상수(Normalizing Constant)** 역할만 합니다.

2.  **뒤쪽 항 (핵심 변화 부분):**
    $$
    \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu + 1}{2}}
    $$
    * **$t^2$**: t-값이 0에서 멀어질수록(양쪽 꼬리로 갈수록) 이 값이 커집니다.
    * **$\nu$ (자유도)**: $\nu$ 값이 작으면(표본이 작으면), 분모 $\nu$가 작아져 전체 항 $(1 + t^2/\nu)$이 **더 빨리 커지고** 결과적으로 $f(t)$ 값은 **더 빨리 작아집니다.** (즉, 꼬리가 두꺼워짐)
    * **지수 $-\frac{\nu + 1}{2}$**: 음수 지수로 인해, 괄호 안의 값이 커질수록 전체 확률 밀도 값 $f(t)$는 빠르게 감소하게 됩니다.

---

t분포 공식은 수학적 배경일 뿐이며, 실제 통계 분석에서는 이 공식을 직접 계산하기보다는 통계 소프트웨어의 t분포표나 함수를 이용하여 $p$-값 또는 임계값($t_{\alpha}$)을 구하게 됩니다.

# 예시: 수업 성적의 평균 차이에 대한 t-검정

## 상황:

* 어떤 학교의 특정 과목의 **모집단 평균 성적이 75점**이라고 알려져 있습니다.
* 이 과목을 새롭게 가르치는 교수의 수업에서 **학생 10명의 표본**이 추출되었고, 이들의 **평균 성적은 80점**입니다.
* 이 교수님의 수업이 모집단의 평균 성적(75점)보다 더 높은 평균 성적을 기록했는지 여부를 확인하기 위해 **단일 표본 t-검정**을 수행합니다.
* 모집단의 표준편차는 알려져 있지 않으므로, **표본의 표준편차를 사용해야 하며 t분포를 사용**합니다.
* **표본 표준편차는 8점**입니다.

## 가설 설정:

* **귀무 가설 (H0):** 교수님의 수업 성적 평균은 모집단 평균과 같다. $H_0 : \mu = 75$
* **대립 가설 (H1):** 교수님의 수업 성적 평균은 모집단 평균보다 높다. $H_1 : \mu > 75$

#  단일 표본 t-검정 계산 과정

## 1. 주어진 정보 정리

t-검정을 위해 필요한 입력 값들을 정리합니다.

| 기호 | 설명 | 값 |
| :--- | :--- | :--- |
| $\mu_0$ | 귀무가설 하의 모집단 평균 (기준값) | 75점 |
| $\bar{X}$ | 표본 평균 | 80점 |
| $s$ | 표본 표준편차 | 8점 |
| $n$ | 표본 크기 | 10명 |

## 2. t-통계량 ($t$) 계산

t-통계량은 표본 평균($\bar{X}$)이 모집단 평균($\mu_0$)으로부터 얼마나 떨어져 있는지를 표준 오차($s/\sqrt{n}$)를 사용하여 표준화한 값입니다.

### 2.1. 표준 오차 (Standard Error, SE) 계산

표본 평균의 표준 오차($SE$)를 먼저 계산합니다.

$$
SE = \frac{s}{\sqrt{n}}
$$

$$
SE = \frac{8}{\sqrt{10}}
$$

$$
SE \approx \frac{8}{3.162} \approx 2.530
$$

### 2.2. t-통계량 계산

이제 표준 오차를 사용하여 t-통계량 $t$를 계산합니다.

$$
t = \frac{\bar{X} - \mu_0}{SE}
$$

$$
t = \frac{80 - 75}{2.530}
$$

$$
t = \frac{5}{2.530} \approx 1.976
$$

**결과:** t-통계량은 약 **1.976**입니다.

## 3. 자유도 ($\nu$) 결정

t분포의 형태를 결정하는 자유도($\nu$)를 계산합니다.

$$
\nu = n - 1
$$

$$
\nu = 10 - 1 = 9
$$

**결과:** 자유도는 **9**입니다.

## 4. 결론 도출 (p-값 또는 임계값 비교)

이제 계산된 $t = 1.976$과 $\nu = 9$를 사용하여 통계적 유의성을 판단해야 합니다.

가설은 $\mathbf{H_1 : \mu > 75}$인 **단측 검정(One-tailed test)**입니다. 일반적으로 유의수준 $\alpha = 0.05$를 사용한다고 가정하겠습니다.

### 4.1. 임계값 (Critical Value) 비교법

* **자유도 $\nu=9$** 이고 **단측 유의수준 $\alpha=0.05$** 일 때, t분포표를 찾아보면 임계값($t_{critical}$)은 약 **1.833**입니다.



* **비교:** 계산된 $t$-통계량과 임계값을 비교합니다.
    $$
    |t_{계산}| \,\, \text{vs.} \,\, |t_{임계}|
    $$
    $$
    1.976 \,\, > \,\, 1.833
    $$

* **판단:** 계산된 t-통계량($1.976$)이 임계값($1.833$)보다 **크므로** 귀무가설($H_0$)을 **기각합니다**.

### 4.2. p-값 (p-value) 비교법

* $\nu=9$일 때 $t=1.976$에 해당하는 단측 p-값을 통계 프로그램에서 찾으면, p-값은 약 **0.0398**입니다.

* **비교:** p-값과 유의수준 $\alpha=0.05$를 비교합니다.
    $$
    \text{p-값} \,\, \text{vs.} \,\, \alpha
    $$
    $$
    0.0398 \,\, < \,\, 0.05
    $$

* **판단:** p-값이 유의수준($0.05$)보다 **작으므로** 귀무가설($H_0$)을 **기각합니다**.

## 5. 최종 해석

계산 결과, 유의수준 5% 하에서 귀무가설($H_0: \mu=75$)을 기각하고 대립가설($H_1: \mu>75$)을 채택합니다.

> **결론:** 이 교수님의 수업 성적 평균(80점)은 모집단의 평균 성적(75점)보다 **통계적으로 유의하게 높다고** 볼 수 있습니다.

##  t분포표와 Z분포표의 차이점

| 특징 | t분포표 (Student's t-distribution Table) | Z분포표 (Standard Normal Table) |
| :--- | :--- | :--- |
| **분포 가정** | **모분산을 모를 때** 사용 (표본 표준편차 $s$ 사용) | **모분산을 알 때** 사용 |
| **결정 요인** | **자유도($\nu$)** | **단일한 분포** (자유도 불필요) |
| **형태** | 자유도에 따라 형태가 변하며, Z분포보다 **꼬리가 두껍습니다** (Fat Tail) | 항상 동일한 **표준 정규 분포** 형태입니다. |
| **표 구성** | **자유도(df)**를 나타내는 행이 존재합니다. | **Z-점수**를 나타내는 행/열만 존재합니다. |

---

* **수렴:** 표본 크기가 커져 **자유도가 무한대($\nu \to \infty$)에 가까워지면**, t분포는 표준정규분포(Z분포)에 수렴하며, t분포표의 맨 마지막 행(자유도 $\infty$)의 값은 Z분포표의 값과 **동일**합니다.

### 일표본 단측 t-검정 수행

```python
import numpy as np
from scipy import stats

# 1. 샘플 데이터 (지우개 무게, 예시 데이터)
# R의 c() 벡터와 동일
sample_data = np.array([51.2, 49.8, 50.1, 50.3, 49.9, 50.7, 50.5, 50.4])

# 2. 일표본 단측 t-검정 (H₁: 평균 > 50)
# t.test(sample_data, mu = 50, alternative = "greater") 와 동일
# scipy.stats.ttest_1samp 함수를 사용하며, alternative 인수를 "greater"로 설정하여 단측 검정 수행
# H₀: mu = 50
# H₁: mu > 50
t_statistic, p_value = stats.ttest_1samp(
    a=sample_data, 
    popmean=50, 
    alternative='greater' # 단측 검정: 대립 가설이 "평균이 50보다 크다"
)

# 3. 결과 출력
print("--- 일표본 단측 t-검정 결과 (H₁: mu > 50) ---")
print(f"샘플 크기 (n): {len(sample_data)}")
print(f"표본 평균: {np.mean(sample_data):.4f}")
print(f"귀무가설 평균 (mu): 50")
print("-" * 35)
print(f"t-통계량 (t-statistic): {t_statistic:.4f}")
print(f"p-값 (p-value): {p_value:.4f}")
print("-" * 35)

# 4. 통계적 해석 (참고용)
alpha = 0.05
print(f"유의수준 (alpha): {alpha}")
if p_value < alpha:
    print("해석: p-값이 유의수준(0.05)보다 작으므로 귀무가설(H₀: mu=50)을 기각하고, 대립가설(H₁: mu>50)을 채택합니다.")
else:
    print("해석: p-값이 유의수준(0.05)보다 크므로 귀무가설(H₀: mu=50)을 기각할 수 없습니다.")
```

### 일표본 양측 t-검정 수행

```python
import numpy as np
from scipy import stats

# 1. 샘플 데이터 (지우개 무게, 예시 데이터)
sample_data = np.array([51.2, 49.8, 50.1, 50.3, 49.9, 50.7, 50.5, 50.4])

# 2. 일표본 양측 t-검정 (H₁: 평균 ≠ 50)
# t.test(sample_data, mu = 50, alternative = "two.sided") 와 동일
# H₀: mu = 50
# H₁: mu ≠ 50
t_statistic, p_value = stats.ttest_1samp(
    a=sample_data, 
    popmean=50, 
    alternative='two-sided' # 양측 검정: 대립 가설이 "평균이 50이 아니다"
)

# 3. 결과 출력
print("--- 일표본 양측 t-검정 결과 (H₁: mu ≠ 50) ---")
print(f"샘플 크기 (n): {len(sample_data)}")
print(f"표본 평균: {np.mean(sample_data):.4f}")
print(f"귀무가설 평균 (mu): 50")
print("-" * 35)
print(f"t-통계량 (t-statistic): {t_statistic:.4f}")
print(f"p-값 (p-value): {p_value:.4f}")
print(f"자유도 (df): {len(sample_data) - 1}") # 자유도 추가
print("-" * 35)

# 4. 통계적 해석 (참고용)
alpha = 0.05
print(f"유의수준 (alpha): {alpha}")
if p_value < alpha:
    print("해석: p-값이 유의수준(0.05)보다 작으므로 귀무가설(H₀: mu=50)을 기각하고, 대립가설(H₁: mu≠50)을 채택합니다.")
else:
    print("해석: p-값이 유의수준(0.05)보다 크므로 귀무가설(H₀: mu=50)을 기각할 수 없습니다.")
```

### 독립 이표본 단측 t-검정 수행

```python
import numpy as np
from scipy import stats

# 1. 샘플 데이터 (예시 데이터)
group1 = np.array([52.1, 51.8, 50.5, 52.4, 51.7])  # 첫 번째 집단 데이터
group2 = np.array([49.8, 50.1, 50.2, 49.5, 50.0])  # 두 번째 집단 데이터

# 2. 이표본 단측 t-검정 (H₁: group1의 평균 > group2의 평균)
# t.test(group1, group2, alternative = "greater") 와 동일
# H₀: mu1 = mu2 (두 그룹 평균은 같다)
# H₁: mu1 > mu2 (group1 평균이 group2 평균보다 크다)
# equal_var=True는 기본값으로, 두 그룹의 분산이 같다고 가정하는 '등분산 t-검정'을 수행합니다.
t_statistic, p_value = stats.ttest_ind(
    a=group1,
    b=group2,
    equal_var=True,      # 분산이 같다고 가정 (R의 기본값)
    alternative='greater'  # 단측 검정: 대립 가설이 "a의 평균 > b의 평균"
)

# 3. 결과 출력
print("--- 독립 이표본 단측 t-검정 결과 (H₁: mu1 > mu2) ---")
print(f"Group 1 (n={len(group1)}) 평균: {np.mean(group1):.4f}")
print(f"Group 2 (n={len(group2)}) 평균: {np.mean(group2):.4f}")
print("-" * 35)
print(f"t-통계량 (t-statistic): {t_statistic:.4f}")
print(f"p-값 (p-value): {p_value:.4f}")
# 두 표본의 자유도는 (n1 + n2 - 2)
df = len(group1) + len(group2) - 2
print(f"자유도 (df): {df}")
print("-" * 35)

# 4. 통계적 해석 (참고용)
alpha = 0.05
print(f"유의수준 (alpha): {alpha}")
if p_value < alpha:
    print("해석: p-값이 유의수준(0.05)보다 작으므로 귀무가설(H₀: mu1=mu2)을 기각하고, 대립가설(H₁: mu1>mu2)을 채택합니다.")
else:
    print("해석: p-값이 유의수준(0.05)보다 크므로 귀무가설(H₀: mu1=mu2)을 기각할 수 없습니다.")
```
