## 1. 확률의 기본 개념 (Basic Concepts of Probability)

**확률**은 어떤 사건이 일어날 가능성을 **수치로 나타낸 것**입니다. 확률은 **0에서 1 사이의 값**을 가지며, 1에 가까울수록 사건이 일어날 가능성이 큽니다.

---

### 예시 1: 주사위 굴리기

주사위를 굴렸을 때 특정 숫자가 나올 확률을 생각해 봅시다. 주사위에는 6개의 면(1, 2, 3, 4, 5, 6)이 있으므로, 각 숫자가 나올 확률은 다음과 같습니다:

* 1이 나올 확률 = $\frac{1}{6}$
* 2가 나올 확률 = $\frac{1}{6}$
* ... 6까지 각 숫자가 나올 확률은 동일하게 $\frac{1}{6}$입니다. 
---

## 2. 조건부 확률의 개념 (Concept of Conditional Probability)

**조건부 확률**은 특정한 조건이 주어졌을 때, 어떤 사건이 일어날 확률을 의미합니다. 예를 들어, **사건 A가 발생했을 때 사건 B가 발생할 확률**을 구하는 것입니다. 
이를 수식으로 표현하면 다음과 같습니다:

$$P(B|A) = \frac{P(A \cap B)}{P(A)}$$

여기서 $P(B|A)$는 **"A가 일어난 후 B가 일어날 확률"**을 의미합니다.

---

### 예시 2: 카드 뽑기

52장의 카드 덱에서 카드를 한 장 뽑는다고 해 봅시다. 여기서 **A**는 뽑은 카드가 "**빨간 카드**"일 확률이고, **B**는 그 카드가 "**하트**"일 확률이라고 가정해 봅시다.

* 덱에 빨간 카드(하트와 다이아몬드)는 총 **26장**입니다.
* 덱에 하트 카드는 총 **13장**입니다.

**조건부 확률**을 계산할 수 있습니다. 만약 뽑은 카드가 **빨간 카드**일 때, 그 카드가 **하트**일 확률은?

$$P(B|A) = \frac{P(A \cap B)}{P(A)}$$

* 빨간 카드이면서 하트인 경우(즉, $A \cap B$)는 **13장**이며, $P(A \cap B) = \frac{13}{52}$입니다.
* 빨간 카드일 확률 $P(A)$는 $\frac{26}{52}$입니다.

따라서 조건부 확률은 다음과 같습니다:

$$P(B|A) = \frac{\frac{13}{52}}{\frac{26}{52}} = \frac{13}{26} = \frac{1}{2}$$

즉, 뽑은 카드가 빨간 카드($A$)라는 조건 하에, 그 카드가 하트($B$)일 확률은 $\frac{1}{2}$입니다.

---

## 3. 독립사건 (Independent Events)

**독립사건**은 한 사건이 일어날 확률이 다른 사건의 결과에 영향을 받지 않는 경우를 말합니다. 두 사건 A와 B가 독립사건일 때, A가 일어났다고 해서 B가 일어날 확률이 바뀌지 않습니다. 수식으로 표현하면 다음과 같습니다:

$$P(A \cap B) = P(A) \times P(B)$$

즉, 두 사건이 동시에 일어날 확률은 각각의 사건이 일어날 확률의 곱입니다.

---

### 예시: 주사위 굴리기와 동전 던지기

* **사건 A**: 주사위를 굴려 6이 나올 확률 ($\frac{1}{6}$)
* **사건 B**: 동전을 던져 앞면이 나올 확률 ($\frac{1}{2}$)

주사위를 굴리는 것과 동전을 던지는 것은 서로 영향을 미치지 않는 **독립사건**입니다. 따라서, 주사위에서 6이 나오고 동전에서 앞면이 나올 확률은:

$$P(A \cap B) = P(A) \times P(B) = \frac{1}{6} \times \frac{1}{2} = \frac{1}{12}$$

즉, 주사위에서 6이 나오고 동시에 동전에서 앞면이 나올 확률은 $\frac{1}{12}$입니다.

## 4. 배반사건 (Mutually Exclusive Events)

**배반사건**은 두 사건이 서로 동시에 일어날 수 없는 경우를 말합니다. 즉, 한 사건이 일어나면 다른 사건은 반드시 일어나지 않는 경우입니다. 두 사건 A와 B가 배반사건일 때, 수식으로는 다음과 같이 표현할 수 있습니다:

$$P(A \cap B) = 0$$

즉, A와 B가 동시에 일어날 확률이 **0**입니다. 
---

### 예시: 카드 뽑기

52장의 카드 덱에서 한 장의 카드를 뽑는 상황을 생각해 봅시다.

* **사건 A**: 뽑은 카드가 "**하트**"일 확률
* **사건 B**: 뽑은 카드가 "**스페이드**"일 확률

하트와 스페이드는 서로 다른 무늬이기 때문에, 같은 카드에서 하트이면서 스페이드일 수는 없습니다. 따라서 **사건 A와 사건 B는 배반사건**입니다. 이때 두 사건이 동시에 일어날 확률은 **0**입니다.

$$P(A \cap B) = 0$$

즉, 카드 한 장을 뽑았을 때, 그 카드가 하트이면서 동시에 스페이드일 확률은 **0**입니다.
