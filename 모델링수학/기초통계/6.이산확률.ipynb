# 이산확률분포

## 베르누이

## 베르누이 분포의 특징

* **확률변수 $X$**: 성공(1) 또는 실패(0)라는 두 가지 값만 가질 수 있는 확률변수입니다.
* **성공의 확률 $p$**: 성공(1)이 나올 확률을 $p$라고 하고, 실패(0)의 확률은 $1 - p$입니다.
* 이 분포는 한 번의 시행에서 두 가지 결과 중 하나가 나오는 사건의 확률을 설명합니다.

---

## 확률질량함수(PMF)

베르누이 분포의 확률질량함수($\text{PMF}$)는 다음과 같이 정의됩니다.

$$
P(X = x) = \begin{cases}
p & \text{if } x = 1 \\
1 - p & \text{if } x = 0
\end{cases}
$$

**여기서:**

* $P(X = 1) = p$: 성공(1)이 나올 확률.
* $P(X = 0) = 1 - p$: 실패(0)이 나올 확률.

## 예시: 동전 던지기 

동전을 던지는 실험에서 **앞면이 나올 확률을 $0.5$**로 가정해보겠습니다.

이때 동전 던지기의 결과는 **앞면(성공, $1$)** 또는 **뒷면(실패, $0$)** 두 가지 중 하나입니다. 이 실험은 **베르누이 분포**를 따릅니다.

* **$X = 1$ (앞면)**: 확률은 $P(X = 1) = 0.5$
* **$X = 0$ (뒷면)**: 확률은 $P(X = 0) = 1 - 0.5 = 0.5$


---

## 1. 기댓값 (Expected Value)

기댓값은 **확률변수의 평균**을 의미합니다. 베르누이 분포에서 기댓값은 **성공할 확률 $p$**에 의해 결정됩니다. 
베르누이 분포의 확률변수 $X$는 성공($1$)과 실패($0$) 두 가지 값만 가질 수 있기 때문에, 기댓값은 다음과 같이 계산됩니다.

$$
E(X) = 1 \cdot P(X = 1) + 0 \cdot P(X = 0) = 1 \cdot p + 0 \cdot (1 - p) = p
$$

즉, 베르누이 분포의 기댓값은 **성공 확률 $p$**입니다.

---

## 예시: 동전 던지기 

동전을 던지는 실험에서 **앞면이 나올 확률을 $0.5$**로 가정해보겠습니다.

이때 동전 던지기의 결과는 **앞면(성공, $1$)** 또는 **뒷면(실패, $0$)** 두 가지 중 하나이며, 이 실험은 베르누이 분포를 따릅니다.

* $X = 1$ (앞면): 확률은 $P(X = 1) = 0.5$
* $X = 0$ (뒷면): 확률은 $P(X = 0) = 1 - 0.5 = 0.5$

### 예시의 기댓값 계산

위 예시에서 성공 확률 $p = 0.5$이므로, 기댓값 $E(X)$는 다음과 같습니다.

$$
E(X) = p = 0.5
$$

---


## 2. 분산 (Variance)

분산은 확률변수가 평균으로부터 얼마나 퍼져 있는지를 나타냅니다. 베르누이 분포의 분산은 다음 공식으로 계산됩니다.

$$
\text{Var}(X) = E(X^2) - (E(X))^2
$$

* **$E(X^2)$ 계산**: 베르누이 분포에서 $X^2$는 $X$가 $0$ 또는 $1$이기 때문에, $E(X^2) = E(X) = p$가 됩니다.
* **따라서 분산은**:

$$
\text{Var}(X) = p - p^2 = p(1 - p)
$$

즉, 베르누이 분포의 분산은 **성공 확률 $p$**와 **실패 확률 $1 - p$**의 곱으로 표현됩니다.

---

## 요약

* **기댓값 $E(X) = p$**: 성공 확률.
* **분산 $\text{Var}(X) = p(1 - p)$**: 성공 확률과 실패 확률의 곱.

### 예시의 기댓값 및 분산 계산

동전 던지기의 경우, 성공 확률 $p = 0.5$일 때, 기댓값과 분산은 다음과 같습니다:

* $E(X) = 0.5$
* $\text{Var}(X) = 0.5(1 - 0.5) = 0.25$

```python
import matplotlib.pyplot as plt
import pandas as pd

# 성공 확률 p
p = 0.5

# X = 0 (실패) 또는 X = 1 (성공)일 확률
x = [0, 1]
probabilities = [1 - p, p]

# 베르누이 분포 시각화
plt.figure(figsize=(7, 5)) # 그래프 크기 설정
plt.bar(x, probabilities, 
        tick_label=["Failure (0)", "Success (1)"], 
        color="skyblue") # 색상 설정
plt.xlabel("Outcome") # x축 레이블
plt.ylabel("Probability") # y축 레이블
plt.title(f"Bernoulli Distribution with p = {p}") # 그래프 제목
plt.ylim(0, 1) # y축 범위 설정 (확률이므로 0과 1 사이)
plt.show() # 그래프 출력

# 데이터 출력
data = pd.DataFrame({
    "Outcome": x, 
    "Probability": probabilities
})
print(data)
```

```python
# 성공 확률 p
p = 0.5

# 베르누이 분포의 기대값 (E[X] = p)
expected_value = p

# 베르누이 분포의 분산 (Var[X] = p * (1 - p))
variance = p * (1 - p)

# 결과 출력
print(f"기대값 (Expected Value): {expected_value}")
print(f"분산 (Variance): {variance}")
```

###  베르누이 분포의 특성

  * **기대값 (Expected Value)**: $E[X] = p$

      * 베르누이 시행에서 \*\*성공(1)\*\*이 나올 확률 \*\*$p$\*\*가 그대로 기대값이 됩니다. (실패는 $0$의 값을 가집니다.)

  * **분산 (Variance)**: $Var[X] = p(1-p)$

      * 분산은 결과가 기대값으로부터 얼마나 퍼져 있는지를 나타냅니다. $p=0.5$일 때 $Var[X] = 0.5 \times 0.5 = 0.25$ 로 최대가 됩니다.


## 이항분포

##  이항분포 (Binomial distribution)

**이항분포 (Binomial distribution)** 는 베르누이 시행을 여러 번 반복했을 때 성공하는 횟수를 나타내는 확률분포입니다. 베르누이 분포가 성공/실패 두 가지 결과 중 하나를 나타낸다면, 이항분포는 그 베르누이 시행을 여러 번 반복했을 때 **성공의 횟수**를 설명하는 분포입니다.

---

### 이항분포의 특징:

* **성공 또는 실패**라는 두 가지 결과가 있는 사건에서 사용됩니다.
* $n$ 번의 **독립적인 시행**을 반복하며, 각 시행에서 성공할 확률은 $p$ 입니다.
* 이항분포는 성공의 **총 횟수**를 나타냅니다.

---

### 이항분포의 확률질량함수 (PMF):

$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

**여기서:**

* $n$: 총 시행 횟수
* $k$: 성공 횟수
* $p$: 성공할 확률
* $1-p$: 실패할 확률
* $\binom{n}{k}$: 조합 ($n$ 번 중 $k$ 번 성공하는 경우의 수)

---


##  예시: 동전 던지기

동전을 5번 던져서 **앞면(성공)** 이 나오는 횟수를 구하는 상황을 생각해봅시다. 동전 던지기의 성공 확률 $p = 0.5$ 이고, 실패 확률 $1 - p = 0.5$ 입니다.

* **$n = 5$**: 동전을 **5번 던집**니다.
* **$p = 0.5$**: 앞면이 나올 **확률은 50%** 입니다.
* 이항분포는 5번 던졌을 때 **앞면이 나오는 횟수**를 나타냅니다.

---


##  조합 (Combination)

조합은 주어진 집합에서 **순서를 고려하지 않고** 특정 개수만큼의 원소를 선택하는 방법의 수를 계산하는 수학적 방법입니다. 조합의 계산법은 $*n$ 개 중에서 $k$ 개를 선택하는 경우의 수*를 의미하며, 이를 **이항계수 (binomial coefficient)** 라고도 부릅니다.

---

### 조합의 공식

조합의 공식은 다음과 같습니다:

$$\binom{n}{k} = \frac{n!}{k!(n-k)!}$$

**여기서:**

* $n$: 전체 원소의 개수
* $k$: 선택할 원소의 개수
* $n!$: $n$ 의 **팩토리얼** (계산: $n \times (n-1) \times (n-2) \times \dots \times 1$)
* $k!$: $k$ 의 팩토리얼
* $(n-k)!$: $(n-k)$ 의 팩토리얼

---

### 예시 1: $\binom{5}{2}$ 계산

5개의 원소 중에서 2개를 선택하는 방법의 수를 계산해 보겠습니다.

$$\binom{5}{2} = \frac{5!}{2!(5-2)!} = \frac{5 \times 4 \times 3 \times 2 \times 1}{(2 \times 1)(3 \times 2 \times 1)} = \frac{120}{2 \times 6} = \frac{120}{12} = 10$$

즉, 5개의 원소 중에서 2개를 선택하는 방법은 **10가지**입니다.

---

조합을 이항분포의 확률질량함수($P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$)에 적용하여 동전 던지기 같은 이항분포 문제를 풀어볼까요?

##  동전 던지기 예시: 확률 계산 과정

동전을 5번 던졌을 때 앞면($k$ 번)이 나올 확률을 계산하는 과정을 이항분포의 확률질량함수($P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$)를 사용하여 살펴보겠습니다. 이때 **성공 확률** $p = 0.5$, **전체 시행 횟수** $n = 5$ 입니다.

---

### 1. $k = 0$: 앞면이 0번 나올 확률

$$P(X = 0) = \binom{5}{0} (0.5)^0 (1-0.5)^{5-0} = 1 \times (0.5)^5 = 1 \times 0.03125 = 0.03125$$

즉, 앞면이 0번 나올 확률은 **0.03125**입니다.

---

### 2. $k = 1$: 앞면이 1번 나올 확률

$$P(X = 1) = \binom{5}{1} (0.5)^1 (1-0.5)^{5-1} = 5 \times (0.5)^1 \times (0.5)^4 = 5 \times 0.5 \times 0.0625 = 0.15625$$

즉, 앞면이 1번 나올 확률은 **0.15625**입니다.

---

### 3. $k = 2$: 앞면이 2번 나올 확률

$$P(X = 2) = \binom{5}{2} (0.5)^2 (1-0.5)^{5-2} = 10 \times (0.5)^2 \times (0.5)^3 = 10 \times 0.25 \times 0.125 = 0.3125$$

즉, 앞면이 2번 나올 확률은 **0.3125**입니다.

---

### 4. $k = 3$: 앞면이 3번 나올 확률

$$P(X = 3) = \binom{5}{3} (0.5)^3 (1-0.5)^{5-3} = 10 \times (0.5)^3 \times (0.5)^2 = 10 \times 0.125 \times 0.25 = 0.3125$$

즉, 앞면이 3번 나올 확률은 **0.3125**입니다.

---

### 5. $k = 4$: 앞면이 4번 나올 확률

$$P(X = 4) = \binom{5}{4} (0.5)^4 (1-0.5)^{5-4} = 5 \times (0.5)^4 \times (0.5)^1 = 5 \times 0.0625 \times 0.5 = 0.15625$$

즉, 앞면이 4번 나올 확률은 **0.15625**입니다.

---

### 6. $k = 5$: 앞면이 5번 나올 확률

$$P(X = 5) = \binom{5}{5} (0.5)^5 (1-0.5)^{5-5} = 1 \times (0.5)^5 \times (0.5)^0 = 1 \times 0.03125 \times 1 = 0.03125$$

즉, 앞면이 5번 나올 확률은 **0.03125**입니다.

---

이 계산 결과를 모두 합하면 $0.03125 + 0.15625 + 0.3125 + 0.3125 + 0.15625 + 0.03125 = 1.0$ 이 되어, 모든 경우의 확률 합이 **1**임을 확인할 수 있습니다.



```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
import pandas as pd

# 총 시행 횟수 n
n = 5

# 성공 확률 p
p = 0.5

# 가능한 성공 횟수 k
k = np.arange(0, n + 1)

# 이항분포 확률 계산
probabilities = binom.pmf(k, n, p)

# 시각화
plt.bar(k, probabilities)
plt.xlabel("Number of Successes (k)")
plt.ylabel("Probability")
plt.title("Binomial Distribution (n = 5, p = 0.5)")
plt.show()

# 데이터 출력
df = pd.DataFrame({"Successes": k, "Probability": probabilities})
df
```
