##  Îã®Ïàú ÏÑ†Ìòï ÌöåÍ∑Ä Í≥µÏãù

Í∏∞Ïö∏Í∏∞($w$)ÏôÄ Ï†àÌé∏($b$)ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏòàÏ∏°Í∞í $\hat{Y}$Î•º ÎÇòÌÉÄÎÇ¥Îäî Í≥µÏãù:

$$
\hat{Y}_i = w X_i + b

### üîç Í≥µÏãù ÏöîÏÜå ÏÑ§Î™Ö

| Í∏∞Ìò∏ | ÏùòÎØ∏ | ÏÑ§Î™Ö |
| :--- | :--- | :--- |
| $\hat{Y}_i$ | $i$Î≤àÏß∏ Îç∞Ïù¥ÌÑ∞Ïùò **ÏòàÏ∏°Í∞í** (Predicted Value) | Î™®Îç∏Ïù¥ ÏûÖÎ†• $X_i$Ïóê ÎåÄÌï¥ ÏòàÏ∏°Ìïú Ï¢ÖÏÜç Î≥ÄÏàòÏùò Í∞í. |
| $X_i$ | $i$Î≤àÏß∏ Îç∞Ïù¥ÌÑ∞Ïùò **ÎèÖÎ¶Ω Î≥ÄÏàò** (Independent Variable) | ÏòàÏ∏°ÏùÑ ÏàòÌñâÌïòÎäî Îç∞ ÏÇ¨Ïö©ÎêòÎäî ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞(ÌäπÏßï)Ïùò Í∞íÏûÖ. |
| $w$ | **Í∞ÄÏ§ëÏπò/Í∏∞Ïö∏Í∏∞** (Weight or Slope) | $X$Í∞Ä Ìïú Îã®ÏúÑ Î≥ÄÌï† Îïå $\hat{Y}$Í∞Ä ÏñºÎßàÎÇò Î≥ÄÌïòÎäîÏßÄÎ•º ÎÇòÌÉÄÎÉÖÎãàÎã§. Î™®Îç∏ ÌïôÏäµÏùÑ ÌÜµÌï¥ Í≤∞Ï†ïÎêòÎäî ÌïµÏã¨ Í≥ÑÏàò. |
| $b$ | **Ï†àÌé∏** (Bias or Intercept) | $X$Í∞Ä 0Ïùº Îïå $\hat{Y}$Ïùò Í∞í. Î™®Îç∏ ÌïôÏäµÏùÑ ÌÜµÌï¥ Í≤∞Ï†ïÎêòÎäî Í∞í. |

##  MSE(ÌèâÍ∑† Ï†úÍ≥± Ïò§Ï∞®) Í≥µÏãù

MSEÎäî ÌöåÍ∑Ä Î∂ÑÏÑùÏóêÏÑú Î™®Îç∏Ïùò ÏòàÏ∏°Í∞íÍ≥º Ïã§Ï†úÍ∞í ÏÇ¨Ïù¥Ïùò **Ïò§Ï∞®Ïùò Ï†úÍ≥±Ïóê ÎåÄÌïú ÌèâÍ∑†**ÏùÑ ÎÇòÌÉÄÎÇ¥Îäî ÏßÄÌëúÏûÖÎãàÎã§.

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2
$$

---

### Í≥µÏãù ÏöîÏÜå ÏÑ§Î™Ö

| Í∏∞Ìò∏ | ÏùòÎØ∏ | ÏÑ§Î™Ö |
| :--- | :--- | :--- |
| $n$ | Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Ïùò Ï¥ù Í∞úÏàò | Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ ÏåçÏùò ÏàòÏûÖÎãàÎã§. |
| $i$ | Í∞úÎ≥Ñ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Ïùò Ïù∏Îç±Ïä§ | 1Î∂ÄÌÑ∞ $n$ÍπåÏßÄÏùò ÏàúÏÑúÎ•º ÎÇòÌÉÄÎÉÖÎãàÎã§. |
| $Y_i$ | $i$Î≤àÏß∏ Îç∞Ïù¥ÌÑ∞Ïùò **Ïã§Ï†úÍ∞í** (Ï∞∏Í∞í) | Î™®Îç∏Ïù¥ ÏòàÏ∏°ÌïòÎ†§Í≥† ÏãúÎèÑÌïòÎäî Ïã§Ï†ú Í¥ÄÏ∏°Í∞íÏûÖÎãàÎã§. |
| $\hat{Y}_i$ | $i$Î≤àÏß∏ Îç∞Ïù¥ÌÑ∞Ïùò **ÏòàÏ∏°Í∞í** | Î™®Îç∏Ïù¥ Í≥ÑÏÇ∞Ìïú Í∞íÏûÖÎãàÎã§. |
| $(Y_i - \hat{Y}_i)^2$ | $i$Î≤àÏß∏ Ïò§Ï∞®Ïùò **Ï†úÍ≥±** | Ïã§Ï†úÍ∞íÍ≥º ÏòàÏ∏°Í∞íÏùò Ï∞®Ïù¥(Ïò§Ï∞®)Î•º Ï†úÍ≥±Ìïú Í∞íÏûÖÎãàÎã§. Ï†úÍ≥±ÏùÑ ÌÜµÌï¥ ÏñëÏàò/ÏùåÏàò Ïò§Ï∞®Î•º ÎèôÏùºÌïòÍ≤å Ï≤òÎ¶¨ÌïòÍ≥† ÌÅ∞ Ïò§Ï∞®Ïóê Îçî ÌÅ∞ ÌéòÎÑêÌã∞Î•º Î∂ÄÏó¨Ìï©ÎãàÎã§. |
| $\sum_{i=1}^{n}$ | Ï¥ùÌï© (ÏãúÍ∑∏Îßà) | $i=1$Î∂ÄÌÑ∞ $n$ÍπåÏßÄ Î™®Îì† Ïò§Ï∞® Ï†úÍ≥±ÏùÑ ÎçîÌï©ÎãàÎã§. |
| $\frac{1}{n}$ | ÌèâÍ∑† | Ï¥ùÌï©ÏùÑ Îç∞Ïù¥ÌÑ∞ Í∞úÏàò $n$ÏúºÎ°ú ÎÇòÎàÑÏñ¥ ÌèâÍ∑†ÏùÑ Íµ¨Ìï©ÎãàÎã§. |

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = [1,2,3]
y_data = [1.5,3,4.5]
```

```python
for i in range(3): #[0,1,2]
    print(i)
```

```python
def mse( x, y, w ):
    c = 0
    for i in range( len(x)):
        hx = w*x[i]
        c = c+ (hx-y[i])**2
    return c/len(x)
```

```python
print( mse( x_data, y_data, -1))
print( mse( x_data, y_data, 0))
print( mse( x_data, y_data, 1))
print( mse( x_data, y_data, 2))
print( mse( x_data, y_data, 3))
print( mse( x_data, y_data, 1.50))
```

```python
plt.scatter( x_data, y_data)
plt.show()
```

```python
for w in np.linspace( -3,5, 50):
    c = mse( x_data, y_data, w)
    print( "w=",w , "cost:", c)
    plt.plot( w,c, 'ro')
plt.xlabel("w")
plt.ylabel("cost")
plt.show()
```
