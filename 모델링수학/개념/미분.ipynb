
#  미분의 기초 개념 정리

## 1. 미분이란 무엇인가? (What is Differentiation?)

미분은 **변화율(Rate of Change)**을 측정하는 수학적 도구입니다.

특정 함수 $f(x)$에 대해, 독립 변수 $x$의 아주 작은 변화($\Delta x$)에 따른 종속 변수 $y=f(x)$의 변화($\Delta y$)의 **순간적인 비율**을 의미합니다.

###  핵심 개념: 순간 변화율 (Instantaneous Rate of Change)

미분의 기하학적 의미는 곡선 $y=f(x)$ 위의 한 점에서의 **접선(Tangent Line)의 기울기**입니다.

* **평균 변화율 (Average Rate of Change):** 두 점 사이의 기울기.
    $$
    \text{평균 변화율} = \frac{\Delta y}{\Delta x} = \frac{f(x + \Delta x) - f(x)}{(x + \Delta x) - x}
    $$

* **순간 변화율 (Instantaneous Rate of Change) / 미분:** $\Delta x$를 0으로 보내는 극한을 취하여 한 점에서의 변화율을 구합니다.
    $$
    \text{미분(도함수)} = f'(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}
    $$

## 2. 미분 표기법 (Notation)

함수 $y = f(x)$를 미분한 결과(도함수)를 나타내는 일반적인 표기법은 다음과 같습니다.

| 표기법 | 읽는 법 | 의미 |
| :---: | :---: | :--- |
| $f'(x)$ | f 프라임 x | 라그랑주(Lagrange) 표기법 |
| $\frac{dy}{dx}$ | 디 와이 디 엑스 | 라이프니츠(Leibniz) 표기법 (가장 일반적) |
| $\frac{d}{dx}f(x)$ | 디 디 엑스 오브 f 엑스 | $f(x)$를 $x$에 대해 미분한다는 연산자 표기 |
| $\dot{y}$ | y 도트 | 뉴턴(Newton) 표기법 (주로 시간에 대한 미분에 사용) |

## 3. 미분 공식 (Differentiation Rules)

가장 기본적인 미분 공식들을 정리합니다.

| 종류 | 함수 $f(x)$ | 도함수 $f'(x)$ |
| :---: | :---: | :---: |
| **상수 함수** | $c$ (c는 상수) | $0$ |
| **멱함수(거듭제곱)** | $x^n$ (n은 실수) | $nx^{n-1}$ |
| **상수배** | $c \cdot f(x)$ | $c \cdot f'(x)$ |
| **덧셈/뺄셈** | $f(x) \pm g(x)$ | $f'(x) \pm g'(x)$ |

###  예시 (Example)

함수 $f(x) = 5x^3 + 2x - 7$을 미분해 봅시다.

$$
\begin{aligned}
f'(x) &= \frac{d}{dx} (5x^3 + 2x - 7) \\
&= \frac{d}{dx}(5x^3) + \frac{d}{dx}(2x) - \frac{d}{dx}(7) \\
&= 5 \cdot \frac{d}{dx}(x^3) + 2 \cdot \frac{d}{dx}(x^1) - 0 \\
&= 5 \cdot (3x^{3-1}) + 2 \cdot (1x^{1-1}) \\
&= 15x^2 + 2x^0 \\
&= 15x^2 + 2
\end{aligned}
$$

따라서 $f'(x) = 15x^2 + 2$ 입니다.

## 4. 미분의 응용 예시 (Application Example)

미분은 **물리학**이나 **경제학** 등 다양한 분야에서 변화율을 계산하는 데 사용됩니다.

###  위치, 속도, 가속도

물체의 **위치**를 나타내는 함수 $s(t)$가 있다고 합시다. ($t$는 시간)

1.  **속도 (Velocity):** 위치의 **시간에 대한** 순간 변화율입니다.
    $$
    v(t) = s'(t) = \frac{ds}{dt}
    $$
2.  **가속도 (Acceleration):** 속도의 **시간에 대한** 순간 변화율입니다. (위치 함수의 **이계 미분**)
    $$
    a(t) = v'(t) = s''(t) = \frac{d^2s}{dt^2}
    $$

###  예시: 로켓의 운동

로켓의 지면으로부터의 높이(위치)가 시간 $t$에 따라 다음 함수로 주어진다고 가정해 봅시다.
$$
s(t) = -t^2 + 10t \quad (\text{미터, 초})
$$

**문제:** 로켓이 발사된 후 3초일 때의 **순간 속도**는 얼마인가요?

**풀이:**

1.  속도 함수 $v(t)$를 구하기 위해 $s(t)$를 미분합니다.
    $$
    v(t) = s'(t) = \frac{d}{dt}(-t^2 + 10t) = -2t + 10
    $$
2.  $t=3$을 속도 함수에 대입합니다.
    $$
    v(3) = -2(3) + 10 = -6 + 10 = 4
    $$

**결론:** 로켓이 발사된 후 3초일 때의 순간 속도는 $4 \text{m/s}$ 입니다.

***

미분을 사용하는 가장 **쉽고 핵심적인 이유**는 바로 **'순간적인 변화'를 정확하게 측정**하기 위해서.

##  미분을 사용하는 이유: '순간'을 포착하기 위해

우리가 일상에서 접하는 대부분의 변화는 일정한 속도가 아니라 **시시각각 변하는** 비선형적인 움직임입니다. 미분은 이 복잡한 변화의 흐름 속에서 **딱 한 지점(순간)**의 **변화하는 빠르기(기울기)**를 알려주는 도구입니다.

### 1. 평균 변화와 순간 변화의 차이

* **평균 변화율 (미분 이전):** 기차로 서울에서 부산까지 400km를 4시간 동안 달렸다면, **평균 속도**는 $100 \text{km/h}$입니다. 하지만 이 기차가 내내 $100 \text{km/h}$로 달렸을까요? 아닙니다. 서행하기도, 급가속하기도 했을 겁니다.
* **순간 변화율 (미분):** 미분은 기차가 **정확히 2시간 30분 째**에 $125 \text{km/h}$로 달리고 있었다는 사실을 포착해 줍니다. 즉, **곡선 위 한 점의 정확한 기울기**를 구합니다.

##  미분의 주요 활용 분야

### 1. 물리/공학: 속도, 가속도 계산

물체의 움직임을 다룰 때 미분은 필수적입니다.

* **위치 함수**를 미분하면 그 물체의 **순간 속도**를 알 수 있습니다.
* **속도 함수**를 다시 미분하면 그 물체의 **순간 가속도**를 알 수 있습니다.

**예시:** 로켓이 발사될 때, **특정 시간**에 연료를 얼마나 더 분사해야 **정확한 궤도**에 진입할 수 있는지 계산하는 데 사용됩니다.

### 2. 최적화: 가장 효율적인 지점 찾기

미분은 함수가 **가장 크거나(극대) 가장 작아지는(극소) 지점**을 찾는 데 사용됩니다. 이는 도함수(미분 값)가 0이 되는 지점을 찾는 것을 의미합니다.

* **경제학:** 특정 제품을 몇 개 생산해야 **이윤(Profit)이 최대**가 되는지 계산합니다.
* **경영:** 비용 함수를 미분하여 **가장 낮은 비용(Cost)이 발생하는 생산량**을 찾습니다.

### 3. 경사하강법 (Gradient Descent)

최근 **인공지능(AI)과 머신러닝**에서 미분은 핵심적인 역할을 합니다.

* AI 모델은 **오차(Cost/Loss)**를 최소화하기 위해 학습합니다.
* 미분은 현재 지점에서 **오차를 가장 빠르게 줄일 수 있는 방향(가장 가파른 경사)**을 알려줍니다. AI는 이 방향을 따라 조금씩 이동하며 학습을 진행합니다.

#  합성함수의 미분 (Chain Rule)

합성함수의 미분은 **연쇄 법칙(Chain Rule)**이라고 불립니다. 함수가 $f(x)$처럼 하나의 변수 $x$로 이루어진 것이 아니라, **함수 안에 또 다른 함수**가 포함되어 있을 때 사용됩니다.

## 1. 연쇄 법칙의 기본 원리

함수 $y = f(g(x))$를 미분할 때, 두 함수의 변화율을 **곱해주는** 방식으로 미분합니다.

1.  **겉 함수** $f$를 미분하고, $g(x)$는 그대로 둡니다.
2.  **속 함수** $g$를 미분한 것을 곱해줍니다.

$$
\frac{d}{dx} [f(g(x))] = f'(g(x)) \cdot g'(x)
$$

## 2. 라이프니츠 표기법을 이용한 설명 (치환을 활용)

제공된 이미지 는 중간 변수 $t$를 사용하여 이 과정을 명확히 보여줍니다.

합성함수를 다음과 같이 정의합니다.
* **겉 함수:** $f(t)$
* **속 함수:** $t = g(x)$

우리가 최종적으로 구하고 싶은 것은 $x$에 대한 $f$의 변화율, 즉 $\frac{df}{dx}$ 입니다.

### 연쇄 법칙의 공식 (라이프니츠 표기법)

$$
\frac{df}{dx} = \frac{df}{dt} \cdot \frac{dt}{dx}
$$

###  예시 적용: $f(x) = (x+2)^2$

함수 $f(x) = (x+2)^2$를 $x$에 대해 미분하는 과정입니다.

#### 1단계: 치환 (Substitution)

속 함수를 $t$로 치환합니다.
$$
\text{Let } t = x+2 \quad (\text{즉, } t = g(x)) \quad
$$
이제 겉 함수는 $f(t) = t^2$ 이 됩니다.

#### 2단계: 각 변수에 대한 미분 계산

**A. 겉 함수 $f(t)$를 $t$에 대해 미분 ($\frac{df}{dt}$)**
$$
\frac{df}{dt} = \frac{\partial}{\partial t} (t^2) = 2t \quad
$$

**B. 속 함수 $t$를 $x$에 대해 미분 ($\frac{dt}{dx}$)**
$$
\frac{dt}{dx} = \frac{\partial}{\partial x} (x+2) = 1 \quad
$$

#### 3단계: 연쇄 법칙 적용 (두 미분 결과를 곱함)

구한 두 결과를 연쇄 법칙 공식에 대입하고, $t$를 다시 $x+2$로 복원합니다.
$$
\begin{aligned}
\frac{df}{dx} &= \frac{df}{dt} \cdot \frac{dt}{dx} \quad \\
&= (2t) \cdot (1) \quad \\
&= 2(x+2) \quad
\end{aligned}
$$

**결론:** $\frac{d}{dx}(x+2)^2 = 2(x+2)$ 입니다.

---

###  요약: 연쇄 법칙의 의미

연쇄 법칙은 **'속 함수가 변하는 빠르기($\frac{dt}{dx}$)'** 와 **'그 변화가 겉 함수에 미치는 영향의 빠르기($\frac{df}{dt}$)'** 를 곱하여, **'가장 바깥 변수($x$)의 변화가 최종 결과($f$)에 미치는 영향'**을 계산하는 방법입니다.
