#  적분 공식 

---

##  적분이란?

> **"아주 작은 조각들을 모아 전체를 구하는 것"**

즉, **넓이**, **길이**, **부피** 등 **연속적인 양을 더하는 과정**입니다.

---

##  적분의 기본 공식


$\int_a^b f(x)\, dx$


| 구성 요소 | 의미 |
|-----------|------|
| $\int$ | 적분하라는 기호 (∫는 S를 늘린 모양, 'sum = 더하라') |
| $f(x)$ | 함수 (예: 높이, 속도 등) |
| $dx$ | $x$에 대해 아주 작게 나눈 조각들 |
| $a$, $b$ | 적분 구간 (어디부터 어디까지 더할지) |

> 이 공식은 **$x = a$부터 $x = b$까지 함수 $f(x)$의 그래프 아래 넓이**를 의미합니다.

---

##  예시: 속도에서 거리 구하기

- 어떤 차가 시속 60km로 2시간 동안 달렸을 때,


$\text{거리} = \int_0^2 60\, dt = 60 \times (2 - 0) = 120\ \text{km}$


- 여기서 60은 속도, $dt$는 시간의 아주 작은 단위
- 결과는 **속도 × 시간 = 거리**

##  면적 구하기 예시

### 함수: $f(x) = x$


$\int_0^3 x\, dx = ?$


- 이건 $x=0$부터 $x=3$까지의 **삼각형 모양 면적**
- 실제로 계산하면:


$\int_0^3 x\, dx = \frac{1}{2} \cdot 3 \cdot 3 = 4.5%$




$\int_0^3 x\, dx$ 는 함수 $f(x) = x$ 와 $x$축, 그리고 수직선 **$x=0$**과 **$x=3$**으로 둘러싸인 영역의 **넓이**를 구하는 것. 

이것이 왜 넓이가 되는지 이해하려면, **정적분**이 태어난 아이디어인 **리만 합**의 개념을 떠올려보면 쉽다.

---

### 1. 쪼개고, 곱하고, 더하기 (리만 합의 아이디어)

정적분은 곡선 아래의 넓이를 정확하게 구하기 위해, 그 넓이를 **매우 얇은 직사각형들**의 합으로 근사하는 것에서 시작됩.

* **쪼개기 (분할):** 구하고자 하는 구간(여기서는 $x=0$부터 $x=3$)을 **아주 얇고 작은 구간**으로 나눕니다. 이 작은 구간의 폭을 $\Delta x$ 라고 합시다.
* **곱하기 (넓이 근사):** 각 작은 구간에서, 함수의 값 $f(x)$를 직사각형의 **높이**로, 구간의 폭 $\Delta x$를 직사각형의 **밑변**으로 삼습니다. 따라서, 직사각형 하나의 넓이는 $\text{넓이} \approx f(x) \cdot \Delta x$ 가 됨.
* **더하기 (총 넓이 근사):** 이 모든 얇은 직사각형들의 넓이를 **모두 더한다** (이것이 $\sum$ 기호의 역할).

$$\text{근사 면적} = \sum_{\text{모든 직사각형}} f(x) \cdot \Delta x$$

---

### 2. 극한을 취하여 '정확한' 넓이로

위의 근사 면적은 얇은 직사각형들이기 때문에, 곡선과 직사각형 윗변 사이에 아주 작은 오차가 생긴다..

이 오차를 없애고 **정확한 면적**을 얻는 방법은, 직사각형들의 밑변 폭 $\Delta x$를 **0에 가깝게** (거의 무한히 얇게) 만드는 것이다.

* 이처럼 **$\Delta x$를 0으로 보내는 극한**을 취하는 순간, **$\sum$ (더하기)** 기호는 **$\int$ (인테그랄)** 기호로 바뀌게 된다..
* **$\Delta x$ (밑변의 폭)**는 **$dx$**로 바뀝니다.
* **$f(x)$ (높이)**와 **$dx$ (밑변)**의 곱을 **무한히 더한다**는 의미로 $\int f(x)\, dx$ 가 된다..

$$\text{정확한 면적} = \lim_{\Delta x \to 0} \sum f(x) \cdot \Delta x = \int_a^b f(x)\, dx$$

### 3. 예시 $f(x)=x$의 경우

사용자님의 예시 $\int_0^3 x\, dx$ 는 결국 다음을 의미:

* **함수:** $f(x) = x$ (높이가 $x$인 직사각형)
* **구간:** $x=0$부터 $x=3$까지
* **의미:** $x$축 위에서 $f(x)=x$의 그래프 아래에 있는 모든 **무한히 얇은 직사각형**들의 넓이를 더하라.

$f(x)=x$의 그래프를 보면, $x=0$에서 $x=3$까지의 영역은 정확히 **밑변과 높이가 3인 직각삼각형** 모양.

* **삼각형 넓이 공식:** $\frac{1}{2} \times \text{밑변} \times \text{높이}$
* **계산:** $\frac{1}{2} \cdot 3 \cdot f(3) = \frac{1}{2} \cdot 3 \cdot 3 = 4.5$

따라서, **정적분 $\int_0^3 x\, dx$는 정확히 이 삼각형의 넓이인 $4.5$를 계산**됨. 정적분은 **넓이를 구하는 수학적 도구**이기 때문.

##  기본적인 적분 공식 요약

| 함수 | 적분 결과 |
|------|------------|
| $\int x^n\, dx$ | $\frac{x^{n+1}}{n+1} + C$  (단, $n \neq -1$) |
| $\int \frac{1}{x}\, dx$ | $\ln|x| + C$   |
| $\int e^x\, dx$ | $e^x + C$ |
| $\int \cos x\, dx$ | $\sin x + C$ |
| $\int \sin x\, dx$ | $- \cos x + C$ |

> 여기서 $C$는 **상수항 (적분 상수)**  
> 왜냐하면 미분하면 상수가 사라지기 때문에, 적분할 땐 항상 더해줘야 합니다.
