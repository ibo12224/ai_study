```python
def sc1(a):
    sc=""
    if a>80:
        sc="A"
    elif a>60:
        sc="B"
    elif a>40:
        sc="C"
    return sc

sc1(75)
```

```python
def ff(kg,cm):
    avg=(cm-100)*0.85
    fat=(kg/avg)*100

    s=""
    if fat<=90:
        s="저체중"
    elif 90<fat<=110:
        s="정상"
    elif 110<fat<=120:
        s="과체중"
    else:
        s="비만"
    return(s)
ff(100,180)
```

```python
def yun(year):
    if year%4==0 and year%100!=0 or year%400==0:
        print('윤년입니다.')
    else:
        print('윤년이 아닙니다.')
yun(400)
```

```python
def yy(year):
    animallist = ['원숭이', '닭', '개', '돼지', '쥐', '소', '호랑이', '토끼', '용', '뱀', '말', '양']
    return f"{year}년은 {animallist[year % 12]}띠입니다."
yy(2000)
```

```python
def mm(money,price):
    change = money - price
    c={}
    for coin in [500, 100, 50, 10]:
        count = change // coin
        change %= coin #change = change % coin
        if count > 0:
            c[coin]=change
    return c
print(mm(1500,500))
```

```python
def num(a):
    num=[]
    for i in range(1,a+1):
        if a%i==0:
            num.append(i)
    return num
num(10)
```
