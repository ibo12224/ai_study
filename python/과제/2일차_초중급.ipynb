```python
kg=int(input("몸무게:"))
cm=int(input("키:"))

avg=(cm-100)*0.85
fat=(kg/avg)*100

s=""
if fat<=90:
    s="저체중"
elif 90<fat<=110:
    s="정상"
elif 110<fat<=120:
    s="과체중"
else:
    s="비만"
print(s)
```
