```python
class MyObj:
    num=0
    def __init__(self):
        MyObj.num+=1
        
    @staticmethod
    def show():
        return MyObj.num

Obj1=MyObj()
Obj2=MyObj()
Obj3=MyObj()

print(Obj1.show())
```

```python
class 차:
    def __init__(self,바퀴,가격):
        self.바퀴=바퀴
        self.가격=가격

class 자전차(차):
    def __init__(self, 바퀴, 가격,구동계):
        super().__init__(바퀴,가격)
        self.구동계=구동계

bicycle=자전차(2,100,"사미노")
print(bicycle.바퀴)
print(bicycle.가격)
print(bicycle.구동계)
```

```python
class Book:
    def __init__(self,title,author):
        self.title=title
        self.author=author
       
    def __repr__(self):
        return f"책제목:{self.title} 저자:{self.author}"
book=Book("데미안","헤르만 헤세")
print(book)
```

```python
class Test:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def setData(self,x,y):
        self.a=x
        self.b=y
    def show(self):
        print(self.a,self.b)
    def __add__(self, other):
        return Test(self.a + other.a, self.b + other.b)

    def __repr__(self):
        return f"({self.a},{self.b})"
obj1=Test(10,20)
obj2=Test(100,200)

rst=obj1+obj2
print(rst)
```

```python
class Student:
    def __init__(self,name,score):
        self.name=name
        self.score=score
    
    def __eq__(self, value):
        return self.score==value.score
    def __lt__(self,value):
        return self.score<value.score

s1=Student("민수",85)
s2=Student("지연",90)
s3=Student("철수",85)

print(s1==s3)
print(s1<s2)
print(s2<s3)
```
