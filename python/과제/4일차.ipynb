```python
def myprint(list,df):
    for i in list:
        df(i)

mylist=[10,20,30]
myprint(mylist,lambda v:print(v))

myprint(mylist,lambda v:print(v,end=' '))
```

```python
class Stock:
    def __init__(self, 종목명, 종목코드, PER, PBR, 배당수익률):
        self.종목명 = 종목명
        self.종목코드 = 종목코드
        self.PER = PER
        self.PBR = PBR
        self.배당수익률 = 배당수익률

    def show(self):
        print(self.종목명,
              self.종목코드,
              self.PER,
              self.PBR,
              self.배당수익률)

삼성 = Stock("삼성전자","005930",15.79,1.33,2.83)
#삼성.__init__(삼성,"삼성전자","005930",15.79,1.33,2.83).show
print(삼성.배당수익률)
삼성.show()
```

```python
class Account:
    def __init__(self, a="은행", b="예금주", c="계좌번호", d="잔고"):
        self.a=a
        self.b=b
        self.c=c
        self.d=d
    def display_info(self):
        print(f"은행이름: {self.a}")
        print(f"예금주: {self.b}")
        print(f"계좌번호: {self.c}")
        print(f"잔고: {format(self.d,',')}원")

# 객체 생성
acc = Account("SC 은행", "파이썬", "111-11-111111", 10000)

# 정보 출력
acc.display_info()
```
