```python
#11. 키와 몸무게를 입력받아 비만도를 구하고 결과를 출력하시오.
height = int(input("키(cm)를 입력하세요: "))
weight = float(input("몸무게(kg)를 입력하세요: "))

standard_weight = (height - 100) * 0.85
obesity = (weight / standard_weight) * 100

print(f"비만도: {obesity:.2f}%")

if obesity <= 90:
    print("저체중")
elif obesity <= 110:
    print("정상")
elif obesity <= 120:
    print("과체중")
else:
    print("비만")
```

```python
#17. 상품가격과 지불액을 입력하여 거스름돈을 최소화하시오.
price = int(input("상품 가격을 입력하세요: "))
money = int(input("지불한 금액을 입력하세요: "))

change = money - price
#print(f"잔돈: {change}원")

for coin in [500, 100, 50, 10]:
    count = change // coin
    change %= coin #change = change % coin
    if count > 0:
        print(f"{coin}원 {count}개", end=' ')
```

```python
#19-1. 환율 관련 어쩌구...
rates = {
    "달러": 1167,
    "엔": 1.096,
    "유로": 1268,
    "위안": 171
}

money, rate = input("입력: ").split()
money = float(money)

if rate in rates:
    print(f"{money * rates[rate]:.3f} 원")
else:
    print("입력이 잘못되었습니다")
```

```python
#19-2. 환율 관련 어쩌구...
money, unit = input("입력: ").split()
money = float(money)

if unit == "달러":
    print(money * 1167, "원")
elif unit == "엔":
    print(money * 1.096, "원")
elif unit == "유로":
    print(money * 1268, "원")
elif unit == "위안":
    print(money * 171, "원")
else:
    print("입력이 잘못되었습니다")
```

```python
#20. 통신사 어쩌구
number = input("휴대전화 번호 입력: ")
frontnum = number.split('-')[0]

if frontnum == '011':
    company = 'SKT'
elif frontnum == '016':
    company = 'KT'
elif frontnum == '019':
    company = 'LGU'
else:
    company = '알수없음'

print(f"당신은 {company} 사용자입니다.")
```
