# 2. 아 래 멤 버 를 가 지 는 Stock 클 래 스 를 정 의 하 시 요
삼성 = Stock 삼성전자 "005930 "", 15.79, 1.33,
prin t 삼성 배당수익률 #2.83 출 력
삼
성 show() 종 목 ~ 배 당 수 익 률 모 두 출 력
항목 정보 종목명 삼성전자 종목코드 005930 PER 15.79
항목 정보 PBR 1.33 배당수익률 2.83

```python
class Stock:
    def __init__(self, 종목명, 종목코드, PER, PBR, 배당수익률):
        self.종목명 = 종목명
        self.종목코드 = 종목코드
        self.PER = PER
        self.PBR = PBR
        self.배당수익률 = 배당수익률

    def show(self):
        print(self.종목명,
              self.종목코드,
              self.PER,
              self.PBR,
              self.배당수익률)

삼성 = Stock("삼성전자","005930",15.79,1.33,2.83)

print(삼성.배당수익률)
삼성.show()
```

# 4
Account 클래스를 만든후 객체를 생성하여 저장된 정보를 출력하는 display_info() 메서드를 이용하여 아래 처럼 출력하시요.

```python
class Account:
    def __init__(self, bank_name, owner, account_number, balance):
        self.bank_name = bank_name
        self.owner = owner
        self.account_number = account_number
        self.balance = balance

    def display_info(self):
        formatted_balance = format(self.balance, ',')  # 천 단위 쉼표 표시
        print(f"은행이름: {self.bank_name}")
        print(f"예금주: {self.owner}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {formatted_balance} 원")

        
# 객체 생성
acc = Account("SC 은행", "파이썬", "111-11-111111", 10000)

# 정보 출력
acc.display_info()
```
