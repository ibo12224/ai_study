### 세트멤버함수

```python
s={10,20,30,40,10,50,30}
print(s)
```

```python
s1={10,20,30}
s2={20,30,40,50}
print(s1.intersection(s2))
print(s2&s1)
print(s1.union(s2))
print(s1|s2)
print(s2.difference(s1))
print(s2-s1)
```
