### 6.슬라이싱
- [시작인덱스:끝인덱스:증가치]
- 시작인덱스 <=인덱스<끝인덱스

```python
my=[10,20,30,40,50]
#print(my[1:4:1])
print(my[0:5:2])
print(my[:4])
print(my[1:])
print(my[0:-1])
print(my[-3:-1])
```
