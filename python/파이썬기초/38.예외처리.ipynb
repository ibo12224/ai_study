## if 문이 많아지면 유지보수가 어렵다.
## 시스템에서 자주 발생하는 예외(에러)는 파이썬 자동처리해 준다.

```python
pizz_slice = 9
person = 0
try:
    per_pizza_slice = pizz_slice/person
    print( per_pizza_slice)
except ZeroDivisionError as err:
    print('예외발생', err)
except Exception as err: #모든 예외를 모두 처리
    print("에러발생",err )
```

```python
pizz_slice = 9
person = -3
try:
    if person<0:
        raise Exception('person이 음수면 않됩니다.')
    per_pizza_slice = pizz_slice/person
    print( per_pizza_slice)
except Exception as err: #모든 예외를 모두 처리
    print("에러발생",err )
```
