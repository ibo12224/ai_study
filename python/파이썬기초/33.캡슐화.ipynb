### 데이터 무결성을 보존

```python
class Calendar:
    def __init__(self):
        self.month = 0

    def setMonth(self, month):
        if month<1 or month>12:
            print( '잘못된 개월')
        else:
            self.month = month
    
    def getMonth(self):
        return self.month
    
cal = Calendar()
# cal.month = 13
#cal.setMonth(13) #cal.setMonth(cal,13)
cal.setMonth(12) 
print( cal.getMonth())
```
