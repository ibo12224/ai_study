```python
class Test:
    st = 100 # static 멤버(객체생성X), 클래스변수
    def __init__(self):
        self.a = 0 # instance변수( 반드시 객체 생성필요)
        self.b = 0 # instance변수( 반드시 객체 생성필요)
    def setData(self, x,y): 
        self.a = x
        self.b = y
    def show(self):
        print( self.a, self.b)
    @staticmethod
    def stFn(aa):
        Test.st=aa
    @staticmethod
    def getSt():
        return Test.st

print( Test.st)
obj = Test()
```

```python
class Student:
    def __init__(self,a,b,c):
        self.name=a
        self.age=b
        self.addr=c
    def show(self):
        print('이름',self.name)
        print('나이',self.age)
        print('주소',self.addr)
```

```python
std1=Student('홍길동',20,'서울시')
std1.show()
```
