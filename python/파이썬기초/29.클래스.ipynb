```python
class Test:
    def __init__(self): ## 생성자 함수
        print('init call')
        self.a = 0 # 간접연산자
        self.b = 0

    def setData(self, x,y): #멤버함수
        self.a = x
        self.b = y

    def show(self): #멤버함수
        print( self.a, self.b)
#   자바: Test obj = new Test()
obj = Test() # Test객체(Test 클래스의 멤버데이터 메모리할당완료),레퍼런스, 변수
# obj.__init__(obj) # <-- 컴파일러가 추가
obj.setData(10,20) # obj.setData(obj, 10,20)
obj.show()#  obj.setData(obj)
print( obj.a , obj.b )
```

```python
class Test:
    def __init__(self):
        print('init self', id(self))
        self.a = 0 
        self.b = 0
    def setData(self, x,y): 
        print('setdata self', id(self))
        self.a = x
        self.b = y
    def show(self):
        print('show self', id(self))
        print( self.a, self.b)

obj = Test()
# obj.__init__(obj) # <-- 컴파일러가 추가
print( 'obj주소', id(obj) )
obj.setData(10,20) # obj.setData(obj, 10,20)
obj.show()#  obj.show(obj)
print( obj.a , obj.b )

obj1 = Test()
# obj1.__init__(obj1) # <-- 컴파일러가 추가
print( 'obj1주소', id(obj1) )
obj1.setData(11,22) # obj1.setData(obj1,11,22)
obj1.show() # obj1.show(obj1)
print( obj1.a, obj1.b)
```
