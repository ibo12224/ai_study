### csv 파일기록

```python
import csv
fp = open('my.csv', 'w', newline='')
wr = csv.writer( fp )
# print( type(wr))
wr.writerow( ['aa',10,20] )
wr.writerow( ['bb',10,20] )
wr.writerow( ['cc',10,20] )
fp.close()
```

```python
fp = open('my.csv','r')
rd = csv.reader(fp)
for a,b,c in rd:
    print( a,b,c )
fp.close()
```
