AWS의 네트워크 기초이자 가상 서버들이 거주하는 집과 같은 개념인 **VPC(Virtual Private Cloud)**

---

# AWS VPC (Virtual Private Cloud)

VPC는 AWS 클라우드 내에서 사용자 전용으로 격리된 **가상 네트워크 공간**을 정의하는 서비스입니다. 사용자는 이 공간 안에서 IP 주소 범위 선택, 서브넷 생성, 라우팅 테이블 및 네트워크 게이트웨이 구성 등 네트워크 환경을 완벽하게 제어할 수 있습니다.


####  "VPC는 내 집의 울타리다"

> "AWS라는 거대한 대지 위에 **나만의 울타리(VPC)**를 치는 것과 같습니다. 울타리 안에 거실(퍼블릭 서브넷)과 안방(프라이빗 서브넷)을 나누고, 누가 들어오고 나갈지 문지기(보안 그룹/NACL)를 세워 안전하게 관리하는 기술입니다."

---


## 1. VPC의 핵심 구성 요소

| 요소 | 설명 | 비유 |
| --- | --- | --- |
| **서브넷 (Subnet)** | VPC의 IP 주소 범위를 더 작은 단위로 나눈 네트워크 층 | 아파트의 각 호실 또는 층 |
| **라우팅 테이블 (Route Table)** | 네트워크 트래픽이 어디로 가야 하는지 알려주는 규칙 모음 | 건물 내 이정표 또는 내비게이션 |
| **인터넷 게이트웨이 (IGW)** | VPC와 인터넷 간의 통신을 가능하게 하는 연결 통로 | 아파트 단지의 정문 출입구 |
| **NAT 게이트웨이** | 프라이빗 서브넷의 자원이 외부 인터넷으로 나가게만 해주는 통로 | 밖으로만 전화를 걸 수 있는 내부 전화기 |

---

## 2. 퍼블릭 서브넷 vs 프라이빗 서브넷

강의 시 보안 관점에서 가장 중요하게 다루어야 할 구분입니다.

* **퍼블릭 서브넷 (Public Subnet):**
* 인터넷 게이트웨이(IGW)를 통해 인터넷과 직접 통신이 가능한 서브넷입니다.
* 주로 웹 서버(Web Server)나 로드 밸런서가 위치합니다.


* **프라이빗 서브넷 (Private Subnet):**
* 외부 인터넷에서 직접 접근할 수 없는 격리된 서브넷입니다.
* 주로 데이터베이스(DB)나 백엔드 로직 서버를 배치하여 보안을 강화합니다.



---

## 3. VPC 보안 도구 (Layered Security)

VPC 내에서는 두 단계의 네트워크 보안 계층을 가집니다.

1. **네트워크 ACL (NACL):**
* **서브넷 수준**의 방화벽입니다.
* 모든 트래픽(인바운드/아웃바운드)에 대해 허용 및 거부 규칙을 설정합니다.
* 상태 비저장(Stateless) 방식입니다.


2. **보안 그룹 (Security Group):**
* **인스턴스(EC2) 수준**의 가상 방화벽입니다.
* 특정 포트나 IP에 대해 허용 규칙만 설정 가능합니다.
* 상태 저장(Stateful) 방식입니다.


## 4. VPC (Virtual Private Cloud) 구조 이해
VPC는 사용자만의 격리된 가상 네트워크 환경을 구축하는 기술입니다.

VPC 대역 설정: 예시에서는 10.0.0.0/16이라는 큰 주소 범위를 할당했습니다.

서브넷(Subnet) 분할: 큰 네트워크를 더 작은 단위로 쪼개어 관리합니다.

Subnet 1: 10.0.1.0/24 대역 할당.

Subnet 2: 10.0.2.0/24 대역 할당.


* **주요 특징**:
* VPC 내에서 각종 리소스(EC2, RDS, ELB 등)를 시작할 수 있으며, 다른 가상 네트워크와 논리적으로 분리되어 있습니다.
* 각 리전(Region)별로 다수의 VPC가 존재할 수 있습니다.
* VPC는 하나의 사설 IP 대역을 보유하고, 서브넷을 생성하여 이 대역의 일부를 나누어 줄 수 있습니다.


* **네트워크 주소 관리 (CIDR)**:
* **CIDR (Classless Inter-Domain Routing)** 방식을 사용하여 IP 주소를 할당합니다.
* 허용된 IP 블록 크기는 `/16` (IP 65,536개)에서 `/28` (IP 16개) 사이입니다.
* **예시 구조**:
* **VPC 대역**: `10.0.0.0/16`
* **Subnet 1**: `10.0.1.0/24`
* **Subnet 2**: `10.0.2.0/24`

VPC 내에 생성된 분리된 네트워크 단위입니다.

AZ 연결: 하나의 서브넷은 하나의 **가용 영역(AZ, Availability Zone)**에 연결됩니다.

자원 할당: 실질적으로 리소스들은 이 서브넷에서 생성되며, 사설 IP를 기본적으로 할당받고 필요에 따라 공인 IP를 할당받습니다.

개별 제어: 하나의 서브넷은 하나의 라우팅 테이블과 하나의 **NACL(Network ACL)**을 가집니다

---

### 5. 인터넷 게이트웨이 (Internet Gateway, IGW)

VPC와 외부 인터넷을 이어주는 연결 고리입니다.

정의: VPC 내 리소스가 외부 인터넷을 사용하고자 할 때 거치는 통로(게이트웨이)입니다.

필수성: 인터넷 게이트웨이가 없으면 VPC 내부 자원은 외부 인터넷과 통신할 수 없습니다.






# AWS 라우팅 테이블 (Route Table)

라우팅 테이블은 VPC 내의 네트워크 트래픽이 어디로 향해야 하는지 결정하는 **이정표(규칙 모음)** 역할을 합니다.

## 1. 라우팅 테이블의 정의

* **역할**: 서브넷이나 게이트웨이에서 나가는 트래픽의 방향을 지정하는 라우팅 규칙(Route)들의 집합입니다.
* **기본 작동**: 트래픽의 목적지 IP 주소를 확인하고, 라우팅 테이블에 정의된 규칙과 비교하여 가장 적합한 경로(Target)로 패킷을 전달합니다.

---

## 2. 주요 구성 요소

| 항목 | 설명 | 예시 |
| --- | --- | --- |
| **대상 (Destination)** | 트래픽이 도달하려는 목적지 IP 범위 (CIDR) | `10.0.0.0/16`, `0.0.0.0/0` |
| **타겟 (Target)** | 트래픽을 보낼 게이트웨이, 인터페이스 또는 연결 | `local`, `igw-id`, `nat-id` |

### 라우팅 규칙 예시

* **Local 경로**: VPC 내부 통신을 위한 규칙입니다. 모든 라우팅 테이블에는 VPC 내부 주소 범위(예: `10.0.0.0/16`)에 대해 `local` 타겟이 자동으로 생성되며, 이는 삭제할 수 없습니다.
* **인터넷 경로**: 외부 인터넷으로 나가는 규칙입니다. 목적지를 `0.0.0.0/0`(모든 IP 주소)으로 설정하고 타겟을 인터넷 게이트웨이(`igw-xxxx`)로 지정합니다.

---

## 3. 서브넷과의 관계 (연결성)

* **1:1 대응**: 하나의 서브넷은 반드시 **하나의 라우팅 테이블**에 연결되어야 합니다.
* **공유 가능**: 하나의 라우팅 테이블을 여러 개의 서브넷이 공유하여 사용할 수 있습니다.
* **기본 라우팅 테이블**: VPC 생성 시 자동으로 생성되는 메인 라우팅 테이블입니다. 별도로 지정하지 않은 서브넷은 모두 이 메인 테이블을 따릅니다.

---

## 4. 퍼블릭 vs 프라이빗 서브넷 결정 기준

라우팅 테이블의 설정에 따라 서브넷의 종류가 결정됩니다.

1. **퍼블릭 서브넷 (Public Subnet)**:
* 라우팅 테이블에 **인터넷 게이트웨이(IGW)**로 향하는 경로(`0.0.0.0/0 -> igw-id`)가 포함된 서브넷입니다.


2. **프라이빗 서브넷 (Private Subnet)**:
* 인터넷 게이트웨이로 향하는 경로가 **없는** 서브넷입니다. 외부에서 직접 접근이 불가능하여 보안이 강화됩니다.



---

##  "라우팅 테이블은 내비게이션이다"

> "우리가 목적지를 입력하면 길을 안내해주는 내비게이션처럼, 데이터 패킷이 '인터넷(IGW)'으로 갈지 '옆 동네 서버(Local)'로 갈지 알려주는 역할을 합니다. 만약 인터넷 게이트웨이로 가는 길(Route)을 적어주지 않으면, 그 서브넷은 외부 세상과 단절된 **프라이빗 서브넷**이 되는 것입니다."

AWS(Amazon Web Services) 환경에서 **고가용성(High Availability)**과 **보안**을 고려한 전형적인 웹 서비스 아키텍처.

---

### 1. 기본 네트워크 구조 (VPC 및 가용 영역)

* **VPC (Virtual Private Cloud):** 가장 바깥쪽 테두리로, AWS 클라우드 내에 나만의 독립된 가상 네트워크 공간을 의미합니다.
* **Availability Zones (AZ 1, 2):** 물리적으로 떨어진 두 개의 데이터 센터를 사용합니다. 한 곳에 장애가 발생해도 서비스가 중단되지 않도록 **이중화**된 구조입니다.

### 2. 서브넷 (Subnet) 구성

네트워크는 보안을 위해 두 종류의 서브넷으로 나뉩니다.

* **Public Subnet (퍼블릭 서브넷):** 인터넷과 직접 통신이 가능한 영역입니다. 외부 사용자가 접속하는 **Web Server(EC2)**가 여기에 위치합니다.
* **Private Subnet (프라이빗 서브넷):** 인터넷에서 직접 접속할 수 없는 안전한 영역입니다. 비즈니스 로직을 처리하는 **App Server**와 핵심 데이터가 담긴 **RDS(데이터베이스)**가 위치합니다.

### 3. 주요 서비스별 역할

| 서비스 | 위치 | 역할 |
| --- | --- | --- |
| **EC2 (Web Server)** | 퍼블릭 서브넷 | 사용자로부터 요청을 받는 첫 번째 관문입니다. |
| **EC2 (App Server)** | 프라이빗 서브넷 | 실제 데이터를 처리하고 DB와 통신하는 핵심 엔진입니다. |
| **RDS (DB Instance)** | 프라이빗 서브넷 | 데이터를 안전하게 저장하며, 두 가용 영역에 걸쳐 있어 복구가 빠릅니다. |
| **S3 Bucket** | VPC 외부 | 이미지, 영상 등 대용량 정적 파일을 저장하는 무제한 스토리지입니다. |

### 4. 트래픽의 흐름

1. **사용자**가 **Internet Gateway**를 통해 들어옵니다.
2. **Web Server(퍼블릭 EC2)**가 요청을 받아 **App Server(프라이빗 EC2)**로 전달합니다.
3. **App Server**는 필요한 정보를 **RDS**에서 가져오거나 **S3**에서 파일을 읽어옵니다.
4. 처리된 결과를 다시 사용자에게 돌려줍니다.

---

**요약하자면:**
이 설계도는 "서버와 DB를 두 곳에 나누어 배치해 **안정성**을 높이고(High Availability), 중요한 정보는 안쪽(Private)에 숨겨 **보안**을 강화한 표준적인 클라우드 설계 방식"을 나타냅니다.
