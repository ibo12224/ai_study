# AWS EC2 (Elastic Compute Cloud)

EC2는 AWS 클라우드에서 제공하는 **확장 가능한 가상 컴퓨팅 리소스(가상 서버)**입니다. 물리적인 서버를 구입할 필요 없이 몇 분 안에 가상 서버를 생성하고 운영체제를 설치하여 사용할 수 있습니다.

## 1. EC2의 주요 특징

* **탄력성(Elasticity):** 사용량에 따라 서버의 성능(CPU, 메모리)을 변경하거나 서버 대수를 자유롭게 조절할 수 있습니다.
* **완전 제어:** 가상 서버에 대한 루트(관리자) 권한을 가지며, 원하는 소프트웨어를 자유롭게 설치할 수 있습니다.
* **다양한 인스턴스 타입:** 연산 위주, 메모리 위주, 그래픽 작업 위주 등 목적에 맞는 최적화된 사양을 선택할 수 있습니다.
* **보안:** Amazon VPC와 결합하여 네트워크 구성, 보안 그룹(방화벽) 설정을 통해 안전한 환경을 구축합니다.

---

## 2. EC2 구성의 4대 요소

| 요소 | 설명 | 비유 |
| --- | --- | --- |
| **AMI** (Amazon Machine Image) | 운영체제(OS), 애플리케이션 등이 포함된 서버 이미지 | 설치용 윈도우/리눅스 USB |
| **인스턴스 유형** (Instance Type) | 서버의 하드웨어 사양 (CPU, RAM 등) | 컴퓨터 본체 사양 (i7, 16GB RAM 등) |
| **EBS** (Elastic Block Store) | 데이터를 저장하는 영구적인 가상 디스크 | 컴퓨터의 SSD / HDD |
| **보안 그룹** (Security Group) | 인바운드/아웃바운드 트래픽을 제어하는 가상 방화벽 | 건물의 보안 검색대 |

---

## 3. EC2 인스턴스 구매 옵션 (비용 절감 전략)

강의 시 수강생들에게 비용 최적화 관점에서 설명하기 좋은 내용입니다.

1. **온디맨드 (On-Demand):** 실행한 초당 또는 시간당 비용 지불. 약정 없이 가장 유연하게 사용.
2. **예약 인스턴스 (Reserved Instances):** 1년 또는 3년 약정을 통해 온디맨드 대비 최대 **75%** 저렴하게 이용.
3. **스팟 인스턴스 (Spot Instances):** AWS의 남는 자원을 경매 방식으로 이용. 최대 **90%** 저렴하지만 AWS에 의해 회수될 수 있음 (중단 가능 업무에 적합).

---

## 4. EC2의 생명 주기 (Lifecycle)

1. **Pending:** 인스턴스 생성 중
2. **Running:** 서버 가동 중 (비용 발생)
3. **Stopping/Stopped:** 서버 종료 중/종료됨 (인스턴스 비용 미발생, EBS 비용 발생)
4. **Terminated:** 서버 영구 삭제 (복구 불가, 모든 비용 중단)

# AWS EBS (Elastic Block Store)

**EBS**는 AWS 클라우드의 EC2 인스턴스에서 사용할 수 있는 **고성능 블록 스토리지 볼륨**입니다. 쉽게 말해, 클라우드 상에서 사용하는 **가상 하드디스크(SSD/HDD)**라고 이해할 수 있습니다.

## 1. EBS의 주요 특징

* **지속성(Persistence):** 인스턴스를 중단(Stop)하거나 종료(Terminate)해도 데이터를 별도로 보존하도록 설정할 수 있습니다.
* **데이터 복제:** EBS 볼륨은 가용 영역(AZ) 내에서 자동으로 복제되어 하드웨어 오류로부터 데이터를 보호하고 높은 가용성을 제공합니다.
* **스냅샷(Snapshot):** 특정 시점의 볼륨 상태를 사진 찍듯 저장하여 S3에 백업할 수 있습니다. 이를 통해 나중에 새로운 볼륨을 생성하거나 복구할 수 있습니다.
* **탄력성:** 볼륨의 크기나 성능(IOPS), 유형을 서비스 중단 없이 실시간으로 변경할 수 있습니다.

---

## 2. EBS 볼륨 유형 (Volume Types)

사용 목적과 비용에 따라 적절한 유형을 선택할 수 있습니다.

| 유형 | 명칭 | 특징 및 용도 |
| --- | --- | --- |
| **SSD** | **gp3 / gp2** | 범용 SSD. 부팅 볼륨, 개발 환경, 일반적인 데이터베이스에 권장. |
| **SSD** | **io2 / io1** | 고성능 SSD. 지연 시간에 민감한 대규모 데이터베이스나 워크로드용. |
| **HDD** | **st1** | 처리량 최적화 HDD. 빅데이터, 로그 처리 등 대용량 스트리밍 작업용. |
| **HDD** | **sc1** | 콜드 HDD. 접근 빈도가 낮은 대용량 데이터 저장용 (가장 저렴). |

---

## 3. EC2와 EBS의 관계

* **1:1 연결:** 일반적으로 하나의 EBS 볼륨은 하나의 EC2 인스턴스에 마운트되어 사용됩니다. (단, Multi-Attach 기능을 지원하는 특정 볼륨은 예외)
* **네트워크 연결:** 물리적으로 서버 안에 꽂혀 있는 디스크가 아니라, **네트워크(AWS 전용망)**를 통해 인스턴스와 연결됩니다. 따라서 인스턴스와 별개로 수명을 관리할 수 있습니다.

---

## 4. 핵심 관리 포인트 

1. **가용 영역(AZ) 제한:** EBS 볼륨은 생성된 특정 AZ 내에서만 존재합니다. 예를 들어 '서울 리전 AZ-A'에서 만든 EBS는 'AZ-B'에 있는 EC2에 바로 붙일 수 없습니다. (이 경우 스냅샷을 찍어 다른 AZ에서 복원해야 함)
2. **인스턴스 종료 시 삭제 설정:** 기본적으로 루트 볼륨(OS 설치된 볼륨)은 인스턴스 종료 시 함께 삭제되도록 설정되어 있는 경우가 많으므로, 중요 데이터라면 이 설정을 해제하거나 데이터용 볼륨을 따로 분리해야 합니다.
