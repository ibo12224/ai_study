

#  클라우드 컴퓨팅 및 AWS의 이해

## 1. 클라우드 컴퓨팅 (Cloud Computing) 이란?

클라우드 컴퓨팅은 IT 리소스를 인터넷을 통해 **주문형(On-demand)**으로 제공하고, **사용량 기반 요금(Pay-as-you-go)**을 지불하는 방식입니다.

* **클라우드 컴퓨팅**은 **인터넷(Network)** 기술을 이용하여 내-외부 고객들에게 **확장성(Scalable)** 있고 **탄력적(Elastic)**인 IT 서비스가 제공되는 방식입니다.

* **Cloud** is a **Style of Computing** where **Scalable** and **Elastic** IT-related capabilities are provided as a service to customer using **Internet** technologies.

---

###  주요 키워드 설명

* **확장성 (Scalable):** 사용자 수나 데이터 양이 증가함에 따라 자원을 유연하게 늘릴 수 있는 능력입니다.
* **탄력성 (Elastic):** 수요에 따라 IT 리소스를 신속하게 확장하거나 축소할 수 있는 성질입니다.
* **인터넷 기술 기반:** 물리적인 장치 연결 없이 네트워크를 통해 서비스 형태로 제공됩니다.

---

**추가로 이 개념과 연결되는 AWS의 주요 서비스(EC2, S3 등)나 클라우드 서비스 모델(IaaS, PaaS, SaaS)에 대한 상세 설명이 필요하신가요?**

### 핵심 특징

* **주문형 셀프 서비스:** 필요할 때 즉시 서버, 저장소 등을 할당 가능.
* **광범위한 네트워크 접속:** 인터넷만 연결되어 있으면 어디서든 접근 가능.
* **리소스 풀링:** 여러 사용자가 물리적 자원을 공유하여 효율성 극대화.
* **신속한 탄력성:** 트래픽 증가 시 확장(Scale-out) 및 감소 시 축소(Scale-in) 용이.

---

### Five Essential Characteristics

클라우드 컴퓨팅이 갖추어야 할 5가지 핵심 요소입니다.

#### 1) 주문형 셀프서비스 (On-Demand Self-Service)

* 사용자가 별도의 기술 습득 없이 필요할 때 온라인으로 즉시 서비스를 사용합니다.

#### 2) 광대역망 액세스 (Broad Network Access)

* 네트워크를 통해 언제, 어디서나, 어떤 기기로든 컴퓨팅 자원에 접근할 수 있습니다 (Any time, Any place, Any device).
* 가상화기술

#### 3) 자원 공동관리 (Resource Pooling)

* 다중 임대 모델(Multi-Tenancy)을 통해 여러 사용자에게 자원을 효율적으로 할당합니다.

#### 4) 빠른 요구 탄력성 (Rapid Elasticity)

* 비즈니스 상황에 맞춰 컴퓨팅 자원을 유연하게 확장하거나 축소하여 사용할 수 있습니다 (Flexibility, Scalability).

#### 5) 도수제 (Measured Service)

* 서비스를 사용한 만큼만 비용을 지불하는 종량제 방식입니다 (Pay-Per-Use, Pay as you go).

---


## 2. 클라우드 서비스 모델 (Service Models)

클라우드는 제공하는 서비스의 범위에 따라 크게 세 가지로 나뉩니다.

| 모델 | 설명 | 예시 |
| --- | --- | --- |
| **IaaS** (Infrastructure) | 가상화된 인프라(서버, 스토리지)만 제공 | AWS EC2, Azure VM |
| **PaaS** (Platform) | 앱 개발에 필요한 플랫폼(런타임, DB) 제공 | AWS Elastic Beanstalk, Heroku |
| **SaaS** (Software) | 완성된 소프트웨어를 웹을 통해 이용 | Google Drive, Slack, Zoom |

---

## 3. AWS (Amazon Web Services) 소개

AWS는 전 세계에서 가장 폭넓게 채택되고 있는 클라우드 플랫폼으로, 200개 이상의 완벽한 기능의 서비스를 제공합니다.

###  AWS 글로벌 인프라 구조

1. **리전 (Region):** 전 세계에 물리적으로 분리된 지리적 영역 (예: 서울 리전).
2. **가용 영역 (Availability Zone, AZ):** 하나의 리전 내에 위치한 격리된 데이터 센터 군락. (장애 대응을 위해 분리됨)
3. **엣지 로케이션 (Edge Location):** 전 세계 사용자에게 빠르게 콘텐츠를 전달하기 위한 캐싱 지점 (CloudFront).

---

## 4. 주요 AWS 핵심 서비스

강의에서 자주 다루는 4대 핵심 서비스입니다.

### 1) 컴퓨팅: EC2 (Elastic Compute Cloud)

* 클라우드에서 제공하는 가상 서버입니다.
* 원하는 운영체제(OS)와 사양을 선택하여 몇 분 안에 구축 가능합니다.

### 2) 스토리지: S3 (Simple Storage Service)

* 업계 최고의 확장성과 데이터 가용성을 제공하는 객체 스토리지입니다.
* 사진, 영상, 정적 웹사이트 호스팅 등에 주로 사용됩니다.

### 3) 데이터베이스: RDS (Relational Database Service)

* MySQL, PostgreSQL 등 관계형 데이터베이스를 쉽게 설정, 운영 및 확장할 수 있는 서비스입니다.

### 4) 네트워크: VPC (Virtual Private Cloud)

* AWS 내 사용자 전용 가상 네트워크 공간을 정의하여 리소스를 안전하게 격리합니다.

---

## 5. 클라우드 도입의 장점

1. **자본 비용의 가변 비용화:** 데이터 센터 구축에 큰돈을 들이지 않고 사용한 만큼만 지불.
2. **규모의 경제:** 많은 사용자가 사용하여 저렴한 단가 유지.
3. **추측 배제:** 서버 용량을 미리 예측할 필요 없이 필요할 때 확장.
4. **속도 및 민첩성 개선:** 개발팀이 IT 리소스를 확보하는 시간을 단축.
5. **글로벌 배포:** 몇 번의 클릭으로 전 세계에 서비스를 배포.

# 가상화(Virtualization): 클라우드의 핵심 엔진

클라우드 컴퓨팅의 5대 특징을 실질적으로 구현하는 기술은 **가상화**입니다.

## 1. 가상화란?

* 물리적인 하드웨어 자원(CPU, 메모리, 저장장치 등)을 추상화하여, 하나의 물리적 자원을 여러 개의 논리적 자원으로 나누거나, 반대로 여러 개의 물리적 자원을 하나의 논리적 자원으로 통합하는 기술입니다.

## 2. 가상화가 클라우드 특징을 구현하는 방법

* **자원 공동관리 (Resource Pooling):** 하이퍼바이저(Hypervisor)라는 소프트웨어를 통해 물리 서버의 자원을 쪼개어 여러 사용자(Multi-tenancy)에게 나누어 줄 수 있습니다.
* **빠른 요구 탄력성 (Rapid Elasticity):** 새로운 서버가 필요할 때 물리적인 하드웨어를 설치할 필요 없이, 가상 머신(VM)을 소프트웨어적으로 복제하여 즉시 생성할 수 있습니다.
* **도수제 (Measured Service):** 가상화 계층에서 각 사용자별 자원 사용량을 정확히 측정할 수 있어 사용한 만큼만 비용을 청구하는 것이 가능해집니다.

## 3. 가상화의 장점

* **자원 효율성:** 서버 한 대의 성능을 최대한 활용하여 유휴 자원을 줄입니다.
* **격리성:** 한 가상 머신에 문제가 생겨도 다른 가상 머신이나 물리 서버 전체에 영향을 주지 않습니다.
* **유연성:** 특정 하드웨어 종속성에서 벗어나 어디서든 동일한 환경을 구동할 수 있습니다.

# 클라우드 서비스 모델 (IaaS, PaaS, SaaS)

클라우드 서비스 모델(IaaS, PaaS, SaaS)은 **"누가 어디까지 관리하느냐"**의 차이입니다. 

클라우드 서비스는 제공되는 자원의 범위와 사용자의 관리 영역에 따라 크게 세 가지로 구분됩니다.

## 1. 한눈에 비교하기 (Pizza as a Service)

이해를 돕기 위해 피자를 먹는 방식에 비유.

| 구분 | 비유 | 설명 |
| --- | --- | --- |
| **IaaS** | 냉동 피자 구매 | 주방(인프라)은 빌리지만, 요리(운영체제, 데이터)는 직접 함 |
| **PaaS** | 피자 배달 | 요리 기구와 재료(플랫폼)는 제공받고, 먹기(애플리케이션 실행)만 함 |
| **SaaS** | 피자 외식 | 식당에 가서 완성된 피자를 먹기만 함 (모든 것을 제공받음) |

---

## 2. 모델별 상세 설명

### 🟢 IaaS (Infrastructure as a Service)

**"인프라를 빌려드립니다"**

* **개념:** 하드웨어 자원(서버, 네트워크, 스토리지)을 가상화하여 제공하는 서비스입니다.
* **특징:** 사용자가 OS(운영체제), 미들웨어, 런타임 등을 직접 설치하고 관리해야 합니다.
* **장점:** 자유도가 가장 높으며, 기존 온프레미스(자체 서버) 환경을 그대로 옮기기에 좋습니다.
* **예시:** AWS EC2, Google Compute Engine(GCE).

### 🔵 PaaS (Platform as a Service)

**"개발에만 집중하세요"**

* **개념:** 애플리케이션을 개발하고 실행할 수 있는 하드웨어와 소프트웨어 플랫폼을 제공합니다.
* **특징:** OS, 서버 관리, 네트워크 설정 등 복잡한 인프라 관리는 클라우드 기업이 담당하고, 사용자는 **코드**와 **데이터**만 관리합니다.
* **장점:** 인프라 관리 비용이 줄어들어 개발 속도가 매우 빠릅니다.
* **예시:** AWS Elastic Beanstalk, Heroku, Google App Engine.

### 🟣 SaaS (Software as a Service)

**"아이디만 있으면 바로 사용 가능"**

* **개념:** 클라우드 인프라와 플랫폼 위에 구축된 완성된 소프트웨어를 제공합니다.
* **특징:** 설치할 필요 없이 웹 브라우저나 앱을 통해 서비스에 접속하기만 하면 됩니다.
* **장점:** 관리할 것이 전혀 없으며 업데이트도 자동화됩니다.
* **예시:** Gmail, Google Drive, Slack, Salesforce.

---

## 3. 관리 범위 요약 (Responsibility)

* **IaaS:** 인프라(하드웨어)만 클라우드 업체가 관리.
* **PaaS:** 인프라 + 실행 환경(OS 등)까지 클라우드 업체가 관리.
* **SaaS:** 애플리케이션을 포함한 모든 것을 클라우드 업체가 관리.

---

AWS와 더불어 전 세계 및 국내에서 많이 활용되는 **클라우드 서비스 제공업체(CSP: Cloud Service Provider)**들의 종류

# 주요 클라우드 서비스 제공업체 (CSP) 종류

클라우드 시장은 글로벌 기업들이 주도하는 **글로벌 CSP**와 특정 국가의 환경에 최적화된 **로컬 CSP**로 나뉩니다.

## 1. 글로벌 BIG 3 (Global Top Tier)

전 세계 시장 점유율의 대부분을 차지하며, 가장 방대한 인프라와 서비스를 제공합니다.

| 기업명 | 서비스 명칭 | 특징 |
| --- | --- | --- |
| **Amazon** | **AWS** (Amazon Web Services) | 세계 시장 점유율 1위. 서비스 종류가장 많고 생태계가 매우 탄탄함. |
| **Microsoft** | **Azure** (애저) | 윈도우(Windows), 오피스365 등 MS 소프트웨어와의 호환성이 강력하여 기업용 시장에서 강세. |
| **Google** | **GCP** (Google Cloud Platform) | 데이터 분석, AI/머신러닝, 컨테이너 기술(Kubernetes) 분야에서 기술적 우위. |

## 2. 기타 글로벌 CSP

* **Oracle Cloud (OCI):** 오라클 데이터베이스(DB)를 사용하는 기업들에게 최적화된 비용과 성능 제공.
* **IBM Cloud:** 대규모 기업용 메인프레임 및 금융권 특화 하이브리드 클라우드 솔루션 강점.

## 3. 국내 주요 CSP (Local CSP)

국내 공공기관, 금융, 교육기관 등 국내 규제(CSAP 인증 등) 준수가 중요한 분야에서 많이 사용됩니다.

* **Naver Cloud Platform (NCP):** 네이버의 서비스 노하우를 바탕으로 하며, 국내 기업들이 가장 많이 사용하는 국내 클라우드.
* **NHN Cloud:** 게임, 공공, 금융 분야에 강점이 있으며 협업 도구(Dooray!)와의 연동이 좋음.
* **KT Cloud:** 국내 최초 클라우드 서비스 시작, 강력한 자체 네트워크(회선) 인프라 기반 공공 클라우드 1위.
* **Kakao i Cloud:** 카카오톡 기반의 편리한 접근성과 AI 서비스 플랫폼 강조.

##  CSP 선택 시 고려사항 


1. **서비스 다양성:** 가장 많은 기능이 필요한가? → **AWS**
2. **기존 환경 호환성:** 사내에서 MS 소프트웨어를 많이 쓰는가? → **Azure**
3. **특수 목적:** AI나 데이터 분석이 핵심인가? → **GCP**
4. **국내 규제 및 지원:** 공공 사업인가, 한국어 기술 지원이 절실한가? → **Naver/NHN/KT Cloud**

# AWS 글로벌 인프라: 리전(Region)과 가용 영역(AZ)

AWS는 전 세계에 걸쳐 물리적으로 분리된 대규모 데이터 센터 인프라를 운영하고 있습니다.

## 1. 리전 (Region) 이란?

* **정의:** AWS 서비스가 운영되는 전 세계의 **지리적 영역**입니다.
* **특징:**
* 각 리전은 다른 리전과 완전히 격리되어 설계되었습니다.
* 리전 내의 데이터는 사용자가 직접 설정하지 않는 한 다른 리전으로 복제되지 않습니다.
* 예: 서울 리전 (ap-northeast-2), 버지니아 북부 리전 (us-east-1).



## 2. 가용 영역 (Availability Zone, AZ)

* **정의:** 하나의 리전 내에 위치하며, 서로 격리된 하나 이상의 **데이터 센터 군락**입니다.
* **특징:**
* 각 리전은 최소 3개 이상의 AZ로 구성됩니다.
* 각 AZ는 물리적으로 수십 킬로미터 떨어져 있어 화재, 침수 등의 재해로부터 서로 영향을 받지 않도록 설계되었습니다.
* AZ 간에는 초고속 전용 네트워크망으로 연결되어 낮은 지연 시간을 보장합니다.



---

## 3. 리전 선택 시 고려해야 할 4가지 기준

강의 시 수강생들에게 리전 선택의 중요성을 설명할 때 유용한 기준입니다.

1. **지연 시간 (Latency):** 사용자와 물리적으로 가장 가까운 리전을 선택하여 서비스 속도를 높여야 합니다. (한국 사용자 대상 서비스라면 서울 리전 선택)
2. **비용 (Cost):** AWS 서비스의 가격은 리전마다 조금씩 다릅니다. (일반적으로 버지니아 북부 리전이 가장 저렴한 편입니다.)
3. **서비스 가용성 (Service Availability):** 최신 서비스나 특정 기능은 일부 리전에서만 먼저 출시될 수 있습니다.
4. **데이터 준수/법적 요구사항 (Compliance):** 데이터가 특정 국가 내에 위치해야 한다는 법적 요구가 있을 경우 해당 국가의 리전을 선택해야 합니다.

---

##  핵심 요약

> **리전**은 커다란 '도시'와 같고, **가용 영역(AZ)**은 그 도시 안에 서로 떨어져 있는 '건물(데이터 센터)'과 같습니다. 만약 한 건물에 문제가 생기더라도 다른 건물의 서버가 살아있기 때문에 서비스가 중단되지 않는 **고가용성(High Availability)**을 확보할 수 있습니다.
