## 1. AWS S3 (Simple Storage Service)

S3는 데이터를 저장하는 **'온라인 데이터 센터'**라고 생각하면 쉽습니다.

* **용도:** 이미지, 동영상, 로그 파일, 정적 웹사이트 호스팅 등 모든 종류의 파일 저장.
* **특징:**
* **무한한 확장성:** 저장 용량 제한이 없습니다.
* **내구성:** 데이터 유실 확률이 거의 제로에 가깝습니다 (99.999999999\% 내구성).
* **객체 스토리지:** 파일 하나하나를 '객체'로 관리하며, 각각 고유한 URL을 가집니다.

## 2. AWS CloudFront (CDN:Content Delivery Network)

<img src='image/cdn.png' width='500' height='500'>

CloudFront는 전 세계 사용자에게 콘텐츠를 빠르게 전달하는 **'물류 배송망(CDN)'**입니다.

* **용도:** S3에 저장된 파일을 전 세계 거점(Edge Location)에 복사해두고, 사용자와 가장 가까운 곳에서 빠르게 전송.
* **특징:**
* **속도 향상:** 한국에 있는 서버의 파일을 미국 사용자가 다운로드할 때, 미국 내 거점을 통해 전송하므로 훨씬 빠릅니다.
* **캐싱:** 한 번 요청된 파일은 일정 기간 저장(캐싱)되어 원본 서버(S3)의 부하를 줄여줍니다.
* **보안:** HTTPS 보안 연결을 쉽게 적용하고 DDoS 공격을 방어합니다.



---

## 3. 왜 둘을 같이 사용하나?

단순히 S3만 써도 파일 공유는 가능하지만, CloudFront를 앞에 두면 다음과 같은 강력한 장점이 생깁니다.

| 비교 항목 | S3 단독 사용 | S3 + CloudFront 조합 |
| --- | --- | --- |
| **속도** | 서버 물리적 위치에 따라 다름 | **전 세계 어디서든 빠름 (Edge 캐싱)** |
| **비용** | 데이터 전송비가 상대적으로 높음 | **AWS 내부 망 이용으로 전송비 절감 가능** |
| **보안** | S3 버킷을 외부에 공개해야 함 | **S3는 숨기고 CloudFront로만 접속 가능 (OAC)** |
| **도메인** | S3 주소 사용 (긴 URL) | **사용자 정의 도메인 및 SSL(HTTPS) 적용 용이** |

---

###  비유

* **S3:** 물건을 쌓아두는 **'중앙 대형 창고'**
* **CloudFront:** 고객 집 근처에 있는 **'편의점'**

고객이 물건을 살 때마다 멀리 있는 중앙 창고까지 갈 필요 없이, 가까운 편의점에 재고가 있다면 바로 가져오는 원리와 같습니다.

---

 **Amazon S3의 버킷 이름은 전 세계(Global)에서 유일해야 합니다.**

단순히 내 AWS 계정 내에서만 겹치지 않으면 되는 것이 아니라, **전 세계 모든 AWS 사용자를 통틀어 단 하나뿐인 이름**이어야 합니다.

---

### 1. 왜 이름이 유일해야 하나요?

S3 버킷은 생성 시 고유한 **URL(Endpoint)** 주소를 가지게 되기 때문입니다.

* **기본 형식:** `https://[버킷이름].s3.[리전].amazonaws.com`
* 인터넷상에서 이 주소는 특정 데이터가 저장된 위치를 가리키는 고유한 식별자 역할을 합니다. 만약 이름이 중복된다면, 주소가 겹쳐서 데이터를 찾아갈 수 없게 됩니다.

### 2. 버킷 이름을 지을 때 규칙 (Naming Convention)

전 세계 유일해야 한다는 점 외에도 몇 가지 중요한 규칙이 있습니다:

* **길이:** 3자에서 63자 사이여야 합니다.
* **구성:** 소문자, 숫자, 점(`.`), 하이픈(`-`)만 사용할 수 있습니다. (대문자 사용 불가)
* **시작과 끝:** 문자와 숫자로 시작하고 끝나야 합니다.
* **IP 형식 금지:** `192.168.0.1` 같은 IP 주소 형식은 사용할 수 없습니다.

### 3. 실무 팁: 이름 중복 피하기

유명한 단어나 서비스 이름(예: `test`, `logs`, `backup`)은 이미 누군가 선점했을 가능성이 매우 높습니다. 그래서 보통 다음과 같은 패턴을 사용합니다:

* **[회사명]-[프로젝트명]-[환경]-[구분]**
* 예: `mycompany-flask-db-prod-01`


* **[도메인 거꾸로]-[구분]**
* 예: `com-example-static-files`
