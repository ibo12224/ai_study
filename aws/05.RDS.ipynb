AWS의 **RDS(Relational Database Service)**는 

클라우드에서 관계형 데이터베이스를 간편하게 설정, 운영 및 확장할 수 있도록 지원하는 관리형 서비스

쉽게 말해, 하드웨어 프로비저닝, 데이터베이스 설정, 패치 및 백업과 같은 **복잡하고 번거로운 관리 작업을 AWS가 대신 처리**해주는 서비스

---

## 1. RDS의 핵심 특징

* **관리형 서비스:** OS 패치, 백업, 복구 등을 자동화하여 사용자는 데이터베이스 설계와 쿼리 최적화에만 집중할 수 있습니다.
* **확장성:** 클릭 몇 번으로 스토리지 용량을 늘리거나, 컴퓨팅 리소스(CPU, RAM)를 업그레이드할 수 있습니다.
* **고가용성 (Multi-AZ):** 여러 가용 영역에 복제본을 두어 데이터 센터 장애 발생 시 자동으로 전환(Failover)합니다.
* **보안:** AWS VPC 내에서 실행되어 네트워크를 격리할 수 있으며, 데이터 암호화(저장 시 및 전송 시)를 지원합니다.

---

## 2. 지원하는 데이터베이스 엔진

RDS는 현재 총 6가지의 주요 데이터베이스 엔진을 지원합니다.

* **상용 엔진:** Oracle, Microsoft SQL Server
* **오픈 소스 엔진:** MySQL, PostgreSQL, MariaDB
* **클라우드 최적화 엔진:** **Amazon Aurora** (MySQL 및 PostgreSQL과 호환되며 더 높은 성능과 가용성을 제공)

---

## 3. 주요 기능 및 아키텍처

### **Multi-AZ (다중 가용 영역)**

* **목적:** 고가용성 및 재해 복구.
* **원리:** 기본(Primary) DB의 데이터를 다른 가용 영역에 있는 예비(Standby) DB로 실시간 동기화 복제합니다. 장애 발생 시 자동으로 예비 DB가 기본 DB 역할을 승계합니다.

### **Read Replica (읽기 전용 복제본)**

* **목적:** 읽기 성능 향상.
* **원리:** 데이터베이스의 읽기 작업이 많을 경우, 복제본을 생성하여 읽기 트래픽을 분산시킵니다. 비동기식 복제 방식을 사용합니다.

---

## 4. RDS vs EC2 설치형 DB 비교

| 구분 | RDS (관리형) | EC2 직접 설치 (자격형) |
| --- | --- | --- |
| **관리 주체** | AWS (자동 패치, 백업 등) | 사용자 직접 관리 |
| **고가용성** | 클릭 한 번으로 설정 (Multi-AZ) | 직접 클러스터링 구성 필요 |
| **확장성** | 간편한 스케일링 | 복잡한 마이그레이션 작업 필요 |
| **제어권** | OS 및 DB 커널 접근 제한 | OS 및 모든 설정 제어 가능 |

---

## 5. 요금 체계

RDS 비용은 주로 다음 요소들에 의해 결정됩니다.

1. **DB 인스턴스 시간:** 선택한 인스턴스 사양 및 가동 시간.
2. **스토리지:** 할당된 스토리지 용량(GB) 및 I/O 요청 횟수.
3. **데이터 전송:** 아웃바운드 데이터 전송량.
4. **추가 옵션:** 백업 저장 공간 초과분, Multi-AZ 구성 여부 등.

---

AWS RDS(Relational Database Service)는 사용자의 요구사항과 기존 환경에 맞춰 선택할 수 있도록 총 **6가지 데이터베이스 엔진**을 제공합니다.

각 엔진은 성능, 비용, 기능 면에서 차이가 있으므로 서비스의 성격에 맞춰 선택하는 것이 중요합니다.

---

## 1. AWS RDS 엔진 종류 및 특징

| 구분 | 엔진 이름 | 주요 특징 및 용도 |
| --- | --- | --- |
| **AWS 최적화** | **Amazon Aurora** | AWS가 클라우드에 최적화하여 재설계한 DB. MySQL/PostgreSQL과 호환되며, 일반 RDS보다 성능이 3~5배 뛰어나고 고가용성이 매우 높음. |
| **오픈 소스** | **MySQL** | 전 세계에서 가장 인기 있는 오픈 소스 DB. 웹 애플리케이션, 전자상거래 등에 널리 사용되며 커뮤니티 지원이 강력함. |
| **오픈 소스** | **PostgreSQL** | 복잡한 쿼리와 데이터 분석에 강점이 있는 객체 관계형 DB. 지리 정보(PostGIS)나 대용량 데이터 처리에 적합함. |
| **오픈 소스** | **MariaDB** | MySQL을 개발한 팀이 만든 엔진으로 MySQL과 높은 호환성을 유지하며 성능과 보안이 개선됨. |
| **상용 DB** | **Oracle** | 대기업 환경에서 주로 사용되는 강력한 기능의 상용 DB. 기존 온프레미스 Oracle 환경을 클라우드로 이전할 때 선택함. |
| **상용 DB** | **SQL Server** | Microsoft의 데이터베이스 엔진. .NET 환경이나 윈도우 기반 애플리케이션과 최상의 궁합을 보여줌. |

---

## 2. 어떤 것을 선택해야 할까요?

* **성능과 안정성이 최우선이라면?** → **Amazon Aurora**
* 데이터가 자동으로 3개의 가용 영역(AZ)에 6개나 복제되어 저장되므로 매우 안전합니다.
* 트래픽이 급증하는 대규모 서비스에 권장됩니다.


* **비용 효율적인 시작을 원한다면?** → **MySQL** 또는 **PostgreSQL**
* 라이선스 비용이 없고 설정이 간편하여 스타트업이나 개인 프로젝트에 적합합니다.


* **기존 기업용 솔루션을 그대로 옮기고 싶다면?** → **Oracle** 또는 **SQL Server**
* 이미 사용 중인 상용 라이선스(BYOL)를 활용하거나 AWS를 통해 라이선스를 포함하여 구매할 수 있습니다.



---

## 3. RDS의 주요 배포 옵션

엔진을 선택한 후에는 인스턴스의 가용성을 결정해야 합니다.

* **단일 인스턴스 (Single-AZ):** 비용이 가장 저렴하지만, 장애 발생 시 복구에 시간이 걸립니다. (테스트/개발용)
* **다중 AZ (Multi-AZ):** 다른 지역에 복제본을 두어 장애 발생 시 자동으로 전환됩니다. (운영/서비스용)
* **읽기 전용 복제본 (Read Replica):** 읽기 트래픽이 많을 때 성능을 분산시키기 위해 생성합니다.

> 2025년 현재, 소규모 프로젝트는 **PostgreSQL**이 기능적 우위로 인해 인기가 높으며, 중대형 이상의 엔터프라이즈급 서비스는 **Amazon Aurora**로 전환하는 추세입니다.

## Relational Database Service (RDS)

### **RDS**

* 백업 및 소프트웨어 패치 등 서비스 유지에 필요한 기능을 AWS가 관리
* SaaS이므로 서비스 최적화에 집중할 수 있음
* **SSH를 통한 접속이 불가하며**, EC2를 통한 접속 혹은 Workbench와 같은 접속 프로그램을 통해서만 가능
* 설정을 변경할 수 있는 파라미터 그룹이 존재하지만 변경 불가능한 설정도 존재

### **DB in EC2**

* EC2 위에 데이터베이스를 직접 올리는 만큼 설정을 마음대로 변경할 수 있고 커스터마이징 또한 가능
* RDS와는 반대로 백업과 패치 등 관리를 직접해야 함
* EC2에 설치하는 것이기에 SSH 접속 가능

---

AWS RDS는 **독립적인 관리형 데이터베이스 서비스**입니다. 반드시 EC2와 연동해야만 사용할 수 있는 것은 아니지만, 실제 서비스 환경에서는 보안과 성능을 위해 EC2와 같은 VPC(가상 네트워크) 내에서 서로 연동하여 사용하는 것이 일반적입니다.



---

### 1. RDS는 "독립적인 서비스"입니다

RDS는 AWS가 제공하는 가상화된 데이터베이스 인스턴스입니다.

* **컴퓨터 한 대를 통째로 빌려주는 것과 유사:** RDS 인스턴스를 생성하면 CPU, 메모리, 스토리지가 할당된 독립적인 환경이 구성됩니다.
* **EC2 없이도 사용 가능:** 본인의 로컬 PC(내 컴퓨터), 외부 서버, 또는 다른 클라우드 서비스에서도 인터넷을 통해 RDS에 직접 접속하여 데이터를 읽고 쓸 수 있습니다.

### 2. 왜 보통 EC2와 연동하나요?

대부분의 가이드는 EC2와의 연동을 강조하는데, 여기에는 몇 가지 중요한 이유가 있습니다.

* **네트워크 보안:** 데이터베이스를 인터넷에 직접 노출(Public)하는 것은 보안상 위험합니다. 보통 데이터베이스는 외부 접근을 차단하고, EC2 서버만 접근할 수 있게 비공개 네트워크(Private VPC)에 가둡니다.
* **성능(낮은 지연 시간):** EC2와 RDS를 같은 가용 영역(AZ)에 두면 네트워크 전송 속도가 매우 빨라져 서비스 성능이 향상됩니다.
* **자동화된 연결:** AWS는 EC2와 RDS를 손쉽게 연결할 수 있도록 보안 그룹(Security Group) 설정을 자동화해주는 기능을 제공합니다.

### 3. "독립 설치(EC2 상의 DB)" vs "RDS" 비교

이미지에서 보신 내용처럼, DB를 어디에 두느냐에 따라 관리 방식이 완전히 달라집니다.

| 구분 | RDS (관리형 서비스) | DB in EC2 (직접 설치) |
| --- | --- | --- |
| **운영체제 접근** | **SSH 접속 불가** (AWS가 관리) | **SSH 접속 가능** (직접 설치 및 제어) |
| **관리 부담** | 백업, 패치, 복구 자동화 | 사용자가 수동으로 모든 관리 수행 |
| **확장성** | 클릭 몇 번으로 용량/성능 확장 | 복잡한 데이터 이전 작업 필요 |

---

**결론적으로:** RDS는 독립적으로 작동하는 서비스이지만, 보안과 속도라는 두 마리 토끼를 잡기 위해 **웹 서버(EC2)와 한 세트처럼 연동해서 사용하는 것이 정석(Best Practice)**입니다.
