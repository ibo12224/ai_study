**MFA(Multi-Factor Authentication, 다요소 인증)** 는 ID와 비밀번호라는 하나의 요소 외에 **추가적인 인증 단계**를 하나 더 거치게 하는 보안 방식입니다. 쉽게 말해 "비밀번호를 알아도 내 핸드폰이 없으면 로그인할 수 없게" 만드는 장치입니다.

AWS 환경에서는 특히 **Root 계정**과 **관리자(Administrator) 계정**에 MFA 설정이 거의 필수입니다.

---

### 1. 인증의 3가지 요소 (Factors)

MFA는 보통 아래 3가지 중 2가지 이상을 조합합니다.

* **지식(Something you know):** 비밀번호, PIN 번호 등
* **소유(Something you have):** 스마트폰(OTP 앱), 보안 키(YubiKey), SMS 인증 코드 등
* **생체(Something you are):** 지문, 안면 인식(FaceID) 등

---

### 2. AWS에서 주로 사용하는 MFA 유형

AWS에서는 주로 다음과 같은 방식을 지원합니다.

| 유형 | 설명 | 장점 |
| --- | --- | --- |
| **가상 MFA 앱** | **Google Authenticator**나 **Authy** 같은 앱을 스마트폰에 설치해 6자리 번호 생성 | 무료이며 설정이 매우 간편함 (가장 많이 쓰임) |
| **FIDO 보안 키** | YubiKey 같은 물리적인 USB 형태의 보안 장치 | 가장 강력한 보안 수준. 해킹/피싱에 매우 강함 |
| **하드웨어 OTP** | 번호가 생성되는 전용 카드나 키홀더형 장치 | 스마트폰 사용이 불가능한 특수 환경에서 사용 |

---

### 3. 왜 MFA를 꼭 써야 할까요?

1. **비밀번호 유출 대비:** 비밀번호가 노출되거나 해킹당해도, 공격자는 내 스마트폰(MFA 앱)이 없으므로 로그인을 완료할 수 없습니다.
2. **클라우드 비용 폭탄 방지:** 관리자 계정이 탈취되면 해커들이 코인 채굴 등을 위해 최고 사양 EC2 수십 대를 생성할 수 있습니다. MFA는 이를 막는 가장 저렴하고 확실한 방어선입니다.
3. **컴플라이언스 준수:** 많은 기업 보안 규정상 관리자 계정의 MFA는 의무 사항입니다.

---

### 4. AWS에서 MFA 설정하는 방법 (관리 콘솔 기준)

1. AWS 상단 우측의 **사용자 이름** 클릭 -> **내 보안 자격 증명(My Security Credentials)** 클릭.
2. **멀티 팩터 인증(MFA)** 섹션에서 **MFA 디바이스 할당** 클릭.
3. **디바이스 이름** 입력 (예: MyPhone) 및 **가상 MFA 앱** 선택.
4. 화면에 나오는 **QR 코드**를 스마트폰의 Authenticator 앱으로 스캔.
5. 앱에 표시되는 **연속된 두 개의 OTP 번호**를 입력하면 설정 완료.

# AWS IAM (Identity and Access Management)

AWS IAM은 AWS 리소스에 대한 액세스를 안전하게 제어할 수 있게 해주는 서비스입니다. "누가(인증)" 무엇을 "할 수 있는지(권한 부여)"를 관리하는 보안의 핵심입니다.

## 1. IAM의 핵심 4요소

| 요소 | 설명 | 비유 |
| --- | --- | --- |
| **사용자 (User)** | 실제 사람이나 애플리케이션을 의미하는 개별 계정 | 회사원 개인 ID |
| **그룹 (Group)** | 사용자들의 집합. 그룹에 권한을 부여하면 소속 사용자 모두에게 적용 | 개발팀, 인사팀 |
| **역할 (Role)** | 특정 시간 동안 리소스에 부여되는 임시 권한 (ID가 없음) | 일일 방문객 출입증 |
| **정책 (Policy)** | 어떤 리소스에 어떤 작업을 허용/거부할지 정의한 문서(JSON 형식) | 출입 통제 지침서 |

---

## 2. IAM 정책(Policy) 예시

IAM 정책은 다음과 같은 **JSON 형식**의 문서로 구성됩니다.

```json
{
    "Version": "2012-10-17",
    "Statement": [
        {
            "Effect": "Allow",
            "Action": [
                "ec2:*",
                "s3:*",
                "rds:*",
                "cloudwatch:*",
                "autoscaling:*",
                "iam:PassRole"
            ],
            "Resource": "*"
        }
    ]
}

---

EC2와 RDS 같은 주요 서비스를 '제대로' 운영하기 위해 반드시 필요한 보조 권한들

1) cloudwatch:(감시와 알람)

CloudWatch는 AWS의 **모니터링 서비스**입니다.

역할: 서버(EC2)가 느려지는지, 데이터베이스(RDS)의 CPU 사용량이 높은지 등을 그래프로 보여주고 기록합니다.

왜 필요한가: 이 권한이 없으면 서버를 만들 수는 있지만, 그 서버가 현재 정상인지 죽었는지 콘솔에서 그래프를 볼 수 없습니다. 또한, 특정 수치가 넘으면 나에게 이메일을 보내는 '경보(Alarm)' 설정도 이 권한이 있어야 가능합니다.

2) autoscaling: (자동 확장)

Auto Scaling은 부하에 따라 **서버 개수를 자동으로 조절**하는 기능입니다.

**역할:** 쇼핑몰에 갑자기 사람이 몰리면 EC2 인스턴스를 2개에서 10개로 늘리고, 밤에 사람이 줄어들면 다시 2개로 줄이는 작업을 관리합니다.
* **왜 필요한가:** 보통 EC2를 사용할 때 단일 서버보다는 Auto Scaling 그룹으로 묶어서 관리하는 것이 표준(Best Practice)입니다. 이 권한이 있어야만 트래픽 변화에 유연하게 대응하는 인프라를 구축할 수 있습니다.

3) iam:PassRole (권한 위임의 핵심)

* **역할:** 내가 만든 **IAM 역할(Role)** 을 AWS 서비스(예: EC2)에게 **"이거 네가 가져가서 써"**라고 건네주는(Pass) 행위입니다.
* **왜 필요한가:** 
1. 예를 들어, EC2 서버가 S3에 있는 파일을 읽어야 한다고 가정해 봅시다.
2. 이때 서버 안에 로그인 아이디/비번을 직접 저장하는 건 위험합니다.
3. 대신 'S3 읽기 권한'이 담긴 **Role**을 만들어 EC2에 입혀줍니다.
4. 이때 사용자가 EC2에게 그 Role을 전달할 수 있는 권한이 바로 `iam:PassRole`입니다.
* **이게 없으면:** EC2를 생성할 때 "IAM 역할을 선택할 권한이 없습니다"라는 오류와 함께 생성이 중단됨.

---

### 요약 테이블

| 항목 | 한 줄 설명 | 없으면 발생하는 문제 |
| --- | --- | --- |
| **CloudWatch** | 시스템 상태 모니터링 | 내 서버 상태(CPU, 메모리 등)를 볼 수 없음 |
| **Auto Scaling** | 서버 개수 자동 조절 | 트래픽 증가 시 서버 자동 확장이 불가능함 |
| **iam:PassRole** | 서비스에 권한 배정 | EC2나 RDS를 생성할 때 역할(Role) 연결 불가 |

**정리:** `ec2:*`, `s3:*`가 **"물건을 만들고 사용하는 권한"**이라면, 위 세 가지는 **"물건이 잘 돌아가게 관리하고 다른 물건과 연결하는 권한"
```

## 3. IAM의 주요 특징

* **공통된 보안 모범 사례 - 최소 권한 원칙 (Least Privilege):** 사용자에게는 업무 수행에 필요한 **최소한의 권한**만 부여해야 합니다.
* **MFA (Multi-Factor Authentication) 지원:** 보안 강화를 위해 비밀번호 외에 추가적인 인증 수단(OTP 등)을 사용할 수 있습니다.
* **자격 증명 연동 (Federation):** 회사 내부 로그인 시스템(Active Directory 등)을 AWS 계정과 연동할 수 있습니다.
* **임시 보안 자격 증명:** 역할(Role) 기능을 통해 영구적인 ID/PW 없이 안전하게 리소스에 접근합니다.

##  실무 보안 가이드


1. **루트(Root) 계정 사용 자제:** 루트 계정은 MFA를 설정한 후 잠가두고, 일상 업무는 별도의 IAM 사용자를 생성해 사용하세요.
2. **그룹 권한 부여:** 사용자 개개인에게 권한을 주기보다, 그룹에 정책을 연결하고 사용자를 그룹에 추가하는 방식이 관리에 효율적입니다.
3. **강력한 암호 정책 설정:** 암호 복잡성, 교체 주기 등을 설정하여 보안 사고를 예방합니다.

## 3. AWS 공동 책임 모델 (Shared Responsibility Model)

AWS 보안의 핵심 원칙으로, AWS와 고객이 보안 책임을 나누어 가집니다.

### 🔒 AWS 공동 책임 모델 개요

* 보안과 규정 준수는 AWS와 고객의 공동 책임입니다.

### 👤 고객의 책임 (Security **IN** the cloud)

* 고객은 AWS 클라우드 내에서 생성하고 배치하는 모든 것의 보안을 책임집니다.
* **주요 관리 대상:** 콘텐츠, 사용하는 AWS 서비스, 해당 콘텐츠에 액세스할 수 있는 사용자 관리 등.

### ☁️ AWS의 책임 (Security **OF** the cloud)

* AWS는 인프라의 모든 계층에서 구성 요소를 운영, 관리 및 제어합니다.
* **주요 관리 대상:** 하드웨어, 소프트웨어, 스토리지, 네트워킹 및 시설, 글로벌 인프라 등 서비스 실행 인프라 보호.

## 4. AWS Identity and Access Management (IAM)

* AWS 리소스에 대한 액세스를 안전하게 제어하고 공동 책임 모델 하에서 **'고객'**측의 보안 책임을 수행하기 위한 핵심 도구입니다.
